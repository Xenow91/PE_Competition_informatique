# 1 Structures de données fondamentales

## 1.1 Structures linéaires (piles, files, listes)
Les structures linéaires sont des structures de données dans lesquelles les éléments sont organisés de manière séquentielle, selon un ordre bien défini.

### 1.1.1 Piles
Une pile est une structure de données linéaire qui suit le principe LIFO (Last In, First
Out)

```python
class Piles :
    def __init__ ( self ) :
        self . data = []
    def push ( self , x ) :
        self . data . append ( x )

    def pop ( self ) :
        if self . data :
            return self . data . pop ()
        return None

    def is_empty ( self ) :
        return len ( self . data ) == 0
```

### 1.1.2 Files
Une file est une structure de données linéaire qui suit le principe FIFO (First In, First
Out)

```python
class File :
    def __init__ ( self ) :
        self . data = []
        
    def enqueue ( self , x ):
        self . data . append ( x )

    def dequeue ( self ) :
        if self . data :
            return self . data . pop (0)
        return None

    def is_empty ( self ):
        return len ( self . data ) == 0
```

éléments.

## 1.2 Arbres (BST, segment trees, Fenwick trees/BIT)
Les arbres sont des structures de données hiérarchiques, organisées sous forme de
nœuds reliés entre eux, permettant une organisation non linéaire des données.

### 1.2.1 BST (ou ABR : arbre binaire de recherche )
Un arbre de recherche binaire est une structure de données arborescente dans laquelle,
pour chaque nœud, les éléments du sous-arbre gauche sont inférieurs à la valeur du nœud,
et ceux du sous-arbre droit sont supérieurs.

```python
class BST:
    def __init__(self):
        self.data = None

    def add(self, x):
        if self.data == None :
            self.data = x
            self.left = BST()
            self.right = BST()
        elif x < self.data :
            if self.left.data == None :
                self.left.add(x)
            else : 
                self.left.add(x)
        else :
            if self.right.data == None :
                bst = BST()
                bst.add(x)
                self.right = bst
            else : 
                self.right.add(x)
                
    def search (self,x) :
        if self.data == None or self == None:
            return False
        elif self.data == x :
            return True
        elif x < self.data :
            return self.left.search(x)
        else : 
            return self.right.search(x)
            
    def delete (self,x) :
        if self.data == None :
            return
        elif self.data == x :
            if self.left.data == None and self.right.data == None :
                self.data = None
            elif self.left.data == None :
                self.data = self.right.data
                self.left = self.right.left
                self.right = self.right.right
            elif self.right.data == None :
                self.data = self.left.data
                self.right = self.left.right
                self.left = self.left.left
            else :
                min_right = self.right.min()
                self.data = min_right
                self.right.delete(min_right)
        elif x < self.data :
            return self.left.delete(x)
        else : 
            return self.right.delete(x)
      
    def min (self) :
        if self.left.data == None:
            return self.data
        else :
            return self.left.min()
            
    def max (self) :
        if self.right.data == None:
            return self.data
        else :
            return self.right.max()

    def is_empty(self):
        return self.data == None
```

## 1.3 Files de priorité et heaps
Une file de priorité est une structure de données permettant d’accéder rapidement à
l’élément de priorité minimale ou maximale, indépendamment de l’ordre d’insertion. Elle
est généralement implémentée à l’aide d’un tas, notamment un tas binaire, qui garantit
des opérations d’insertion et d’extraction en temps logarithmique.

```python
class priorityQueue:
    def __init__(self):
        self.heap = []

    def parent(self, index):
        return (index - 1) // 2
    
    def leftChild(self, index):
        return 2 * index + 1
    
    def rightChild(self, index):
        return 2 * index + 2
    
    def isEmpty(self):
        return len(self.heap) == 0

    def insert(self, item): # item is a tuple (priority, value)
        self.heap.append(item)
        self.heapifyUp(len(self.heap) - 1)

    def heapifyUp(self, index):
        parentIndex = self.parent(index)
        if index > 0 and self.heap[parentIndex][0] > self.heap[index][0]:
            self.heap[parentIndex], self.heap[index] = self.heap[index], self.heap[parentIndex]
            self.heapifyUp(parentIndex)
    
    def heapifyDown(self, index):
        leftChildIndex = self.leftChild(index)
        rightChildIndex = self.rightChild(index)
        smallestIndex = index

        if leftChildIndex < len(self.heap) and self.heap[leftChildIndex][0] < self.heap[smallestIndex][0]:
            smallestIndex = leftChildIndex

        if rightChildIndex < len(self.heap) and self.heap[rightChildIndex][0] < self.heap[smallestIndex][0]:
            smallestIndex = rightChildIndex

        if smallestIndex != index:
            self.heap[index], self.heap[smallestIndex] = self.heap[smallestIndex], self.heap[index]
            self.heapifyDown(smallestIndex)

    def pop(self):
        if self.isEmpty():
            raise IndexError("La file de priorité est vide")
        minItem = self.heap[0]
        lastItem = self.heap.pop()
        if not self.isEmpty():
            self.heap[0] = lastItem
            self.heapifyDown(0)
        return minItem 
```

Les algorithmes de parcours permettent d’explorer systématiquement l’ensemble des
sommets d’un graphe, soit en largeur (BFS), soit en profondeur (DFS).

```python
# Parcours en largeur (Breadth-First Search) 
def bfs(graph, start):  # Le graphe est represente sous forme d'une liste d'adjacence
    visited = [False for i in range(len(graph))] 
    queue = []  

    visited[start] = True  #
    queue.append(start)  

    while queue:  # Tant que la file d'attente n'est pas vide
        vertex = queue.pop(0)  
        print(vertex) 

        for neighbor in graph[vertex]: 
            if not visited[neighbor]:  
                visited[neighbor] = True
                queue.append(neighbor) 

# Parcours en profondeur (Depth-First Search)
def dfs(graph, start):
    visited = [False for i in range(len(graph))] 
    stack = []  

    visited[start] = True  
    stack.append(start)  

    while stack:  # Tant que la pile n'est pas vide
        vertex = stack.pop()  
        print(vertex) 

        for neighbor in graph[vertex]: 
            if not visited[neighbor]:  
                visited[neighbor] = True
                stack.append(neighbor)
                
```

Diviser pour régner est un paradigme de programmation : les algorithmes séparent le problème en plusieurs sous-problèmes semblables au problème initial mais de taille moindre, résolvent les sous-problèmes de façon récursive, puis combinent toutes les solutions pour produire la solution du problème original.

Le paradigme diviser-pour-régner implique trois étapes à chaque niveau de la récursivité :
- **Diviser** le problème en un certain nombre de sous-problèmes.
- **Régner** sur les sous-problèmes en les résolvant de manière récursive. Si la taille d’un sous-problème est suffisamment réduite, on peut toutefois le résoudre directement.
- **Combiner** les solutions des sous-problèmes pour produire la solution du problème originel.

## 4.1 Algorithmes de recherche de motifs

Les algorithmes de recherche de motifs permettent de localiser les occurrences d’une sous-chaîne (appelée *motif*) dans un texte.

Une approche naïve consiste à comparer le motif avec toutes les positions possibles dans le texte.

Cependant, cette méthode a une complexité \(O(nm)\), où :

- \(n\) est la longueur du texte  
- \(m\) est la longueur du motif  

Lorsque les chaînes sont très longues, cette complexité devient inefficace.  
Des algorithmes plus avancés, comme KMP et Z-algorithm permettent de résoudre ce problème en temps linéaire ou quasi linéaire.

```python
def naive_pattern_search(text, pattern):

    # Longueur du texte
    n = len(text)
    # Longueur du motif à rechercher
    m = len(pattern)

    # Liste pour stocker les positions où le motif apparaît
    positions = []

    # On parcourt toutes les positions possibles du texte
    # où le motif peut commencer
    for i in range(n - m + 1):

        # On compare la sous-chaîne du texte avec le motif
        if text[i:i+m] == pattern:
            positions.append(i)

    return positions
# Complexité : O(nm)
```

## 4.2 Hachage de chaînes
Le hachage de chaînes consiste à associer une valeur numérique à une chaîne de caractères afin de comparer efficacement des sous-chaînes ou de détecter des égalités.

Cette technique est particulièrement utile pour :

- comparer rapidement deux sous-chaînes  
- détecter des sous-chaînes identiques  
- résoudre des problèmes comme la recherche de motifs ou la comparaison de sous-chaînes  

Le principe repose généralement sur un polynomial rolling hash, qui transforme une chaîne en une valeur numérique calculée à partir des caractères et de puissances d’une base.

Le hachage de chaînes nécessite un prétraitement en \(O(n)\) et permet ensuite de comparer deux sous-chaînes en \(O(1)\).

```python
class StringHash:
    def __init__(self, s):
        self.s = s
        self.n = len(s)
        self.p = 31
        self.mod = 10**9 + 7
        
        self.p_pow = [1] * (self.n + 1)
        self.h = [0] * (self.n + 1)
        
        for i in range(1, self.n + 1):
            self.p_pow[i] = (self.p_pow[i-1] * self.p) % self.mod
        
        for i in range(self.n):
            self.h[i+1] = (self.h[i] + (ord(s[i]) - ord('a') + 1) * self.p_pow[i]) % self.mod
    
    def get_hash(self, l, r):
        hash_value = (self.h[r+1] - self.h[l] + self.mod) % self.mod
        return (hash_value * pow(self.p, self.n-l-1, self.mod)) % self.mod
```

## 4.3 KMP
L’algorithme *KMP (Knuth–Morris–Pratt)* permet de rechercher un motif dans un texte en temps linéaire.

L’idée principale est d’éviter de recommencer les comparaisons depuis le début lorsqu’un échec se produit.  
Pour cela, on exploite les préfixes qui sont aussi suffixes grâce à un tableau appelé fonction préfixe dont la fonctionnement est suivant.

Pour chaque position i, π[i] représente la longueur du plus long préfixe propre de la chaîne qui est aussi suffixe du préfixe se terminant en i.

L’algorithme KMP est utile lorsque :

- on doit trouver toutes les occurrences d’un motif dans un texte  
- les chaînes sont très longues  
- on veut éviter la complexité \(O(nm)\) de la méthode naïve  

Exemples de problèmes :

- recherche de motifs  
- détection de répétitions dans une chaîne  
- construction d’automates de chaînes 
La complexité de l’algorithme est \[O(n + m)\] où \(n\) est la longueur du texte et \(m\) est la longueur du motif  

```python
def prefix_function(s):
    n = len(s)
    pi = [0]*n
    
    for i in range(1,n):
        j = pi[i-1]
        
        while j > 0 and s[i] != s[j]:
            j = pi[j-1]
        
        if s[i] == s[j]:
            j += 1
        
        pi[i] = j
        
    return pi


def kmp_search(text, pattern):
    s = pattern + "#" + text
    pi = prefix_function(s)
    
    result = []
    m = len(pattern)
    
    for i in range(len(s)):
        if pi[i] == m:
            result.append(i - 2*m)
    
    return result
```

## 4.4 Z-algorithm
L’algorithme Z calcule, pour chaque position \(i\), la longueur du plus long préfixe de la chaîne qui commence à la position \(i\).
On construit ainsi un tableau \(Z\) tel que :
\[
Z[i] = \text{longueur du plus long préfixe de la chaîne commençant en } i
\]

Cet algorithme permet de résoudre efficacement plusieurs problèmes liés aux chaînes de caractères, notamment la recherche de motifs.
L’algorithme Z est utilisé pour :

- la recherche de motifs dans un texte  
- la détection de répétitions dans une chaîne  
- le calcul de bordures de chaînes  
- l’analyse de structures répétitives  
- La complexité de l’algorithme est :
\[
O(n)
\]
où \(n\) est la longueur de la chaîne.

```python
def z_algorithm(s):
    n = len(s)
    z = [0]*n
    l = r = 0
    
    for i in range(1,n):
        if i <= r:
            z[i] = min(r-i+1, z[i-l])
        
        while i+z[i] < n and s[z[i]] == s[i+z[i]]:
            z[i] += 1
        
        if i+z[i]-1 > r:
            l = i
            r = i+z[i]-1
    
    return z
```
