# 1 Structures de données fondamentales

## 1.1 Structures linéaires (piles, files, listes)

Les structures linéaires sont des structures de données dans lesquelles les éléments sont organisés de manière séquentielle, selon un ordre bien défini.

### 1.1.1 Piles

Une pile est une structure de données linéaire qui suit le principe LIFO (Last In, First
Out)

````python
class Piles :
    def __init__ ( self ) :
        self . data = []

    #Ajoute un élément au sommet de la pile
    def push ( self , x ) :
        self . data . append ( x )

    #Retire et retourner l'élément au sommet de la pile
    def pop ( self ) :
        if self . data :
            return self . data . pop ()
        return None

    def is_empty ( self ) :
        return len ( self . data ) == 0
````

### 1.1.2 Files

Une file est une structure de données linéaire qui suit le principe FIFO (First In, First Out)

````python
class File :
    def __init__ ( self ) :
        self . data = []
    
    #Ajoute un élément à la fin de la file
    def enqueue ( self , x ):
        self . data . append ( x )

    #Retire et retourne le premier élément de la file
    def dequeue ( self ) :
        if self . data :
            return self . data . pop (0)
        return None

    def is_empty ( self ):
        return len ( self . data ) == 0
````

### 1.1.3 Listes

Une liste est une structure de données linéaire permettant un accès séquentiel aux éléments.
Les opérations élémentaires de cette structure sont déjà présentes dans la librairie initiale de Python.

````python
class Liste :
    def __init__(self):
        self.data = []

    #Ajoute un élément à la fin de la liste
    def ajouter(self, x):
        self.data.append(x)

    #Supprime et retourne l'élément à la position donnée
    def supprimer(self, index):
        if 0 <= index < len(self.data):
            return self.data.pop(index)
        return None

    def acceder(self, index):
        if 0 <= index < len(self.data):
            return self.data[index]
        return None

    def est_vide(self):
        return len(self.data) == 0
````

## 1.2 Arbres (BST, segment trees, Fenwick trees/BIT)

Les arbres sont des structures de données hiérarchiques, organisées sous forme de nœuds reliés entre eux, permettant une organisation non linéaire des données.

### 1.2.1 BST (ou ABR : arbre binaire de recherche )

Un arbre de recherche binaire est une structure de données arborescente dans laquelle, pour chaque nœud, les éléments du sous-arbre gauche sont inférieurs à la valeur du nœud, et ceux du sous-arbre droit sont supérieurs.

````python
class BST:
    def __init__(self):
        self.data = None

    #Ajoute un élément dans l'arbre en respectant la propriété du BST
    def add(self, x):
        if self.data == None :
            self.data = x
            self.left = BST() #Sous-arbre gauche
            self.right = BST() #Sous-arbre droit
        elif x < self.data :
            #Si x est plus petit que la valeur du noeud courant, on l'insère dans le sous-arbre gauche
            if self.left.data == None :
                self.left.add(x)
            else : 
                self.left.add(x)
        else :
            #Sinon, on l'insère dans le sous-arbre droit
            if self.right.data == None :
                bst = BST()
                bst.add(x)
                self.right = bst
            else : 
                self.right.add(x)
                
    def search (self,x) :
        if self.data == None or self == None:
            return False
        elif self.data == x :
            return True
        elif x < self.data :
            return self.left.search(x)
        else : 
            return self.right.search(x)
            
    def delete (self,x) :
        if self.data == None :
            return
        elif self.data == x :
            #Cas 1 : le noeud est une feuille
            if self.left.data == None and self.right.data == None :
                self.data = None
            #Cas 2 : un seul enfant à droite
            elif self.left.data == None :
                self.data = self.right.data
                self.left = self.right.left
                self.right = self.right.right
            #Cas 3 : un seul enfant à gauche
            elif self.right.data == None :
                self.data = self.left.data
                self.right = self.left.right
                self.left = self.left.left
            #Cas 4 : deux enfants
            else :
                min_right = self.right.min()
                self.data = min_right
                self.right.delete(min_right)
        elif x < self.data :
            return self.left.delete(x)
        else : 
            return self.right.delete(x)
      
    def min (self) :
        if self.left.data == None:
            return self.data
        else :
            return self.left.min()
            
    def max (self) :
        if self.right.data == None:
            return self.data
        else :
            return self.right.max()

    def is_empty(self):
        return self.data == None
````

### 1.2.2 segment trees
Un arbre de segments est une structure de données arborescente permettant de repré-
senter des intervalles et de répondre efficacement à des requêtes sur des segments d’un
tableau, telles que des sommes, minimums ou maximums.

````python
class SegmentTree:
    def __init__(self, arr):
        self.n = len(arr)
        self.tree = [0] * (4 * self.n) 
        self.build(arr, 0, 0, self.n - 1)

    def build(self, arr, node, start, end):
        if start == end:
            #Noeud feuille : correspond à un élément du tableau
            self.tree[node] = arr[start]
        else:
            mid = (start + end) // 2
            left = 2 * node + 1
            right = 2 * node + 2

            #Construction des sous-arbres gauche et droit
            self.build(arr, left, start, mid)
            self.build(arr, right, mid + 1, end)

            #Le noeud courant stocke la somme des deux enfants
            self.tree[node] = self.tree[left] + self.tree[right]

    # Requête : somme sur l'intervalle [l, r]
    def query(self, node, start, end, l, r):
        if r < start or end < l:
            #Aucun recouvrement
            return 0
        if l <= start and end <= r:
            #Recouvrement total
            return self.tree[node]

        #Recouvrement partiel
        mid = (start + end) // 2
        left_sum = self.query(2 * node + 1, start, mid, l, r)
        right_sum = self.query(2 * node + 2, mid + 1, end, l, r)

        return left_sum + right_sum

    def update(self, node, start, end, idx, val):
        if start == end:
            #Mise à jour du noeud feuille
            self.tree[node] = val
        else:
            mid = (start + end) // 2

            if idx <= mid:
                self.update(2 * node + 1, start, mid, idx, val)
            else:
                self.update(2 * node + 2, mid + 1, end, idx, val)

            #Mise à jour du noeud courant
            self.tree[node] = self.tree[2 * node + 1] + self.tree[2 * node + 2]
````

### 1.2.3 Fenwick trees/BIT
Un arbre de Fenwick est une structure de données arborescente optimisée pour le calcul
et la mise à jour rapide de sommes préfixes dans un tableau.

````python
class FenwickTree:
    def __init__(self, n):
        self.n = n
        self.tree = [0] * (n + 1)

    def update(self, i, val):
        while i <= self.n:
            self.tree[i] += val
            i += i & (-i)  #Passe au noeud suivant responsable de cet intervalle

    #Calcule la somme préfixe de 1 à i
    def prefix_sum(self, i):
        s = 0
        while i > 0:
            s += self.tree[i]
            i -= i & (-i)  #Remonte dans l'arbre
        return s

    #Calcule la somme sur l'intervalle [l, r]
    def range_sum(self, l, r):
        return self.prefix_sum(r) - self.prefix_sum(l - 1)
````

## 1.3 Files de priorité et heaps
Une file de priorité est une structure de données permettant d’accéder rapidement à
l’élément de priorité minimale ou maximale, indépendamment de l’ordre d’insertion. Elle
est généralement implémentée à l’aide d’un tas, notamment un tas binaire, qui garantit
des opérations d’insertion et d’extraction en temps logarithmique.

````python
class priorityQueue:
    def __init__(self):
        self.heap = []

    def parent(self, index):
        return (index - 1) // 2
    
    def leftChild(self, index):
        return 2 * index + 1
    
    def rightChild(self, index):
        return 2 * index + 2
    
    def isEmpty(self):
        return len(self.heap) == 0

    def insert(self, item): # item is a tuple (priority, value)
        self.heap.append(item)
        self.heapifyUp(len(self.heap) - 1)

    def heapifyUp(self, index):
        parentIndex = self.parent(index)
        if index > 0 and self.heap[parentIndex][0] > self.heap[index][0]:
            self.heap[parentIndex], self.heap[index] = self.heap[index], self.heap[parentIndex]
            self.heapifyUp(parentIndex)
    
    def heapifyDown(self, index):
        leftChildIndex = self.leftChild(index)
        rightChildIndex = self.rightChild(index)
        smallestIndex = index

        if leftChildIndex < len(self.heap) and self.heap[leftChildIndex][0] < self.heap[smallestIndex][0]:
            smallestIndex = leftChildIndex

        if rightChildIndex < len(self.heap) and self.heap[rightChildIndex][0] < self.heap[smallestIndex][0]:
            smallestIndex = rightChildIndex

        if smallestIndex != index:
            self.heap[index], self.heap[smallestIndex] = self.heap[smallestIndex], self.heap[index]
            self.heapifyDown(smallestIndex)

    def pop(self):
        if self.isEmpty():
            raise IndexError("La file de priorité est vide")
        minItem = self.heap[0]
        lastItem = self.heap.pop()
        if not self.isEmpty():
            self.heap[0] = lastItem
            self.heapifyDown(0)
        return minItem 
````

# 2 Graphes

La théorie des graphes est un pilier fondamental de l'algorithmique, fournissant un langage mathématique pour modéliser des relations et des interactions entre des objets distincts. Un graphe est défini par un couple $G = (V, E)$, où $V$ représente un ensemble de **sommets** (nœuds) et $E$ un ensemble d'**arêtes** (liaisons) reliant ces sommets.

Que ce soit pour modéliser des réseaux sociaux, optimiser des infrastructures de transport ou analyser des dépendances logicielles, les graphes permettent de transformer des problèmes réels complexes en structures exploitables par des algorithmes de parcours, d'optimisation de flux ou de recherche de chemins.



**Modélisation Mathématique Générale**
Pour manipuler un graphe en programmation, nous définissons généralement les éléments suivants :

* **État :** La structure est représentée soit par une **liste d'adjacence** (plus efficace en mémoire pour les graphes creux) soit par une **matrice d'adjacence** (permettant un accès en $O(1)$ pour vérifier l'existence d'une arête).
* **Transition :** Le passage d'un état à un autre correspond au parcours d'une arête $(u, v) \in E$. Ce mouvement peut être bidirectionnel (graphe non-orienté) ou unidirectionnel (graphe orienté).
* **Pondération :** Chaque arête peut être associée à un poids $w(u, v)$, représentant un coût, une distance ou une capacité.

**Propriétés Fondamentales :**
* **Connectivité :** Capacité d'atteindre n'importe quel sommet à partir d'un autre.
* **Cyclicité :** Présence ou absence de chemins fermés (cycles), définissant notamment les DAG (*Directed Acyclic Graphs*).
* **Degré :** Nombre d'arêtes incidentes à un sommet, crucial pour les propriétés d'Euler ou de Hamilton.


## 2.1 Parcours (BFS, DFS)

Les algorithmes de parcours permettent d’explorer systématiquement les sommets d’un graphe. Le **BFS** explore par "couches" successives, tandis que le **DFS** s'aventure le plus loin possible avant de rebrousser chemin.

**Modélisation Mathématique**
* **BFS (Breadth-First Search) :** Utilise une file (FIFO). Idéal pour trouver le plus court chemin dans un graphe non pondéré.
* **DFS (Depth-First Search) :** Utilise une pile (LIFO) ou la récursion. Essentiel pour la détection de cycles ou le tri topologique.


````python
# Parcours en largeur (Breadth-First Search) 
def bfs(graph, start):  # Le graphe est represente sous forme d'une liste d'adjacence
    visited = [False for i in range(len(graph))] 
    queue = []  

    visited[start] = True  #
    queue.append(start)  

    while queue:  # Tant que la file d'attente n'est pas vide
        vertex = queue.pop(0)  
        print(vertex) 

        for neighbor in graph[vertex]: 
            if not visited[neighbor]:  
                visited[neighbor] = True
                queue.append(neighbor) 

# Parcours en profondeur (Depth-First Search)
def dfs(graph, start):
    visited = [False for i in range(len(graph))] 
    stack = []  

    visited[start] = True  
    stack.append(start)  

    while stack:  # Tant que la pile n'est pas vide
        vertex = stack.pop()  
        print(vertex) 

        for neighbor in graph[vertex]: 
            if not visited[neighbor]:  
                visited[neighbor] = True
                stack.append(neighbor)
                
````

## 2.2 Plus courts chemins (Dijkstra, Bellman-Ford, Floyd-Warshall)

Cette section traite de la recherche de la distance minimale entre des sommets dans un graphe pondéré $G = (V, E)$. Le choix de l'algorithme dépend de la présence de poids négatifs et du nombre de sources à traiter.



### 2.2.1 Algorithme de Dijkstra (Source unique, poids positifs)
Dijkstra est un algorithme **glouton** qui maintient un ensemble de sommets dont la distance minimale est connue.

**Modélisation Mathématique**
* **État :** $dist[u]$ représente la distance minimale connue de la source à $u$.
* **Transition (Relâchement) :** On extrait $u$ avec $dist[u]$ minimal, puis pour chaque voisin $v$ :
    $$dist[v] = \min(dist[v], dist[u] + w(u, v))$$
* **Cas de base :** $dist[source] = 0$, et $dist[u] = \infty$ pour $u \neq source$.


````python
import heapq

def dijkstra(n, adj, start):
    dist = [float('inf')] * n
    dist[start] = 0
    pq = [(0, start)]
    while pq:
        d, u = heapq.heappop(pq)
        if d > dist[u]: continue
        for v, weight in adj[u]:
            if dist[u] + weight < dist[v]:
                dist[v] = dist[u] + weight
                heapq.heappush(pq, (dist[v], v))
    return dist
````

**Analyse de complexité Dijkstra :**

* Temporelle $\mathcal{O}((V + E) \log V)$
* Spatiale $\mathcal{O}(V + E)$.

### 2.2.2 Algorithme de Bellman-Ford (Source unique, poids négatifs)
Basé sur la **programmation dynamique**, il supporte les poids négatifs et détecte les cycles négatifs.

**Modélisation Mathématique**
* **État :** $dp[i][v]$ est la distance minimale de la source à $v$ en utilisant au plus $i$ arêtes.
* **Transition :** Pour chaque arête $(u, v)$ de poids $w$ :
    $$dist[v] = \min(dist[v], dist[u] + w)$$
    On répète cette opération $V-1$ fois.

````python
def bellman_ford(n, edges, start):
    dist = [float('inf')] * n
    dist[start] = 0
    for _ in range(n - 1):
        for u, v, w in edges:
            if dist[u] + w < dist[v]:
                dist[v] = dist[u] + w
    # Vérification de cycle négatif
    for u, v, w in edges:
        if dist[u] + w < dist[v]: return None
    return dist
````

**Analyse de complexité Bellman-Ford :**

* Temporelle $\mathcal{O}(V \cdot E)$
* Spatiale $\mathcal{O}(V)$.

Cet algorithme de **programmation dynamique** calcule les plus courts chemins entre **tous les couples** de sommets.

**Modélisation Mathématique**
* **État :** $dp[k][i][j]$ est la distance de $i$ à $j$ en passant par les sommets $\{1..k\}$.
* **Transition :** On vérifie si passer par l'étape $k$ est optimal :
    $$dp[k][i][j] = \min(dp[k-1][i][j], dp[k-1][i][k] + dp[k-1][k][j])$$

````python
def floyd_warshall(n, matrix):
    dist = [row[:] for row in matrix]
    for k in range(n):
        for i in range(n):
            for j in range(n):
                dist[i][j] = min(dist[i][j], dist[i][k] + dist[k][j])
    return dist
````

## 2.3 Arbre couvrant minimum (Kruskal, Prim)

Les algorithmes d’arbre couvrant minimum (MST) permettent de relier tous les sommets d’un graphe pondéré en minimisant le poids total des arêtes sélectionnées. Un arbre couvrant est un sous-graphe connexe sans cycle contenant tous les sommets du graphe original.

### 2.3.1 Algorithme de Kruskal

L'algorithme de Kruskal suit une approche **gloutonne (greedy)** basée sur les arêtes. Il construit l'arbre en sélectionnant les arêtes de plus petit poids qui ne créent pas de cycle.

**Modélisation Mathématique**
* **État :** Un ensemble de composantes connexes géré par une structure de données **Union-Find** (ou ensembles disjoints).
* **Transition :** On trie l'ensemble des arêtes $E$ par poids croissant. Pour chaque arête $(u, v)$ de poids $w$ :
    Si $find(u) \neq find(v)$ (les sommets appartiennent à des arbres différents) :
    $$MST = MST \cup \{(u, v)\}$$
    $$Union(u, v)$$
* **Cas de base :** Chaque sommet $v \in V$ est initialement sa propre composante connexe.



````python
def kruskal(n, edges):
    # edges: liste de (u, v, poids)
    edges.sort(key=lambda x: x[2])
    parent = list(range(n))
    
    def find(i):
        if parent[i] == i: return i
        parent[i] = find(parent[i])
        return parent[i]

    mst_weight = 0
    count = 0
    for u, v, w in edges:
        root_u, root_v = find(u), find(v)
        if root_u != root_v:
            parent[root_u] = root_v
            mst_weight += w
            count += 1
            if count == n - 1: break
            
    return mst_weight
````

**Analyse de la complexité**
* **Temporelle :** $\mathcal{O}(E \log E)$ ou $\mathcal{O}(E \log V)$. Le tri des arêtes est l'étape dominante.
* **Spatiale :** $\mathcal{O}(V + E)$ pour stocker la liste des arêtes et le tableau des parents pour l'Union-Find.

### 2.3.2 Algorithme de Prim

L'algorithme de Prim adopte une approche **gloutonne locale**. Il fait croître un arbre unique à partir d'un sommet source en ajoutant à chaque étape le sommet le plus proche de l'arbre déjà construit.

**Modélisation Mathématique**
* **État :** Un ensemble $S$ de sommets inclus dans le MST et une file de priorité $PQ$ contenant les arêtes $(u, v)$ telles que $u \in S$ et $v \notin S$.
* **Transition :** Extraire de $PQ$ l'arête $(u, v)$ de poids minimal $w$. Si $v \notin S$ :
    $$S = S \cup \{v\}$$
    $$MST\_Weight = MST\_Weight + w$$
    Ajouter toutes les arêtes incidentes à $v$ menant à des sommets hors de $S$ dans $PQ$.
* **Cas de base :** $S = \{source\}$, $PQ$ contient les arêtes incidentes à la source.

````python
import heapq

def prim(n, adj, start=0):
    # adj: liste d'adjacence [u] = [(v, poids), ...]
    visited = [False] * n
    pq = [(0, start)]
    mst_weight = 0
    nodes_included = 0
    
    while pq and nodes_included < n:
        w, u = heapq.heappop(pq)
        if visited[u]: continue
        
        visited[u] = True
        mst_weight += w
        nodes_included += 1
        
        for v, weight in adj[u]:
            if not visited[v]:
                heapq.heappush(pq, (weight, v))
                
    return mst_weight
````


**Analyse de la complexité**

* **Temporelle :** $\mathcal{O}(E \log V)$ avec l'utilisation d'un tas binaire (file de priorité).
* **Spatiale :** $\mathcal{O}(V + E)$ pour stocker la liste d'adjacence et les éléments de la file de priorité.

## 2.4 Flot maximum et coupe minimum

Le problème consiste à déterminer la quantité maximale de flux pouvant circuler d'une source $S$ vers un puits $T$ à travers un réseau de capacités.

**Modélisation Mathématique (Edmonds-Karp)**

* **État :** Un graphe résiduel $G_f$ représentant les capacités restantes.
* **Transition :** Tant qu'il existe un chemin augmentant $P$ de $S$ à $T$ dans $G_f$ (via BFS) :
    $$f_{max} = f_{max} + \min_{e \in P} (cap(e))$$
* **Cas de base :** Le flot initial est nul pour toutes les arêtes.

````python
from collections import deque

def edmonds_karp(capacity, s, t):
    n = len(capacity)
    flow = 0
    parent = [-1] * n
    while True:
        parent = [-1] * n
        queue = deque([s])
        parent[s] = s
        while queue:
            u = queue.popleft()
            for v in range(n):
                if parent[v] == -1 and capacity[u][v] > 0:
                    parent[v] = u; queue.append(v)
        if parent[t] == -1: break
        
        path_flow = float('inf')
        curr = t
        while curr != s:
            path_flow = min(path_flow, capacity[parent[curr]][curr]); curr = parent[curr]
        flow += path_flow
        curr = t
        while curr != s:
            u, v = parent[curr], curr
            capacity[u][v] -= path_flow; capacity[v][u] += path_flow; curr = parent[curr]
    return flow
````

**Complexité**

* **Temporelle :** $\mathcal{O}(V \cdot E^2)$ — Pour l'implémentation d'Edmonds-Karp via BFS.
* **Spatiale :** $\mathcal{O}(V^2)$ ou $\mathcal{O}(V + E)$ — Dépend de la représentation (matrice ou liste d'adjacence).

## 2.5 Composantes fortement connexes (SCC)

Une SCC est un sous-ensemble de sommets d'un graphe orienté où chaque sommet est accessible depuis tous les autres.

**Modélisation Mathématique (Algorithme de Kosaraju)**

* **État :** Un ordre de traitement obtenu par un premier parcours DFS.
* **Transition :** Inverser toutes les arêtes du graphe, puis lancer des DFS sur les sommets selon l'ordre inverse de leur temps de fin.
* **Cas de base :** Tous les sommets sont initialement non visités.

````python
def kosaraju(n, adj, adj_rev):
    visited, stack = [False] * n, []
    def dfs1(u):
        visited[u] = True
        for v in adj[u]:
            if not visited[v]: dfs1(v)
        stack.append(u)
    
    for i in range(n):
        if not visited[i]: dfs1(i)
    
    visited = [False] * n
    sccs = []
    def dfs2(u, curr_scc):
        visited[u] = True; curr_scc.append(u)
        for v in adj_rev[u]:
            if not visited[v]: dfs2(v, curr_scc)

    while stack:
        u = stack.pop()
        if not visited[u]:
            curr_scc = []; dfs2(u, curr_scc); sccs.append(curr_scc)
    return sccs
````


**Complexité**

* **Temporelle :** $\mathcal{O}(V + E)$ — Deux parcours complets (graphe original et inversé).
* **Spatiale :** $\mathcal{O}(V + E)$ — Stockage du graphe transposé et de la pile.

## 2.6 Détection de cycles

Il s'agit de déterminer si un graphe contient au moins un chemin fermé.

**Modélisation Mathématique**

* **État :** Trois couleurs/états par sommet : *Blanc* (non visité), *Gris* (en cours de visite), *Noir* (visite terminée).
* **Transition :** Si lors du DFS d'un sommet $u$, on rencontre un voisin $v$ de couleur *Gris*, un cycle est détecté (back-edge).
* **Cas de base :** Tous les sommets commencent en *Blanc*.




````python
def has_cycle(u, adj, visited, rec_stack):
    visited[u] = rec_stack[u] = True
    for v in adj[u]:
        if not visited[v]:
            if has_cycle(v, adj, visited, rec_stack): return True
        elif rec_stack[v]: return True
    rec_stack[u] = False
    return False
````

**Complexité**

* **Temporelle :** $\mathcal{O}(V + E)$ — Basé sur un parcours DFS.
* **Spatiale :** $\mathcal{O}(V)$ — Pour stocker l'état des sommets et la pile de récursion.

## 2.7 Cycle eulérien

Un cycle eulérien est un parcours fermé passant par chaque arête exactement une fois.

**Modélisation Mathématique (Algorithme de Hierholzer)**

* **État :** Un sommet courant $u$ et un circuit partiel.
* **Transition :** Suivre les arêtes non utilisées depuis $u$ jusqu'à être bloqué, puis remonter pour intégrer les sous-circuits.
* **Cas de base :** Condition d'existence : tous les sommets ont un degré pair (graphe non orienté).


````python
def find_eulerian_circuit(adj, start_node):
    stack, circuit = [start_node], []
    while stack:
        u = stack[-1]
        if adj[u]:
            v = adj[u].pop()
            stack.append(v)
        else:
            circuit.append(stack.pop())
    return circuit[::-1]
````

**Complexité**

* **Temporelle :** $\mathcal{O}(V + E)$ — Chaque arête est parcourue et supprimée une seule fois.
* **Spatiale :** $\mathcal{O}(V + E)$ — Stockage du circuit et de la pile.

## 2.8 Équilibre d’arbre

Maintenir une hauteur logarithmique permet de garantir des performances optimales pour les opérations de base.

**Modélisation Mathématique (Rotation AVL)**

* **État :** Un facteur d'équilibre $bf(u) = height(left) - height(right)$.
* **Transition :** Si $|bf(u)| > 1$, on effectue des rotations (gauche ou droite) pour restaurer l'équilibre.
* **Cas de base :** Un arbre vide ou une feuille est équilibré.

````python
def rotate_right(y):
    x = y.left
    T2 = x.right
    x.right = y; y.left = T2
    # Mise à jour des hauteurs ici
    return x

def rotate_left(x):
    y = x.right
    T2 = y.left
    y.left = x; x.right = T2
    # Mise à jour des hauteurs ici
    return y
````


**Complexité**

* **Temporelle :** $\mathcal{O}(1)$ pour une rotation ; $\mathcal{O}(\log N)$ pour rééquilibrer après insertion/suppression.
* **Spatiale :** $\mathcal{O}(1)$ — Les rotations ne consomment pas de mémoire supplémentaire.

# 3 Programmation dynamique

## 3.1 Introduction à la DP : Top-Down vs Bottom-Up

La Programmation Dynamique (DP) est une méthode d'optimisation mathématique et algorithmique. Elle s'applique aux problèmes qui présentent deux caractéristiques fondamentales :
1. **Sous-structure optimale :** La solution optimale du problème global peut être construite à partir des solutions optimales de ses sous-problèmes.
2. **Chevauchement des sous-problèmes :** L'algorithme naïf résout les mêmes sous-problèmes de multiples fois.

La DP résout ce problème de redondance en stockant les résultats des sous-problèmes (mémoïsation ou tabulation) pour ne les calculer qu'une seule fois. Il existe deux approches principales pour implémenter une DP :

| Caractéristique | Top-Down (Récursif + Mémoïsation) | Bottom-Up (Itératif + Tabulation) |
| :--- | :--- | :--- |
| **Principe** | On part du problème global et on descend récursivement. On stocke le résultat de chaque appel. | On part des cas de base et on remplit un tableau itérativement jusqu'au problème global. |
| **Avantage principal** | Ne calcule que les états strictement nécessaires. Transition très naturelle à écrire. | Pas d'overhead lié à la pile d'appels récursifs. Souvent plus rapide en temps d'exécution constant. |
| **Inconvénient** | Risque de dépassement de la pile d'appels (`RecursionError` en Python) sur de grandes entrées. | Oblige à calculer tous les états, même ceux qui pourraient être inaccessibles. |
| **Optimisation spatiale** | Difficile à mettre en place. | Très facile (ex: ne conserver que la ligne précédente dans une matrice 2D). |

**Exemple d'application : Nombre de chemins dans une grille**
*Problème :* Trouver le nombre de chemins uniques du point $(0, 0)$ au point $(M-1, N-1)$ dans une grille, en ne se déplaçant que vers la droite ou vers le bas. Certains obstacles bloquent le passage.

* **État :** $dp[i][j]$ représente le nombre de chemins pour atteindre la case $(i, j)$.
* **Transition :** Si la case $(i, j)$ n'est pas un obstacle : 
    $dp[i][j] = dp[i-1][j] + dp[i][j-1]$
* **Cas de base :** $dp[0][0] = 1$ (1 façon d'être au départ). La première ligne et la première collone n'ont qu'une seule provenance possible.

````python
# 1 indique une case libre, 0 indique un obstacle

def top_down(grid : list[list[int]]) -> int :
    grid = grid
    M = len(grid)
    N = len(grid[0]) if M > 0 else 0

    if M == 0 or N == 0 or grid[0][0] == 0:
            return 0
    
    memo = [[-1] * N for _ in range(M)]

    def dp(i: int, j: int) -> int:
            # Hors limites ou obstacle
            if i < 0 or j < 0 or grid[i][j] == 0:
                return 0
            # Cas de base
            if i == 0 and j == 0:
                return 1
            # Retourner si déjà calculé
            if memo[i][j] != -1:
                return memo[i][j]
                
            # Transition : venant du haut + venant de la gauche
            memo[i][j] = dp(i - 1, j) + dp(i, j - 1)
            return memo[i][j]
            
    return dp(M - 1, N - 1)
````

Analysons maintenant les performances de cet algorithme. On a une complexité en temps et en espace en $O(M*N)$. 

````python
# 1 indique une case libre, 0 indique un obstacle

def bottom_up(grid : list[list[int]]) -> int :
    grid = grid
    M = len(grid)
    N = len(grid[0]) if M > 0 else 0

    if M == 0 or N == 0 or grid[0][0] == 0:
            return 0
    
    memo = [[0] * N for _ in range(M)]
    # On initialise la première ligne et première collone de notre tableau
    for i in range (M):
        if grid[i][0]==1:
            memo[i][0]=1
        else:
             break
    for j in range (N):
        if grid[0][j]==1:
            memo[0][j]=1
        else:
            break
    
    # On effectue la récurence 
    for i in range (1,M):
          for j in range(1,N):
                if grid[i][j]==1:
                      memo[i][j]=memo[i-1][j]+memo[i][j-1]

    return memo[M-1][N-1]
````

La complexité est la même que pour l'algorithme précédent. En revanche comme il ne fait pas d'appel récursif il n'y a aucun risque de dépasssement de la pile d'appel.

## 3.2 Programmation Dynamique Classique

### 3.2.1 Le problème du Sac à Dos (0/1 Knapsack)

Le problème classique du sac à dos (0/1 car un objet est soit pris, soit laissé) consiste à maximiser la valeur totale des objets placés dans un sac de capacité limitée, sans la dépasser. 

**Modélisation Mathématique**
Soit un ensemble de $N$ objets, chacun ayant un poids $w_i$ et une valeur $v_i$. La capacité maximale du sac est $W$.

* **État :** $dp[i][j]$ représente la valeur maximale que l'on peut obtenir en choisissant parmi les $i$ premiers objets, avec une capacité exacte ou inférieure à $j$.
* **Transition :** Pour l'objet $i$, deux choix s'offrent à nous (le prendre ou l'ignorer).
  Si $j \ge w_i$ (l'objet rentre dans le sac) :
  $$dp[i][j] = \max(dp[i-1][j], dp[i-1][j-w_i] + v_i)$$
  Sinon (l'objet est trop lourd) :
  $$dp[i][j] = dp[i-1][j]$$
* **Cas de base :** $dp[0][j] = 0$ pour tout $j \in [0, W]$. Zéro objet donne une valeur de zéro.



**Optimisation Spatiale :**
La relation de récurrence montre que la ligne $i$ ne dépend que de la ligne $i-1$. Il n'est donc pas nécessaire de stocker une matrice entière de taille $N \times W$. Un simple tableau 1D de taille $W+1$ suffit, à condition de le parcourir **à l'envers** (de $W$ vers $w_i$) pour ne pas écraser les données de l'itération précédente prématurément.

* **Complexité temporelle :** $\mathcal{O}(N \cdot W)$
* **Complexité spatiale (Optimisée) :** $\mathcal{O}(W)$

````python
def Knapsack(capacity: int, weights: list[int], values: list[int]) -> int:
    n = len(weights)
    if n == 0 or capacity <= 0:
        return 0
            
    # Initialisation du tableau 1D pour l'optimisation spatiale
    dp = [0] * (capacity + 1)
     
    # Remplissage itératif (Bottom-Up)
    for i in range(n):
        weight = weights[i]
        value = values[i]
        
        # Parcours inversé crucial pour garantir que chaque objet 
        # n'est pris qu'une seule fois (0/1 Knapsack)
        for w in range(capacity, weight - 1, -1):
            dp[w] = max(dp[w], dp[w - weight] + value)
    return dp[capacity]
````

Ici dans l'équation de récurrence 1D on utilises bien "la case du dessus" en ayant $dp[w]$ comme argument dans la fonction max.

### 3.2.2 Plus Longue Sous-séquence Croissante (LIS)

Le problème de la Plus Longue Sous-séquence Croissante (Longest Increasing Subsequence) consiste à trouver la longueur de la plus longue sous-séquence d'un tableau donné telle que tous les éléments de cette sous-séquence soient triés par ordre strictement croissant.

**1. L'approche classique (Approche $\mathcal{O}(N^2)$)**
* **État :** $dp[i]$ représente la longueur de la LIS qui se termine exactement à l'index $i$.
* **Transition :** Pour chaque élément $i$, on regarde tous les éléments précédents $j$ (avec $j < i$). Si l'élément en $j$ est strictement inférieur à l'élément en $i$, on peut étendre la séquence :
  $$dp[i] = \max(dp[i], dp[j] + 1) \quad \text{pour tout } j < i \text{ tel que } arr[j] < arr[i]$$
* **Cas de base :** $dp[i] = 1$ pour tout $i$, car chaque élément forme au minimum une sous-séquence de taille 1.

**2. L'approche optimale par Dichotomie (Approche $\mathcal{O}(N \log N)$)**
Pour rendre l'algorithme scalable sur de grandes entrées ($N \ge 10^5$), on change la définition de l'état en utilisant le principe du *Patience Sorting*.

* **Nouvel État :** On maintient un tableau `tails` où `tails[k]` stocke la **plus petite valeur finale** d'une sous-séquence croissante de longueur $k+1$.
* **Propriété clé :** Le tableau `tails` sera toujours strictement croissant. Cela permet d'utiliser une recherche dichotomique (Binary Search) pour trouver rapidement où insérer ou remplacer un élément, réduisant la transition de $\mathcal{O}(N)$ à $\mathcal{O}(\log N)$.

````python

def LIS(arr : list[int]) -> int :
    if not arr:
        return 0
    
    # tails[i] stocke la plus petite valeur de fin pour une LIS de longueur i+1
    tails = []

    for x in arr:
        # Implémentation manuelle de la recherche dichotomique (lower bound)
        # Cherche le premier index 'left' tel que tails[left] >= x
        left = 0
        right = len(tails)
            
        while left < right:
            mid = (left + right) // 2
            if tails[mid] < x:
                left = mid + 1
            else:
                right = mid
            
        idx = left

        # Si x est strictement plus grand que tous les éléments, 
        # il prolonge la plus longue séquence actuelle.
        if idx == len(tails):
            tails.append(x)
        # Sinon, il remplace une valeur de fin existante pour optimiser
        # le potentiel des futurs éléments.
        else:
            tails[idx] = x

    return len(tails)
````

Cette implémentation repose sur une redéfinition de l'état. Plutôt que de calculer naïvement la longueur maximale pour chaque indice, nous maintenons un tableau d'états $tails$.

Dans ce tableau, $tails[i]$ stocke la plus petite valeur finale connue pour une sous-séquence croissante de longueur $i + 1$. Maintenir ces valeurs terminales aussi petites que possible est la clé de l'algorithme : cela maximise les chances de pouvoir raccorder et prolonger la séquence avec les éléments futurs.

Par construction, le tableau tails maintient une propriété de stricte croissance. Cette monotonie nous permet de remplacer le classique parcours linéaire (en $O(N)$) par une recherche dichotomique (en $O(\log N)$) pour traiter chaque nouvel élément $x$ :
* **Extension :** Si $x$ est strictement supérieur au dernier élément de tails, il prolonge la plus longue sous-séquence connue. Nous l'ajoutons à la fin du tableau.
* **Optimisation locale :** Sinon, la dichotomie identifie le premier élément dans tails qui est supérieur ou égal à $x$, et le remplace. Cette opération ne rallonge pas la séquence immédiatement, mais elle "abaisse" la valeur finale d'une sous-séquence existante. Cela améliore son potentiel d'extension pour les itérations suivantes.

### 3.2.3 Plus Longue Sous-séquence Commune (LCS)

Le problème de la Plus Longue Sous-séquence Commune consiste à déterminer la longueur de la plus longue séquence de caractères qui apparaît dans le même ordre d'apparition dans deux chaînes distinctes, sans être nécessairement contiguë.

**Modélisation Mathématique :**
Soient deux chaînes $S_1$ de taille $N$ et $S_2$ de taille $M$.

* **État :** $dp[i][j]$ représente la longueur de la LCS pour les préfixes $S_1[0..i-1]$ et $S_2[0..j-1]$.
* **Transition :** Pour chaque paire de caractères, l'approche se divise en deux cas de figure :
  * Si les caractères correspondent ($S_1[i-1] == S_2[j-1]$), on prolonge la séquence commune trouvée jusqu'aux indices précédents :
    $$dp[i][j] = dp[i-1][j-1] + 1$$
  * Si les caractères diffèrent, la séquence ne peut pas être prolongée. On conserve alors le maximum des sous-séquences communes trouvées en ignorant le caractère actuel de l'une ou l'autre des chaînes :
    $$dp[i][j] = \max(dp[i-1][j], dp[i][j-1])$$
* **Cas de base :** $dp[0][j] = 0$ et $dp[i][0] = 0$. Une chaîne vide n'a aucune sous-séquence commune avec une autre chaîne.

**Complexité :**
* **Temporelle :** $\mathcal{O}(N \cdot M)$
* **Spatiale (Optimisée) :** $\mathcal{O}(\min(N, M))$ en ne conservant en mémoire que la ligne précédente lors de l'itération. *Attention : la matrice complète $\mathcal{O}(N \cdot M)$ reste nécessaire si l'objectif est de reconstruire (backtracker) la chaîne exacte et pas seulement d'obtenir sa longueur.*

````python
def longest_common_subsequence(s1: str, s2: str) -> int:
    """
    Calcule la longueur de la Plus Longue Sous-séquence Commune (LCS).
    Implémente une optimisation spatiale stricte.
    """
    # Optimisation spatiale : on s'assure que s2 est toujours la chaîne 
    # la plus courte pour allouer le plus petit tableau possible en mémoire.
    if len(s1) < len(s2):
        s1, s2 = s2, s1
        
    n, m = len(s1), len(s2)
    
    # dp représente la ligne i-1 de la matrice de programmation dynamique
    dp = [0] * (m + 1)
    curr_dp = [0] * (m + 1)

    for i in range(1, n + 1):
        # curr_dp représente la ligne i en cours de calcul
        curr_dp[0]=0
        for j in range(1, m + 1):
            if s1[i - 1] == s2[j - 1]:
                curr_dp[j] = dp[j - 1] + 1
            else:
                if curr_dp[j - 1] > dp[j]: 
                    curr_dp[j] = curr_dp[j - 1]
                else :
                    curr_dp[j]=dp[j]
        # Le calcul de la ligne courante est terminé, elle devient la ligne précédente
        dp = curr_dp
        
    return dp[m]
````

**Logique de l'optimisation spatiale :**
L'équation de transition mathématique montre que pour calculer la valeur de la case `dp[i][j]`, l'algorithme n'a besoin que de consulter la case située juste au-dessus (`dp[i-1][j]`), la case à sa gauche (`dp[i][j-1]`), et la case en diagonale (`dp[i-1][j-1]`). 

Il est donc inutile de stocker l'historique de toutes les lignes précédentes. En maintenant uniquement deux tableaux unidimensionnels — un pour la ligne précédente (`dp`) et un pour la ligne en cours d'évaluation (`curr_dp`) — l'empreinte mémoire (footprint) est drastiquement réduite de $\mathcal{O}(N \cdot M)$ à $\mathcal{O}(\min(N, M))$. Le "swap" final (`dp = curr_dp`) s'effectue simplement par réaffectation de référence sans coût de copie supplémentaire, garantissant une exécution performante.

## 3.3 DP sur les intervalles (Interval DP)

La DP sur les intervalles est utilisée lorsque la solution d'un problème sur un segment continu `[l, r]` dépend de la combinaison des solutions optimales de ses propres sous-segments. La particularité architecturale de cette technique réside dans le flux de contrôle : on n'itère pas de gauche à droite, mais par **longueur croissante de l'intervalle**. On résout d'abord tous les segments de taille 1, puis 2, etc., jusqu'à la taille $N$.

### Exemple classique : Multiplication d'une chaîne de matrices

**Exemple illustratif : L'impact critique du parenthésage**\
Considérons la multiplication de trois matrices $A$, $B$ et $C$ avec les dimensions suivantes :
* **$A$** : $10 \times 100$
* **$B$** : $100 \times 5$
* **$C$** : $5 \times 50$

Pour rappel, multiplier une matrice de dimension $p \times q$ par une matrice $q \times r$ requiert exactement $p \cdot q \cdot r$ opérations scalaires (ils existent des algorithmes le faisant de manières plus efficaces : [Wikipédia](https://fr.wikipedia.org/wiki/Complexit%C3%A9_de_la_multiplication_de_matrices)). Le produit matriciel étant associatif, nous pouvons calculer $A \cdot B \cdot C$ de deux manières. Voyons l'impact sur les performances :

**Option 1 : Parenthésage $(A \cdot B) \cdot C$**
1. Calcul de $(A \cdot B)$ : $10 \cdot 100 \cdot 5 = 5000$ opérations. La matrice résultante est de dimension $10 \times 5$.
2. Multiplication par $C$ : $10 \cdot 5 \cdot 50 = 2500$ opérations.
3. **Coût total :** $5000 + 2500 = 7500$ opérations.

**Option 2 : Parenthésage $A \cdot (B \cdot C)$**
1. Calcul de $(B \cdot C)$ : $100 \cdot 5 \cdot 50 = 25000$ opérations. La matrice résultante est de dimension $100 \times 50$.
2. Multiplication par $A$ : $10 \cdot 100 \cdot 50 = 50000$ opérations.
3. **Coût total :** $25000 + 50000 = 75000$ opérations.

**Conclusion :**
Le second choix de parenthésage nécessite **10 fois plus d'opérations** (75 000 contre 7 500) pour aboutir strictement au même résultat mathématique. Cet écart de performance croît de manière exponentielle avec le nombre de matrices (nombre de Catalan), ce qui rend une recherche exhaustive en force brute totalement inenvisageable. C'est précisément pour cela que l'optimisation par la DP sur les intervalles est requise.

Le problème consiste donc à trouver l'ordre optimal (le parenthésage) pour multiplier une suite de matrices afin de minimiser le nombre total d'opérations scalaires.

**Modélisation Mathématique**
Soit un tableau de dimensions $P$ de taille $N+1$, où la matrice $A_i$ a pour dimension $P[i-1] \times P[i]$.

* **État :** $dp[i][j]$ représente le coût minimum pour multiplier la sous-chaîne de matrices allant de l'indice $i$ à $j$ (inclus).
* **Transition :** Pour trouver le coût optimal sur le segment $[i, j]$, on teste tous les points de coupure possibles $k$ (avec $i \le k < j$).
  $$dp[i][j] = \min_{i \le k < j} \left( dp[i][k] + dp[k+1][j] + P[i-1] \cdot P[k] \cdot P[j] \right)$$
* **Cas de base :** $dp[i][i] = 0$ pour tout $i$. Le coût pour multiplier une matrice par elle-même n'existe pas.

**Complexité**
* **Temporelle :** $\mathcal{O}(N^3)$ (deux boucles pour définir les bornes de l'intervalle, une boucle interne pour chercher le point de coupure optimal).
* **Spatiale :** $\mathcal{O}(N^2)$ (la matrice des états).

````python
def matrix_chain_multiplication(p: list[int]) -> int:
    """
    Calcule le coût minimum pour multiplier une chaîne de matrices.
    Le tableau p contient les dimensions : la matrice i a pour taille p[i] x p[i+1].
    """
    # Le nombre de matrices est la longueur du tableau des dimensions moins 1
    n = len(p) - 1
    if n <= 0:
        return 0
        
    # Initialisation de la matrice DP. 
    # dp[i][j] stockera le coût min pour multiplier les matrices de i à j.
    # On utilise l'infini comme valeur de base pour la minimisation.
    dp = [[float('inf')] * n for _ in range(n)]
    
    # Cas de base : le coût pour multiplier une matrice seule est de 0
    for i in range(n):
        dp[i][i] = 0
        
    # Workflow spécifique à l'Interval DP : itération sur la LONGUEUR du sous-segment
    # La longueur varie de 2 (plus petit calcul possible) à n (le problème global)
    for length in range(2, n + 1):
        for i in range(n - length + 1):
            j = i + length - 1
            
            # Recherche du point de coupure k optimal divisant [i, j] en [i, k] et [k+1, j]
            for k in range(i, j):
                # Le coût combine la résolution des deux sous-problèmes plus le coût 
                # de la multiplication des deux matrices résultantes.
                cost = dp[i][k] + dp[k + 1][j] + p[i] * p[k + 1] * p[j + 1]
                
                if cost < dp[i][j]:
                    dp[i][j] = cost
                    
    return dp[0][n - 1]
````

## 3.4 DP sur Arbres et graphes

La programmation dynamique sur les arbres exploite le fait qu'un arbre (un graphe connexe sans cycle) possède une hiérarchie naturelle. En choisissant un sommet arbitraire comme racine, on peut définir un ordre topologique implicite. Les sous-problèmes se définissent alors sur les sous-arbres, et les solutions remontent des feuilles vers la racine via un parcours en profondeur (DFS).

### Exemple classique : Ensemble Indépendant Maximum (Maximum Independent Set)

Le problème consiste à trouver la taille maximale d'un sous-ensemble de sommets tel qu'aucune paire de sommets de ce sous-ensemble ne soit reliée par une arête.

**Modélisation Mathématique**
Soit un arbre de $N$ sommets. Pour chaque sommet $u$, son état dépend de la décision de l'inclure ou non dans notre ensemble.

* **État :** Nous avons besoin de deux états par sommet pour capturer la contrainte d'adjacence.
  * $dp[u][0]$ : La taille maximale de l'ensemble indépendant dans le sous-arbre enraciné en $u$, **si l'on n'inclut pas** le sommet $u$.
  * $dp[u][1]$ : La taille maximale de l'ensemble indépendant dans le sous-arbre enraciné en $u$, **si l'on inclut** le sommet $u$.
* **Transition :** Soit $v$ un enfant direct du sommet $u$.
  * Si l'on **n'inclut pas** $u$, on est libre de prendre ou non l'enfant $v$. L'optimisation locale consiste à prendre le meilleur des deux cas pour chaque enfant :
    $$dp[u][0] = \sum_{v \in enfants(u)} \max(dp[v][0], dp[v][1])$$
  * Si l'on **inclut** $u$, la contrainte stricte de l'ensemble indépendant nous interdit de prendre l'enfant $v$. On est obligé de s'appuyer sur l'état où $v$ est exclu :
    $$dp[u][1] = 1 + \sum_{v \in enfants(u)} dp[v][0]$$
* **Cas de base :** Pour une feuille $f$ (un sommet sans enfant), les sommes sont vides. Donc $dp[f][0] = 0$ et $dp[f][1] = 1$.

**Complexité**
* **Temporelle :** $\mathcal{O}(N)$ car chaque sommet et chaque arête est visité exactement une fois lors du parcours DFS.
* **Spatiale :** $\mathcal{O}(N)$ pour la matrice DP et la pile d'appels récursifs.

````python
def tree_max_independent_set(n: int, edges: list[list[int]]) -> int:
    """
    Calcule la taille de l'ensemble indépendant maximum dans un arbre.
    Implémente un Tree DP via un parcours en profondeur (DFS).
    """
    if n <= 0:
        return 0
        
    # Construction de la liste d'adjacence pour modéliser le graphe
    adj = [[] for _ in range(n)]
    for u, v in edges:
        adj[u].append(v)
        adj[v].append(u)
        
    # dp[u][0] : max en excluant u
    # dp[u][1] : max en incluant u
    dp = [[0, 0] for _ in range(n)]
    
    def dfs(node: int, parent: int):
        # Initialisation du cas de base pour le nœud courant
        dp[node][0] = 0
        dp[node][1] = 1
        
        for neighbor in adj[node]:
            if neighbor == parent:
                continue
                
            # Descente récursive vers les feuilles
            dfs(neighbor, node)
            
            # Remontée : Application des transitions
            # Si on exclut 'node', on prend le meilleur choix pour 'neighbor'
            dp[node][0] += dp[neighbor][0] if dp[neighbor][0] > dp[neighbor][1] else dp[neighbor][1]
            
            # Si on inclut 'node', on est obligé d'exclure 'neighbor'
            dp[node][1] += dp[neighbor][0]

    # On lance l'algorithme depuis la racine arbitraire (sommet 0), sans parent (-1)
    dfs(0, -1)
    
    # La solution optimale globale est le maximum à la racine
    return dp[0][0] if dp[0][0] > dp[0][1] else dp[0][1]
````

**Avertissement technique pour Python :**
Bien que la complexité algorithmique soit strictement en $\mathcal{O}(N)$, l'implémentation récursive standard pose un problème d'architecture en Python. La limite de profondeur de la pile d'appels (Call Stack) est par défaut fixée à 1000. Pour un arbre très déséquilibré (ex: une ligne droite de $10^5$ sommets), cela provoquera une erreur `RecursionError`.

### Plus Long Chemin dans un DAG


Contrairement aux arbres, un graphe orienté peut posséder des chemins qui se rejoignent. Cependant, tant qu'il n'y a pas de cycles (DAG - Directed Acyclic Graph), on conserve un ordre topologique valide qui garantit qu'un sous-problème ne dépendra jamais de lui-même. 

Le problème de la recherche du plus long chemin est NP-Complet dans un graphe général, mais se résout en temps linéaire sur un DAG grâce à la mémoïsation.

**Modélisation Mathématique**
Soit un graphe orienté acyclique de $N$ sommets.

* **État :** $dp[u]$ représente la longueur (en nombre d'arêtes) du plus long chemin orienté qui **commence** au sommet $u$.
* **Transition :** Pour trouver le plus long chemin depuis $u$, on regarde tous ses voisins directs $v$, on calcule leur propre plus long chemin, et on ajoute 1 (l'arête reliant $u$ à $v$). On prend ensuite le maximum de ces choix :
  $$dp[u] = \max_{v \in voisins(u)} (1 + dp[v])$$
* **Cas de base :** Si le sommet $u$ est un puits (il n'a aucune arête sortante), la boucle de transition ne s'exécute pas. Son chemin sortant maximum est naturellement $0$.

**Complexité**
* **Temporelle :** $\mathcal{O}(V + E)$ où $V$ est le nombre de sommets et $E$ le nombre d'arêtes. Chaque sommet et chaque arête n'est évalué qu'une seule fois.
* **Spatiale :** $\mathcal{O}(V + E)$ pour stocker le graphe (liste d'adjacence), le tableau $dp$, et la pile d'appels récursifs.

````python
def longest_path_in_dag(n: int, edges: list[list[int]]) -> int:
    """
    Calcule la longueur du plus long chemin dans un Graphe Orienté Acyclique (DAG).
    Utilise un parcours en profondeur (DFS) avec mémoïsation.
    """
    if n <= 0:
        return 0

    # Construction de la liste d'adjacence pour un graphe orienté
    adj = [[] for _ in range(n)]
    for u, v in edges:
        adj[u].append(v)

    # Initialisation du tableau de mémoïsation avec -1 (état non calculé)
    dp = [-1] * n

    def dfs(node: int) -> int:
        # Retourne la valeur immédiatement si l'état a déjà été calculé
        if dp[node] != -1:
            return dp[node]

        max_path = 0
        for neighbor in adj[node]:
            # La distance est 1 (l'arête u->v) + le plus long chemin depuis v
            path_length = 1 + dfs(neighbor)
            if path_length > max_path:
                max_path = path_length

        # Mémoïsation du résultat optimal pour ce nœud
        dp[node] = max_path
        return dp[node]

    # Le graphe n'étant pas nécessairement connexe, on doit potentiellement 
    # lancer le DFS depuis chaque sommet non visité pour trouver le max global.
    global_max = 0
    for i in range(n):
        if dp[i] == -1:
            current_path_length = dfs(i)
            if current_path_length > global_max:
                global_max = current_path_length

    return global_max
````

## 3.5 DP Combinatoire avec Masques de Bits (Bitmask DP)

La technique du masque de bits repose sur une idée architecturale simple mais puissante : utiliser la représentation binaire d'un entier pour encoder un ensemble d'états booléens. Si le $k$-ième bit d'un entier est à $1$, cela signifie que le $k$-ième élément est inclus dans notre sous-ensemble.

Cette approche permet d'utiliser les opérations bit à bit (`&`, `|`, `^`, `<<`), qui s'exécutent en un seul cycle d'horloge au niveau du processeur, rendant les transitions d'états extrêmement rapides.

### Exemple classique : Le Problème du Voyageur de Commerce (TSP)
Le problème consiste à trouver le cycle de coût minimum passant par tous les sommets d'un graphe pondéré exactement une fois, et revenant à son point de départ.

**Modélisation Mathématique**
Soit un graphe complet de $N$ sommets et une matrice de distances $D$. Un sous-ensemble de sommets visités est représenté par un entier $mask$.

* **État :** $dp[mask][i]$ représente le coût minimum pour visiter le sous-ensemble de villes défini par $mask$, en terminant le trajet actuel à la ville $i$.
* **Transition :** Pour calculer $dp[mask][i]$, on regarde toutes les villes $j$ qui appartiennent au sous-ensemble précédent (c'est-à-dire $mask$ sans la ville $i$). On choisit la ville $j$ minimisant le coût de transition vers $i$ :
  $$dp[mask][i] = \min_{j \in (mask \setminus \{i\})} \left( dp[mask \setminus \{i\}][j] + D[j][i] \right)$$
* **Cas de base :** On fixe arbitrairement la ville $0$ comme point de départ. Le masque initial ne contient que cette ville (soit $1 \ll 0$, donc $1$).
  $$dp[1][0] = 0$$
  Tous les autres états sont initialisés à $+\infty$.

**Complexité**
* **Temporelle :** $\mathcal{O}(2^N \cdot N^2)$. Il y a $2^N$ masques, $N$ villes d'arrivée $i$, et $N$ villes de provenance $j$. C'est une amélioration drastique par rapport à la force brute en $\mathcal{O}(N!)$.
* **Spatiale :** $\mathcal{O}(2^N \cdot N)$ pour stocker la matrice des états.

````python
def tsp_bitmask(n: int, dist: list[list[int]]) -> int:
    """
    Résout le Problème du Voyageur de Commerce (TSP) via Bitmask DP.
    Retourne le coût minimum pour visiter toutes les villes et revenir au départ.
    """
    if n <= 1:
        return 0

    # 1 << n est équivalent à 2^n. 
    # dp[mask][i] stocke le coût min pour le sous-ensemble 'mask' finissant en 'i'.
    total_masks = 1 << n
    dp = [[float('inf')] * n for _ in range(total_masks)]
    
    # Point de départ : ville 0. Le masque binaire est 000...001 (valeur 1).
    dp[1][0] = 0
    
    # On itère sur tous les états possibles des sous-ensembles
    for mask in range(1, total_masks):
        for i in range(n):
            # Vérification : la ville 'i' doit appartenir au sous-ensemble actuel 'mask'
            if (mask & (1 << i)) != 0:
                
                # prev_mask est l'état avant d'avoir visité la ville 'i' (opération XOR)
                prev_mask = mask ^ (1 << i)
                
                # Si le masque précédent est vide, on ne peut pas faire de transition
                if prev_mask == 0:
                    continue
                    
                # Recherche de la meilleure ville 'j' depuis laquelle venir
                for j in range(n):
                    # La ville 'j' devait être dans l'état précédent
                    if (prev_mask & (1 << j)) != 0:
                        cost = dp[prev_mask][j] + dist[j][i]
                        if cost < dp[mask][i]:
                            dp[mask][i] = cost

    # Reconstruire le cycle : relier la dernière ville visitée à la ville de départ (0)
    # Le masque final (toutes les villes visitées) est composé de N bits à 1.
    full_mask = total_masks - 1
    min_cycle_cost = float('inf')
    
    for i in range(1, n):
        # On ajoute le coût de retour D[i][0] pour fermer le cycle
        cycle_cost = dp[full_mask][i] + dist[i][0]
        if cycle_cost < min_cycle_cost:
            min_cycle_cost = cycle_cost
            
    return min_cycle_cost
````

## 3.6 Optimisations avancées de Programmation Dynamique

Lorsque l'on fait face à des limites de temps très strictes sur des tableaux de grande taille ($N \ge 10^5$), les transitions classiques en $\mathcal{O}(N^2)$ deviennent un goulot d'étranglement. Il faut alors exploiter les propriétés mathématiques de la fonction de coût pour réduire l'espace de recherche.

### 3.6.1 L'optimisation Divide and Conquer (D&C DP)


Cette optimisation s'applique aux problèmes de partitionnement (ex: diviser un tableau en $K$ sous-segments continus pour minimiser un coût total). 

**Modélisation Mathématique**
La relation de récurrence typique est de la forme :
$$dp[i][j] = \min_{k < j} (dp[i-1][k] + C(k, j))$$
Où $dp[i][j]$ est le coût minimum pour partitionner les $j$ premiers éléments en $i$ groupes, et $C(k, j)$ est le coût du segment allant de l'indice $k+1$ à $j$.

**La condition stricte (Monotonie du point de coupure) :**
Pour appliquer l'optimisation D&C, la fonction de coût $C$ doit vérifier l'inégalité quadrangulaire. En termes simples, si $opt(i, j)$ est le point de coupure $k$ optimal pour $dp[i][j]$, alors ces points optimaux doivent être monotones :
$$opt(i, j) \le opt(i, j+1)$$

**L'architecture de l'algorithme :**
Puisque le point optimal pour $j$ est contraint, si l'on calcule le point optimal $opt(mid)$ pour le milieu d'un intervalle $[j_{left}, j_{right}]$, on sait de manière déterministe que :
1. Pour tout $j < mid$, le point optimal se trouve dans $[k_{left}, opt(mid)]$.
2. Pour tout $j > mid$, le point optimal se trouve dans $[opt(mid), k_{right}]$.
On peut donc diviser le problème en deux et résoudre récursivement, ce qui réduit la recherche du point de coupure d'une complexité linéaire à une complexité logarithmique.

````python
def solve_divide_and_conquer_dp(n: int, k_groups: int, cost_function) -> int:
    """
    Template pour l'optimisation Divide and Conquer DP.
    Calcule le coût minimum pour diviser N éléments en K groupes.
    """
    if n == 0 or k_groups == 0:
        return 0

    # dp[i][j] stockera les résultats, mais pour l'optimisation spatiale,
    # on ne conserve que la ligne précédente (prev_dp) et la courante (curr_dp).
    prev_dp = [float('inf')] * (n + 1)
    curr_dp = [float('inf')] * (n + 1)
    
    # Initialisation : 1 seul groupe (i=1)
    for j in range(1, n + 1):
        prev_dp[j] = cost_function(0, j)
        
    def compute(j_left: int, j_right: int, k_left: int, k_right: int):
        """
        Calcule curr_dp[mid] et divise l'espace de recherche récursivement.
        - [j_left, j_right] : intervalle des indices du tableau à calculer.
        - [k_left, k_right] : intervalle de recherche du point de coupure optimal.
        """
        if j_left > j_right:
            return
            
        mid = (j_left + j_right) // 2
        best_cost = float('inf')
        best_k = -1
        
        # Le point de coupure ne peut pas dépasser 'mid' ou 'k_right'
        # On s'assure de respecter l'ordre topologique : k doit être < mid
        search_limit = mid if mid < k_right else k_right
        
        for k in range(k_left, search_limit + 1):
            cost = prev_dp[k] + cost_function(k, mid)
            if cost < best_cost:
                best_cost = cost
                best_k = k
                
        curr_dp[mid] = best_cost
        
        # Division et appels récursifs en exploitant la monotonie
        compute(j_left, mid - 1, k_left, best_k)
        compute(mid + 1, j_right, best_k, k_right)

    # Remplissage itératif pour chaque nombre de groupes (de 2 à K)
    for i in range(2, k_groups + 1):
        # On lance le D&C pour calculer la ligne 'curr_dp' à partir de 'prev_dp'
        compute(1, n, 0, n)
        
        # Swap des références pour l'itération suivante
        prev_dp = curr_dp
        curr_dp = [float('inf')] * (n + 1)

    return prev_dp[n]
````

**Pourquoi cette architecture est-elle performante ?**

L'approche classique $\mathcal{O}(K \cdot N^2)$ nécessiterait trois boucles imbriquées : une pour le nombre de groupes $K$, une pour la fin du segment $j$, et une pour chercher le point de coupure $k$. 

La fonction récursive `compute` remplace la recherche linéaire du point $k$. À chaque niveau de l'arbre de récursivité, le milieu `mid` divise l'intervalle d'indices $[j_{left}, j_{right}]$ par deux. La somme des tailles des intervalles de recherche $[k_{left}, k_{right}]$ à chaque niveau de profondeur de l'arbre est bornée par $\mathcal{O}(N)$. Puisque l'arbre a une profondeur de $\log N$, le temps de calcul pour une ligne complète passe de $\mathcal{O}(N^2)$ à $\mathcal{O}(N \log N)$. 

Multiplié par $K$ itérations, la complexité globale chute à $\mathcal{O}(K \cdot N \log N)$, assurant la scalabilité de l'algorithme sur des jeux de données massifs.

### 3.6.2 Exercice d'application : Partitionnement sous contrainte de temps

Pour bien comprendre l'impact critique de cette optimisation, étudions un problème classique de compétition (similaire à "Guardians of the Lunatics").

**Le Problème :**
On te donne un tableau de `N` éléments représentant des valeurs de "folie" de prisonniers alignés dans des cellules adjacentes. Tu dois diviser ces prisonniers en exactement `K` secteurs contigus. 
Le coût d'un secteur allant de l'indice $i$ à $j$ est défini par le nombre de prisonniers dans ce secteur multiplié par la somme de leur folie :
$$C(i, j) = (j - i + 1) \times \sum_{x=i}^j A[x]$$
Ton objectif est de minimiser la somme totale des coûts des `K` secteurs.

**Les Contraintes (Le piège de la scalabilité) :**
* `N` (Nombre d'éléments) $\le 8000$
* `K` (Nombre de secteurs) $\le 800$
* Temps d'exécution maximum : 2.0 secondes.

**Analyse de la complexité :**
* **Approche DP classique $\mathcal{O}(K \cdot N^2)$ :** L'algorithme nécessiterait environ $800 \times \frac{8000^2}{2} \approx 2.5 \times 10^{10}$ opérations. En Python (et même en C++), cela garantit un **Time Limit Exceeded (TLE)** massif, car un processeur standard exécute environ $10^8$ opérations par seconde.
* **Approche optimisée D&C $\mathcal{O}(K \cdot N \log N)$ :** La fonction de coût satisfait l'inégalité quadrangulaire. L'optimisation réduit le nombre d'opérations à $800 \times 8000 \times \log_2(8000) \approx 8.3 \times 10^7$ opérations. L'algorithme passe très largement sous la barre des 2 secondes (Accepted).

**Architecture de la solution :**
1. **Sommes préfixes :** Pour calculer le coût $C(i, j)$ en $\mathcal{O}(1)$ et ne pas rajouter une boucle interne, on pré-calcule un tableau de sommes préfixes.
2. **Mémoire :** On utilise deux tableaux 1D (`prev_dp` et `curr_dp`) pour respecter une empreinte mémoire en $\mathcal{O}(N)$.
3. **Récursivité D&C :** On utilise le template Divide and Conquer pour trouver les points de coupure optimaux.

````python
def solve_lunatics_partition(n: int, k_groups: int, arr: list[int]) -> int:
    """
    Résout le problème de partitionnement avec l'optimisation Divide and Conquer.
    Minimise la somme des coûts des secteurs.
    """
    if n == 0 or k_groups == 0:
        return 0
        
    # k_groups ne peut pas être supérieur au nombre d'éléments
    k_groups = k_groups if k_groups < n else n

    # Pré-calcul des sommes préfixes pour obtenir la somme d'un segment en O(1)
    # prefix[i] contiendra la somme des éléments de arr[0] à arr[i-1]
    prefix = [0] * (n + 1)
    for i in range(1, n + 1):
        prefix[i] = prefix[i - 1] + arr[i - 1]

    def cost(i: int, j: int) -> int:
        """
        Calcule le coût du secteur allant de l'indice i à j (indices basés sur 1).
        Coût = (longueur) * (somme des éléments).
        """
        if i > j:
            return 0
        length = j - i + 1
        segment_sum = prefix[j] - prefix[i - 1]
        return length * segment_sum

    # Optimisation spatiale : on ne garde que la ligne précédente et la ligne courante
    prev_dp = [float('inf')] * (n + 1)
    curr_dp = [float('inf')] * (n + 1)

    # Initialisation du cas de base : coût si tout est dans 1 seul groupe
    for j in range(1, n + 1):
        prev_dp[j] = cost(1, j)

    def compute(j_left: int, j_right: int, k_left: int, k_right: int):
        if j_left > j_right:
            return

        mid = (j_left + j_right) // 2
        best_cost = float('inf')
        best_k = -1

        # Le point de coupure k définit la fin du groupe précédent.
        # Il doit être avant 'mid' (k < mid) pour laisser au moins 1 élément 
        # dans le groupe actuel allant de k+1 à mid.
        search_limit = mid - 1 if (mid - 1) < k_right else k_right
        
        # On ne peut pas couper avant k_left
        start_search = k_left if k_left > 1 else 1

        for k in range(start_search, search_limit + 1):
            # Coût = coût minimal avec un groupe en moins jusqu'à k 
            #        + coût du nouveau groupe de k+1 à mid
            current_cost = prev_dp[k] + cost(k + 1, mid)
            
            if current_cost < best_cost:
                best_cost = current_cost
                best_k = k

        curr_dp[mid] = best_cost

        # Diviser pour régner : on exploite la monotonie du point de coupure
        if best_k != -1:
            compute(j_left, mid - 1, k_left, best_k)
            compute(mid + 1, j_right, best_k, k_right)

    # Remplissage itératif pour chaque nombre de groupes de 2 à K
    for g in range(2, k_groups + 1):
        compute(1, n, 1, n)
        
        # Swap des tableaux pour passer à l'itération suivante
        prev_dp = curr_dp
        curr_dp = [float('inf')] * (n + 1)

    return int(prev_dp[n])
````

## 4.1 Algorithmes de recherche de motifs

Les algorithmes de recherche de motifs permettent de localiser les occurrences d’une sous-chaîne (appelée *motif*) dans un texte.

Une approche naïve consiste à comparer le motif avec toutes les positions possibles dans le texte.

Cependant, cette méthode a une complexité \(O(nm)\), où :

- \(n\) est la longueur du texte  
- \(m\) est la longueur du motif  

Lorsque les chaînes sont très longues, cette complexité devient inefficace.  
Des algorithmes plus avancés, comme KMP et Z-algorithm permettent de résoudre ce problème en temps linéaire ou quasi linéaire.

````python
def naive_pattern_search(text, pattern):

    # Longueur du texte
    n = len(text)
    # Longueur du motif à rechercher
    m = len(pattern)

    # Liste pour stocker les positions où le motif apparaît
    positions = []

    # On parcourt toutes les positions possibles du texte
    # où le motif peut commencer
    for i in range(n - m + 1):

        # On compare la sous-chaîne du texte avec le motif
        if text[i:i+m] == pattern:
            positions.append(i)

    return positions
# Complexité : O(nm)
````

## 4.2 Hachage de chaînes
Le hachage de chaînes consiste à associer une valeur numérique à une chaîne de caractères afin de comparer efficacement des sous-chaînes ou de détecter des égalités.

Cette technique est particulièrement utile pour :

- comparer rapidement deux sous-chaînes  
- détecter des sous-chaînes identiques  
- résoudre des problèmes comme la recherche de motifs ou la comparaison de sous-chaînes  

Le principe repose généralement sur un polynomial rolling hash, qui transforme une chaîne en une valeur numérique calculée à partir des caractères et de puissances d’une base.

Le hachage de chaînes nécessite un prétraitement en \(O(n)\) et permet ensuite de comparer deux sous-chaînes en \(O(1)\).

````python
class StringHash:
    def __init__(self, s):
        self.s = s #chaîne de caractères dont on veut calculer le hash
        self.n = len(s) #longueur de la chaîne
        self.p = 31 #base utilisée pour le hash polynomial
        self.mod = 10**9 + 7 #modulo pour éviter que les nombres deviennent trop grands
        
        #tableau contenant les puissances de p
        self.p_pow = [1] * (self.n + 1)
        #tableau des hash préfixes
        self.h = [0] * (self.n + 1)
        
        #calcul des puissances de p
        for i in range(1, self.n + 1):
            self.p_pow[i] = (self.p_pow[i-1] * self.p) % self.mod
        #calcul des hash préfixes
        for i in range(self.n):
            self.h[i+1] = (self.h[i] + (ord(s[i]) - ord('a') + 1) * self.p_pow[i]) % self.mod
    
    def get_hash(self, l, r):
        #fonction qui calcule le hash de la sous-chaîne
        #on utilise la différence des hash préfixes
        hash_value = (self.h[r+1] - self.h[l] + self.mod) % self.mod
        #normalisation pour pouvoir comparer des sous-chaînes
        return (hash_value * pow(self.p, self.n-l-1, self.mod)) % self.mod
````

## 4.3 KMP
L’algorithme *KMP (Knuth–Morris–Pratt)* permet de rechercher un motif dans un texte en temps linéaire.

L’idée principale est d’éviter de recommencer les comparaisons depuis le début lorsqu’un échec se produit.  
Pour cela, on exploite les préfixes qui sont aussi suffixes grâce à un tableau appelé fonction préfixe dont la fonctionnement est suivant.

Pour chaque position i, π[i] représente la longueur du plus long préfixe propre de la chaîne qui est aussi suffixe du préfixe se terminant en i.

L’algorithme KMP est utile lorsque :

- on doit trouver toutes les occurrences d’un motif dans un texte  
- les chaînes sont très longues  
- on veut éviter la complexité \(O(nm)\) de la méthode naïve  

Exemples de problèmes :

- recherche de motifs  
- détection de répétitions dans une chaîne  
- construction d’automates de chaînes 
La complexité de l’algorithme est \[O(n + m)\] où \(n\) est la longueur du texte et \(m\) est la longueur du motif  

````python
def prefix_function(s):
    n = len(s) #longueur de la chaîne
    pi = [0]*n #tableau de la fonction préfixe
    
    #on parcourt la chaîne à partir du deuxième caractère
    for i in range(1,n):
        j = pi[i-1] #longueur du plus long préfixe-suffixe pour la position précédente
        #si les caractères ne correspondent pas, on recule en utilisant les valeurs précédentes de pi
        while j > 0 and s[i] != s[j]:
            j = pi[j-1]
        
        #si les caractères correspondent, on étend le préfixe
        if s[i] == s[j]:
            j += 1
        
        pi[i] = j
        
    return pi


def kmp_search(text, pattern):
    #on concatène le motif et le texte avec un séparateur
    s = pattern + "#" + text
    #calcul de la fonction préfixe pour toute la chaîne
    pi = prefix_function(s)
    
    result = [] #liste des positions où le motif apparaît
    m = len(pattern) #longueur du motif
    
    #on parcourt le tableau pi
    for i in range(len(s)):
        #si pi[i] = m, cela signifie que le motif a été trouvé
        if pi[i] == m:
            #position de début du motif dans le texte
            result.append(i - 2*m)
    
    return result
````

## 4.4 Z-algorithm
L’algorithme Z calcule, pour chaque position \(i\), la longueur du plus long préfixe de la chaîne qui commence à la position \(i\).
On construit ainsi un tableau \(Z\) tel que :
\[
Z[i] = \text{longueur du plus long préfixe de la chaîne commençant en } i
\]

Cet algorithme permet de résoudre efficacement plusieurs problèmes liés aux chaînes de caractères, notamment la recherche de motifs.
L’algorithme Z est utilisé pour :

- la recherche de motifs dans un texte  
- la détection de répétitions dans une chaîne  
- le calcul de bordures de chaînes  
- l’analyse de structures répétitives  
- La complexité de l’algorithme est :
\[
O(n)
\]
où \(n\) est la longueur de la chaîne.

````python
def z_algorithm(s):
    n = len(s) #longueur de la chaîne
    z = [0]*n #tableau Z : z[i] = longueur du plus long préfixe
    l = r = 0
    
    for i in range(1,n): # on commence à 1 car z[0] n'est généralement pas utilisé
        #si i est à l'intérieur de la Z-box actuelle
        if i <= r:
            #on peut réutiliser l'information déjà calculée
            z[i] = min(r-i+1, z[i-l])
        
        #on essaie d'étendre la correspondance caractère par caractère
        while i+z[i] < n and s[z[i]] == s[i+z[i]]:
            z[i] += 1
        
        #si on dépasse r, on met à jour la nouvelle Z-box
        if i+z[i]-1 > r:
            l = i
            r = i+z[i]-1
    
    return z
````

## 6.1 Points, vecteurs, produits scalaire/vectoriel

Les notions de points et de vecteurs, ainsi que les produits scalaire et vectoriel, constituent les bases pour modéliser et analyser des objets géométriques dans le plan ou l’espace.

Dans un plan 2D, un point et un vecteur sont mathématiquement équivalents. Pour améliorer la réutilisabilité, nous définissons une classe avec surcharge d'opérateurs, s'inspirant de l'implémentation du Vecteur2D vue en cours .

Le produit scalaire est défini par $\vec{A} \cdot \vec{B} = x_1 x_2 + y_1 y_2$. Géométriquement, son signe reflète le cosinus de l'angle entre deux vecteurs : positif pour un angle aigu, négatif pour un angle obtus, et nul pour l'orthogonalité. Il est utile pour calculer les normes ou les projections.

Le produit vectoriel est l'âme de la géométrie algorithmique. En 2D, le signe de $x_1 y_2 - x_2 y_1$ détermine la rotation relative. Si $P \times Q > 0$, $Q$ est à gauche de $P$ (anti-horaire). C'est la base de tous les tests d'orientation.

La complexité temporelle des opérations de base est de $O(1)$, car le coût de calcul est constant quel que soit le nombre de points. La complexité spatiale est également de $O(1)$, ne nécessitant qu'une mémoire constante pour stocker les variables de coordonnées.

````python
class Point:
    """Classe de base pour les points et vecteurs"""
    __slots__ = ('x', 'y') # Optimisation de la mémoire

    def __init__(self, x, y): # Constructeur de la classe [cite: 353, 743, 823]
        self.x = x
        self.y = y

    def __sub__(self, other): # Surcharge de l'opérateur de soustraction [cite: 875]
        # Retourne un vecteur pointant de 'other' vers 'self'
        return Point(self.x - other.x, self.y - other.y)

    def __repr__(self): # Redéfinition de la méthode de conversion en chaîne [cite: 879, 880]
        return f"({self.x}, {self.y})"

def cross_product(a, b):
    """Calcul du produit vectoriel en 2D"""
    # Mesure l'aire signée du parallélogramme formé par les vecteurs a et b
    return a.x * b.y - a.y * b.x

def get_orientation(p, q, r):
    """
    Test d'orientation de trois points
    0 -> Colinéaire , 1 -> Horaire , 2 -> Anti-horaire 
    """
    val = cross_product(q - p, r - q)
    if val == 0: return 0
    return 1 if val < 0 else 2
````

## 6.2 Enveloppe convexe

L’enveloppe convexe d’un ensemble de points est le plus petit polygone convexe contenant tous les points, et peut être calculée efficacement à l’aide d’algorithmes dédiés.

Le problème de l'enveloppe convexe consiste à trouver le plus petit polygone convexe contenant tous les points. L'algorithme d'Andrew est privilégié en compétition pour sa clarté et sa performance. Il construit la frontière via un double balayage (coques inférieure et supérieure) avec une complexité de $O(n \log n)$.

L'algorithme trie d'abord les points par $x$, puis par $y$. Ensuite, on utilise une pile pour maintenir les points de la coque. Pour chaque point, tant que le nouveau point ne respecte pas la règle "anti-horaire", on retire le sommet de la pile. Cette approche gère parfaitement les points en double ou colinéaires.

````python
def convex_hull(points):
    """
    Calcul de l'enveloppe convexe (Andrew's Monotone Chain)
    Complexité : O(N log N) dû au tri initial
    """
    n = len(points)
    if n <= 2: return points
    
    # 1. Tri par coordonnées x (puis y en cas d'égalité)
    points.sort(key=lambda p: (p.x, p.y))
    
    def build_half_hull(sequence):
        hull = []
        for p in sequence:
            # Tant que les 3 derniers points ne forment pas un virage à gauche, on dépile
            while len(hull) >= 2 and get_orientation(hull[-2], hull[-1], p) != 2:
                hull.pop()
            hull.append(p)
        return hull

    # 2. Construction des coques inférieure et supérieure
    lower = build_half_hull(points)
    upper = build_half_hull(reversed(points))
    
    # Fusionner en omettant le dernier point de chaque coque (doublons)
    return lower[:-1] + upper[:-1]
````

## 6.3 Intersection de segments

Les algorithmes d’intersection de segments permettent de déterminer si deux segments se coupent, en s’appuyant sur des tests d’orientation et des comparaisons géométriques.

En compétition, pour vérifier si deux segments $P_1Q_1$ et $P_2Q_2$ se croisent, on privilégie le test d'enjambement basé sur le signe du produit vectoriel. Cette méthode détermine les positions relatives des extrémités par rapport à l'autre segment, offrant une robustesse maximale.

La logique se divise en deux : dans le cas général, l'intersection existe si $P_1, Q_1$ sont de part et d'autre de $P_2Q_2$ et inversement. Dans le cas colinéaire, si trois points sont alignés, on vérifie si le point appartient à l'intervalle des coordonnées $x$ et $y$ de l'autre segment.

Le processus de décision ne comporte qu'un nombre fixe d'opérations arithmétiques et de branchements conditionnels, sa complexité temporelle est donc de $O(1)$. Sa complexité spatiale est également de $O(1)$, ne nécessitant qu'un espace constant pour stocker les indicateurs de changement de direction intermédiaires.


````python
def is_on_segment(p, q, r):
    """Vérifie si q est sur le segment pr"""
    return min(p.x, r.x) <= q.x <= max(p.x, r.x) and \
           min(p.y, r.y) <= q.y <= max(p.y, r.y)

def do_intersect(p1, q1, p2, q2):
    """
    Vérifie si les segments p1q1 et p2q2 se coupent
    """
    o1 = get_orientation(p1, q1, p2)
    o2 = get_orientation(p1, q1, q2)
    o3 = get_orientation(p2, q2, p1)
    o4 = get_orientation(p2, q2, q1)

    # Cas général : les segments s'enjambent mutuellement
    if o1 != o2 and o3 != o4:
        return True

    # Cas particuliers : un point est colinéaire et situé sur l'autre segment
    if o1 == 0 and is_on_segment(p1, p2, q1): return True
    if o2 == 0 and is_on_segment(p1, q2, q1): return True
    if o3 == 0 and is_on_segment(p2, p1, q2): return True
    if o4 == 0 and is_on_segment(p2, q1, q2): return True

    return False
````

## 6.4 Aire d'un polygone


Le calcul de l'aire d'un polygone quelconque (concave ou convexe) est une exigence courante en compétition. La méthode la plus robuste est la formule de la dentelle (Shoelace Formula). L'idée centrale consiste à décomposer le polygone en plusieurs triangles ayant l'origine pour sommet et à utiliser le produit vectoriel pour calculer la somme des aires signées de ces triangles. Comme le produit vectoriel porte un signe, les parties chevauchantes à l'intérieur du polygone s'annulent automatiquement, et la valeur absolue du résultat final donne l'aire totale.

Lors de l'implémentation, nous parcourons les sommets du polygone dans l'ordre et calculons le produit vectoriel des vecteurs formés par deux points consécutifs. La formule est $Aire = \frac{1}{2} | \sum_{i=1}^{n} (x_i y_{i+1} - x_{i+1} y_i) |$, avec la convention que le dernier point est relié au premier. Cette méthode est non seulement minimaliste en termes de code, mais elle traite également parfaitement tout polygone complexe non auto-intersectant.

La complexité temporelle est $O(N)$ car un seul parcours linéaire des $N$ sommets est nécessaire.La complexité spatiale est $O(1)$ car seule une variable auxiliaire constante est requise pour stocker l'aire accumulée.

````python
def get_polygon_area(points):
    """
    Calcul de l'aire par la formule de la dentelle
    Points: Liste des sommets (objets Point)
    """
    area = 0
    n = len(points)
    for i in range(n):
        p1 = points[i]
        p2 = points[(i + 1) % n] # Fermeture automatique au premier point
        # Accumulation de la contribution du produit vectoriel
        area += cross_product(p1, p2)
    
    # Valeur absolue divisée par 2
    return abs(area) / 2.0
````

## 6.5 Point dans un polygone

Déterminer si un point se trouve à l'intérieur d'un polygone est l'un des problèmes les plus classiques. L'algorithme le plus universel est celui du lancer de rayons (Ray Casting Algorithm). La logique est intuitive : on trace une demi-droite partant du point cible vers l'infini (généralement horizontalement vers la droite) et on compte le nombre d'intersections avec les bords du polygone. Si le nombre est impair, le point est à l'intérieur ; s'il est pair, il est à l'extérieur.

Pour garantir la robustesse, il faut gérer les cas où le point est sur un bord ou si le rayon passe par un sommet. On convient généralement que le rayon ne croise que les segments "fermés en bas et ouverts en haut" pour éviter les doubles comptages.

La complexité temporelle est $O(N)$ car chaque arête du polygone doit être inspectée pour détecter une intersection.La complexité spatiale est $O(1)$ car seule une variable booléenne est nécessaire pour suivre l'état de la parité.

````python
def is_inside_polygon(poly, p):
    """
    Test du point dans le polygone via Ray Casting
    """
    n = len(poly)
    inside = False
    for i in range(n):
        p1 = poly[i]
        p2 = poly[(i + 1) % n]
        
        # Vérifier si le point est sur le bord
        if get_orientation(p1, p2, p) == 0 and is_on_segment(p1, p, p2):
            return True
            
        # 2. Cœur de la logique du lancer de rayons
        # Vérifier si le rayon traverse le segment actuel
        if ((p1.y > p.y) != (p2.y > p.y)) and \
           (p.x < (p2.x - p1.x) * (p.y - p1.y) / (p2.y - p1.y) + p1.x):
            inside = not inside # Inversion parité
            
    return inside
````
