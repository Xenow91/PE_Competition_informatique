**1 Structures de données fondamentales**

**1.1 Structures linéaires (piles, files, listes)**\
Les structures linéaires sont des structures de données dans lesquelles les éléments sont organisés de manière séquentielle, selon un ordre bien défini.

**1.1.1 Piles**\
Une pile est une structure de données linéaire qui suit le principe LIFO (Last In, First
Out)

```python
class Piles :
    def __init__ ( self ) :
        self . data = []
    def push ( self , x ) :
        self . data . append ( x )

    def pop ( self ) :
        if self . data :
            return self . data . pop ()
        return None

    def is_empty ( self ) :
        return len ( self . data ) == 0
```

**1.1.2 Files**\
Une file est une structure de données linéaire qui suit le principe FIFO (First In, First
Out)

```python
class File :
    def __init__ ( self ) :
        self . data = []
        
    def enqueue ( self , x ):
        self . data . append ( x )

    def dequeue ( self ) :
        if self . data :
            return self . data . pop (0)
        return None

    def is_empty ( self ):
        return len ( self . data ) == 0
```

**1.2 Arbres (BST, segment trees, Fenwick trees/BIT)**\
Les arbres sont des structures de données hiérarchiques, organisées sous forme de
nœuds reliés entre eux, permettant une organisation non linéaire des données.

**1.2.1 BST (ou ABR : arbre binaire de recherche )**\
Un arbre de recherche binaire est une structure de données arborescente dans laquelle,
pour chaque nœud, les éléments du sous-arbre gauche sont inférieurs à la valeur du nœud,
et ceux du sous-arbre droit sont supérieurs.

```python
class BST:
    def __init__(self):
        self.data = None

    def add(self, x):
        if self.data == None :
            self.data = x
            self.left = BST()
            self.right = BST()
        elif x < self.data :
            if self.left.data == None :
                self.left.add(x)
            else : 
                self.left.add(x)
        else :
            if self.right.data == None :
                bst = BST()
                bst.add(x)
                self.right = bst
            else : 
                self.right.add(x)
                
    def search (self,x) :
        if self.data == None or self == None:
            return False
        elif self.data == x :
            return True
        elif x < self.data :
            return self.left.search(x)
        else : 
            return self.right.search(x)
            
    def delete (self,x) :
        if self.data == None :
            return
        elif self.data == x :
            if self.left.data == None and self.right.data == None :
                self.data = None
            elif self.left.data == None :
                self.data = self.right.data
                self.left = self.right.left
                self.right = self.right.right
            elif self.right.data == None :
                self.data = self.left.data
                self.right = self.left.right
                self.left = self.left.left
            else :
                min_right = self.right.min()
                self.data = min_right
                self.right.delete(min_right)
        elif x < self.data :
            return self.left.delete(x)
        else : 
            return self.right.delete(x)
      
    def min (self) :
        if self.left.data == None:
            return self.data
        else :
            return self.left.min()
            
    def max (self) :
        if self.right.data == None:
            return self.data
        else :
            return self.right.max()

    def is_empty(self):
        return self.data == None
```

**1.3 Files de priorité et heaps**\
Une file de priorité est une structure de données permettant d’accéder rapidement à
l’élément de priorité minimale ou maximale, indépendamment de l’ordre d’insertion. Elle
est généralement implémentée à l’aide d’un tas, notamment un tas binaire, qui garantit
des opérations d’insertion et d’extraction en temps logarithmique.

```python
class priorityQueue:
    def __init__(self):
        self.heap = []

    def parent(self, index):
        return (index - 1) // 2
    
    def leftChild(self, index):
        return 2 * index + 1
    
    def rightChild(self, index):
        return 2 * index + 2
    
    def isEmpty(self):
        return len(self.heap) == 0

    def insert(self, item): # item is a tuple (priority, value)
        self.heap.append(item)
        self.heapifyUp(len(self.heap) - 1)

    def heapifyUp(self, index):
        parentIndex = self.parent(index)
        if index > 0 and self.heap[parentIndex][0] > self.heap[index][0]:
            self.heap[parentIndex], self.heap[index] = self.heap[index], self.heap[parentIndex]
            self.heapifyUp(parentIndex)
    
    def heapifyDown(self, index):
        leftChildIndex = self.leftChild(index)
        rightChildIndex = self.rightChild(index)
        smallestIndex = index

        if leftChildIndex < len(self.heap) and self.heap[leftChildIndex][0] < self.heap[smallestIndex][0]:
            smallestIndex = leftChildIndex

        if rightChildIndex < len(self.heap) and self.heap[rightChildIndex][0] < self.heap[smallestIndex][0]:
            smallestIndex = rightChildIndex

        if smallestIndex != index:
            self.heap[index], self.heap[smallestIndex] = self.heap[smallestIndex], self.heap[index]
            self.heapifyDown(smallestIndex)

    def pop(self):
        if self.isEmpty():
            raise IndexError("La file de priorité est vide")
        minItem = self.heap[0]
        lastItem = self.heap.pop()
        if not self.isEmpty():
            self.heap[0] = lastItem
            self.heapifyDown(0)
        return minItem 
```

**1.4 Union-Find**\
L’Union-Find, également appelé structure d’ensembles disjoints, est une structure de
données permettant de gérer efficacement une partition d’éléments en ensembles disjoints.
Elle supporte principalement deux opérations : la recherche de l’ensemble auquel appar-
tient un élément (find) et la fusion de deux ensembles (union). Grâce à l’utilisation de la
compression de chemin et de l’union par rang, ces opérations s’effectuent en temps quasi
constant.

**1.5 Tries et arbres de suffixes**\
Les tries et les arbres de suffixes sont des structures de données arborescentes spécialisées pour la manipulation efficace de chaînes de caractères. Un trie permet de représenter un ensemble de mots en factorisant leurs préfixes communs, tandis qu’un arbre de suffixes représente l’ensemble des suffixes d’une chaîne, facilitant les recherches de motifs et les opérations sur les sous-chaînes en temps linéaire par rapport à la taille de la chaîne.

**2.1 Parcours (BFS,DFS)**\
Les algorithmes de parcours permettent d’explorer systématiquement l’ensemble des
sommets d’un graphe, soit en largeur (BFS), soit en profondeur (DFS).

```python
# Parcours en largeur (Breadth-First Search) 
def bfs(graph, start):  # Le graphe est represente sous forme d'une liste d'adjacence
    visited = [False for i in range(len(graph))] 
    queue = []  

    visited[start] = True  #
    queue.append(start)  

    while queue:  # Tant que la file d'attente n'est pas vide
        vertex = queue.pop(0)  
        print(vertex) 

        for neighbor in graph[vertex]: 
            if not visited[neighbor]:  
                visited[neighbor] = True
                queue.append(neighbor) 

# Parcours en profondeur (Depth-First Search)
def dfs(graph, start):
    visited = [False for i in range(len(graph))] 
    stack = []  

    visited[start] = True  
    stack.append(start)  

    while stack:  # Tant que la pile n'est pas vide
        vertex = stack.pop()  
        print(vertex) 

        for neighbor in graph[vertex]: 
            if not visited[neighbor]:  
                visited[neighbor] = True
                stack.append(neighbor)
                
```
