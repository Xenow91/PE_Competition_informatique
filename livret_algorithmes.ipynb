# 1 Structures de données fondamentales

## 1.1 Structures linéaires (piles, files, listes)
Les structures linéaires sont des structures de données dans lesquelles les éléments sont organisés de manière séquentielle, selon un ordre bien défini.

### 1.1.1 Piles
Une pile est une structure de données linéaire qui suit le principe LIFO (Last In, First
Out)

```python
class Piles :
    def __init__ ( self ) :
        self . data = []
    def push ( self , x ) :
        self . data . append ( x )

    def pop ( self ) :
        if self . data :
            return self . data . pop ()
        return None

    def is_empty ( self ) :
        return len ( self . data ) == 0
```

### 1.1.2 Files
Une file est une structure de données linéaire qui suit le principe FIFO (First In, First
Out)

```python
class File :
    def __init__ ( self ) :
        self . data = []
        
    def enqueue ( self , x ):
        self . data . append ( x )

    def dequeue ( self ) :
        if self . data :
            return self . data . pop (0)
        return None

    def is_empty ( self ):
        return len ( self . data ) == 0
```

Une liste est une structure de données linéaire permettant un accès séquentiel aux
éléments.

## 1.2 Arbres (BST, segment trees, Fenwick trees/BIT)
Les arbres sont des structures de données hiérarchiques, organisées sous forme de

### 1.2.1 BST (ou ABR : arbre binaire de recherche )
Un arbre de recherche binaire est une structure de données arborescente dans laquelle,
pour chaque nœud, les éléments du sous-arbre gauche sont inférieurs à la valeur du nœud,
et ceux du sous-arbre droit sont supérieurs.

```python
class BST:
    def __init__(self):
        self.data = None

    def add(self, x):
        if self.data == None :
            self.data = x
            self.left = BST()
            self.right = BST()
        elif x < self.data :
            if self.left.data == None :
                self.left.add(x)
            else : 
                self.left.add(x)
        else :
            if self.right.data == None :
                bst = BST()
                bst.add(x)
                self.right = bst
            else : 
                self.right.add(x)
                
    def search (self,x) :
        if self.data == None or self == None:
            return False
        elif self.data == x :
            return True
        elif x < self.data :
            return self.left.search(x)
        else : 
            return self.right.search(x)
            
    def delete (self,x) :
        if self.data == None :
            return
        elif self.data == x :
            if self.left.data == None and self.right.data == None :
                self.data = None
            elif self.left.data == None :
                self.data = self.right.data
                self.left = self.right.left
                self.right = self.right.right
            elif self.right.data == None :
                self.data = self.left.data
                self.right = self.left.right
                self.left = self.left.left
            else :
                min_right = self.right.min()
                self.data = min_right
                self.right.delete(min_right)
        elif x < self.data :
            return self.left.delete(x)
        else : 
            return self.right.delete(x)
      
    def min (self) :
        if self.left.data == None:
            return self.data
        else :
            return self.left.min()
            
    def max (self) :
        if self.right.data == None:
            return self.data
        else :
            return self.right.max()

    def is_empty(self):
        return self.data == None
```

## 1.3 Files de priorité et heaps
Une file de priorité est une structure de données permettant d’accéder rapidement à
l’élément de priorité minimale ou maximale, indépendamment de l’ordre d’insertion. Elle
est généralement implémentée à l’aide d’un tas, notamment un tas binaire, qui garantit
des opérations d’insertion et d’extraction en temps logarithmique.

```python
class priorityQueue:
    def __init__(self):
        self.heap = []

    def parent(self, index):
        return (index - 1) // 2
    
    def leftChild(self, index):
        return 2 * index + 1
    
    def rightChild(self, index):
        return 2 * index + 2
    
    def isEmpty(self):
        return len(self.heap) == 0

    def insert(self, item): # item is a tuple (priority, value)
        self.heap.append(item)
        self.heapifyUp(len(self.heap) - 1)

    def heapifyUp(self, index):
        parentIndex = self.parent(index)
        if index > 0 and self.heap[parentIndex][0] > self.heap[index][0]:
            self.heap[parentIndex], self.heap[index] = self.heap[index], self.heap[parentIndex]
            self.heapifyUp(parentIndex)
    
    def heapifyDown(self, index):
        leftChildIndex = self.leftChild(index)
        rightChildIndex = self.rightChild(index)
        smallestIndex = index

        if leftChildIndex < len(self.heap) and self.heap[leftChildIndex][0] < self.heap[smallestIndex][0]:
            smallestIndex = leftChildIndex

        if rightChildIndex < len(self.heap) and self.heap[rightChildIndex][0] < self.heap[smallestIndex][0]:
            smallestIndex = rightChildIndex

        if smallestIndex != index:
            self.heap[index], self.heap[smallestIndex] = self.heap[smallestIndex], self.heap[index]
            self.heapifyDown(smallestIndex)

    def pop(self):
        if self.isEmpty():
            raise IndexError("La file de priorité est vide")
        minItem = self.heap[0]
        lastItem = self.heap.pop()
        if not self.isEmpty():
            self.heap[0] = lastItem
            self.heapifyDown(0)
        return minItem 
```

Les algorithmes de parcours permettent d’explorer systématiquement l’ensemble des
sommets d’un graphe, soit en largeur (BFS), soit en profondeur (DFS).

```python
# Parcours en largeur (Breadth-First Search) 
def bfs(graph, start):  # Le graphe est represente sous forme d'une liste d'adjacence
    visited = [False for i in range(len(graph))] 
    queue = []  

    visited[start] = True  #
    queue.append(start)  

    while queue:  # Tant que la file d'attente n'est pas vide
        vertex = queue.pop(0)  
        print(vertex) 

        for neighbor in graph[vertex]: 
            if not visited[neighbor]:  
                visited[neighbor] = True
                queue.append(neighbor) 

# Parcours en profondeur (Depth-First Search)
def dfs(graph, start):
    visited = [False for i in range(len(graph))] 
    stack = []  

    visited[start] = True  
    stack.append(start)  

    while stack:  # Tant que la pile n'est pas vide
        vertex = stack.pop()  
        print(vertex) 

        for neighbor in graph[vertex]: 
            if not visited[neighbor]:  
                visited[neighbor] = True
                stack.append(neighbor)
                
```

# 3 Programmation dynamique

## 3.1 Introduction à la DP : Top-Down vs Bottom-Up

La Programmation Dynamique (DP) est une méthode d'optimisation mathématique et algorithmique. Elle s'applique aux problèmes qui présentent deux caractéristiques fondamentales :
1. **Sous-structure optimale :** La solution optimale du problème global peut être construite à partir des solutions optimales de ses sous-problèmes.
2. **Chevauchement des sous-problèmes :** L'algorithme naïf résout les mêmes sous-problèmes de multiples fois.

La DP résout ce problème de redondance en stockant les résultats des sous-problèmes (mémoïsation ou tabulation) pour ne les calculer qu'une seule fois. Il existe deux approches principales pour implémenter une DP :

| Caractéristique | Top-Down (Récursif + Mémoïsation) | Bottom-Up (Itératif + Tabulation) |
| :--- | :--- | :--- |
| **Principe** | On part du problème global et on descend récursivement. On stocke le résultat de chaque appel. | On part des cas de base et on remplit un tableau itérativement jusqu'au problème global. |
| **Avantage principal** | Ne calcule que les états strictement nécessaires. Transition très naturelle à écrire. | Pas d'overhead lié à la pile d'appels récursifs. Souvent plus rapide en temps d'exécution constant. |
| **Inconvénient** | Risque de dépassement de la pile d'appels (`RecursionError` en Python) sur de grandes entrées. | Oblige à calculer tous les états, même ceux qui pourraient être inaccessibles. |
| **Optimisation spatiale** | Difficile à mettre en place. | Très facile (ex: ne conserver que la ligne précédente dans une matrice 2D). |

**Exemple d'application : Nombre de chemins dans une grille**
*Problème :* Trouver le nombre de chemins uniques du point $(0, 0)$ au point $(M-1, N-1)$ dans une grille, en ne se déplaçant que vers la droite ou vers le bas. Certains obstacles bloquent le passage.

* **État :** $dp[i][j]$ représente le nombre de chemins pour atteindre la case $(i, j)$.
* **Transition :** Si la case $(i, j)$ n'est pas un obstacle : 
    $dp[i][j] = dp[i-1][j] + dp[i][j-1]$
* **Cas de base :** $dp[0][0] = 1$ (1 façon d'être au départ). La première ligne et la première collone n'ont qu'une seule provenance possible.

```python
# 1 indique une case libre, 0 indique un obstacle

def top_down(grid : list[list[int]]) -> int :
    grid = grid
    M = len(grid)
    N = len(grid[0]) if M > 0 else 0

    if M == 0 or N == 0 or grid[0][0] == 0:
            return 0
    
    memo = [[-1] * N for _ in range(M)]

    def dp(i: int, j: int) -> int:
            # Hors limites ou obstacle
            if i < 0 or j < 0 or grid[i][j] == 0:
                return 0
            # Cas de base
            if i == 0 and j == 0:
                return 1
            # Retourner si déjà calculé
            if memo[i][j] != -1:
                return memo[i][j]
                
            # Transition : venant du haut + venant de la gauche
            memo[i][j] = dp(i - 1, j) + dp(i, j - 1)
            return memo[i][j]
            
    return dp(M - 1, N - 1)
```

Analysons maintenant les performances de cet algorithme. On a une complexité en temps et en espace en $O(M*N)$. 

```python
# 1 indique une case libre, 0 indique un obstacle

def bottom_up(grid : list[list[int]]) -> int :
    grid = grid
    M = len(grid)
    N = len(grid[0]) if M > 0 else 0

    if M == 0 or N == 0 or grid[0][0] == 0:
            return 0
    
    memo = [[0] * N for _ in range(M)]
    # On initialise la première ligne et première collone de notre tableau
    for i in range (M):
        if grid[i][0]==1:
            memo[i][0]=1
        else:
             break
    for j in range (N):
        if grid[0][j]==1:
            memo[0][j]=1
        else:
            break
    
    # On effectue la récurence 
    for i in range (1,M):
          for j in range(1,N):
                if grid[i][j]==1:
                      memo[i][j]=memo[i-1][j]+memo[i][j-1]

    return memo[M-1][N-1]
```

La complexité est la même que pour l'algorithme précédent. En revanche comme il ne fait pas d'appel récursif il n'y a aucun risque de dépasssement de la pile d'appel.

## 3.2 Programmation Dynamique Classique

### 3.2.1 Le problème du Sac à Dos (0/1 Knapsack)

Le problème classique du sac à dos (0/1 car un objet est soit pris, soit laissé) consiste à maximiser la valeur totale des objets placés dans un sac de capacité limitée, sans la dépasser. 

**Modélisation Mathématique**
Soit un ensemble de $N$ objets, chacun ayant un poids $w_i$ et une valeur $v_i$. La capacité maximale du sac est $W$.

* **État :** $dp[i][j]$ représente la valeur maximale que l'on peut obtenir en choisissant parmi les $i$ premiers objets, avec une capacité exacte ou inférieure à $j$.
* **Transition :** Pour l'objet $i$, deux choix s'offrent à nous (le prendre ou l'ignorer).
  Si $j \ge w_i$ (l'objet rentre dans le sac) :
  $$dp[i][j] = \max(dp[i-1][j], dp[i-1][j-w_i] + v_i)$$
  Sinon (l'objet est trop lourd) :
  $$dp[i][j] = dp[i-1][j]$$
* **Cas de base :** $dp[0][j] = 0$ pour tout $j \in [0, W]$. Zéro objet donne une valeur de zéro.



**Optimisation Spatiale :**
La relation de récurrence montre que la ligne $i$ ne dépend que de la ligne $i-1$. Il n'est donc pas nécessaire de stocker une matrice entière de taille $N \times W$. Un simple tableau 1D de taille $W+1$ suffit, à condition de le parcourir **à l'envers** (de $W$ vers $w_i$) pour ne pas écraser les données de l'itération précédente prématurément.

* **Complexité temporelle :** $\mathcal{O}(N \cdot W)$
* **Complexité spatiale (Optimisée) :** $\mathcal{O}(W)$

```python
def Knapsack(capacity: int, weights: list[int], values: list[int]) -> int:
    n = len(weights)
    if n == 0 or capacity <= 0:
        return 0
            
    # Initialisation du tableau 1D pour l'optimisation spatiale
    dp = [0] * (capacity + 1)
     
    # Remplissage itératif (Bottom-Up)
    for i in range(n):
        weight = weights[i]
        value = values[i]
        
        # Parcours inversé crucial pour garantir que chaque objet 
        # n'est pris qu'une seule fois (0/1 Knapsack)
        for w in range(capacity, weight - 1, -1):
            dp[w] = max(dp[w], dp[w - weight] + value)
    return dp[capacity]
```

Ici dans l'équation de récurrence 1D on utilises bien "la case du dessus" en ayant $dp[w]$ comme argument dans la fonction max.

### 3.2.2 Plus Longue Sous-séquence Croissante (LIS)

Le problème de la Plus Longue Sous-séquence Croissante (Longest Increasing Subsequence) consiste à trouver la longueur de la plus longue sous-séquence d'un tableau donné telle que tous les éléments de cette sous-séquence soient triés par ordre strictement croissant.

**1. L'approche classique (Approche $\mathcal{O}(N^2)$)**
* **État :** $dp[i]$ représente la longueur de la LIS qui se termine exactement à l'index $i$.
* **Transition :** Pour chaque élément $i$, on regarde tous les éléments précédents $j$ (avec $j < i$). Si l'élément en $j$ est strictement inférieur à l'élément en $i$, on peut étendre la séquence :
  $$dp[i] = \max(dp[i], dp[j] + 1) \quad \text{pour tout } j < i \text{ tel que } arr[j] < arr[i]$$
* **Cas de base :** $dp[i] = 1$ pour tout $i$, car chaque élément forme au minimum une sous-séquence de taille 1.

**2. L'approche optimale par Dichotomie (Approche $\mathcal{O}(N \log N)$)**
Pour rendre l'algorithme scalable sur de grandes entrées ($N \ge 10^5$), on change la définition de l'état en utilisant le principe du *Patience Sorting*.

* **Nouvel État :** On maintient un tableau `tails` où `tails[k]` stocke la **plus petite valeur finale** d'une sous-séquence croissante de longueur $k+1$.
* **Propriété clé :** Le tableau `tails` sera toujours strictement croissant. Cela permet d'utiliser une recherche dichotomique (Binary Search) pour trouver rapidement où insérer ou remplacer un élément, réduisant la transition de $\mathcal{O}(N)$ à $\mathcal{O}(\log N)$.

```python

def LIS(arr : list[int]) -> int :
    if not arr:
        return 0
    
    # tails[i] stocke la plus petite valeur de fin pour une LIS de longueur i+1
    tails = []

    for x in arr:
        # Implémentation manuelle de la recherche dichotomique (lower bound)
        # Cherche le premier index 'left' tel que tails[left] >= x
        left = 0
        right = len(tails)
            
        while left < right:
            mid = (left + right) // 2
            if tails[mid] < x:
                left = mid + 1
            else:
                right = mid
            
        idx = left

        # Si x est strictement plus grand que tous les éléments, 
        # il prolonge la plus longue séquence actuelle.
        if idx == len(tails):
            tails.append(x)
        # Sinon, il remplace une valeur de fin existante pour optimiser
        # le potentiel des futurs éléments.
        else:
            tails[idx] = x

    return len(tails)
```

Cette implémentation repose sur une redéfinition de l'état. Plutôt que de calculer naïvement la longueur maximale pour chaque indice, nous maintenons un tableau d'états $tails$.

Dans ce tableau, $tails[i]$ stocke la plus petite valeur finale connue pour une sous-séquence croissante de longueur $i + 1$. Maintenir ces valeurs terminales aussi petites que possible est la clé de l'algorithme : cela maximise les chances de pouvoir raccorder et prolonger la séquence avec les éléments futurs.

Par construction, le tableau tails maintient une propriété de stricte croissance. Cette monotonie nous permet de remplacer le classique parcours linéaire (en $O(N)$) par une recherche dichotomique (en $O(\log N)$) pour traiter chaque nouvel élément $x$ :
* **Extension :** Si $x$ est strictement supérieur au dernier élément de tails, il prolonge la plus longue sous-séquence connue. Nous l'ajoutons à la fin du tableau.
* **Optimisation locale :** Sinon, la dichotomie identifie le premier élément dans tails qui est supérieur ou égal à $x$, et le remplace. Cette opération ne rallonge pas la séquence immédiatement, mais elle "abaisse" la valeur finale d'une sous-séquence existante. Cela améliore son potentiel d'extension pour les itérations suivantes.

### 3.2.3 Plus Longue Sous-séquence Commune (LCS)

Le problème de la Plus Longue Sous-séquence Commune consiste à déterminer la longueur de la plus longue séquence de caractères qui apparaît dans le même ordre d'apparition dans deux chaînes distinctes, sans être nécessairement contiguë.

**Modélisation Mathématique :**
Soient deux chaînes $S_1$ de taille $N$ et $S_2$ de taille $M$.

* **État :** $dp[i][j]$ représente la longueur de la LCS pour les préfixes $S_1[0..i-1]$ et $S_2[0..j-1]$.
* **Transition :** Pour chaque paire de caractères, l'approche se divise en deux cas de figure :
  * Si les caractères correspondent ($S_1[i-1] == S_2[j-1]$), on prolonge la séquence commune trouvée jusqu'aux indices précédents :
    $$dp[i][j] = dp[i-1][j-1] + 1$$
  * Si les caractères diffèrent, la séquence ne peut pas être prolongée. On conserve alors le maximum des sous-séquences communes trouvées en ignorant le caractère actuel de l'une ou l'autre des chaînes :
    $$dp[i][j] = \max(dp[i-1][j], dp[i][j-1])$$
* **Cas de base :** $dp[0][j] = 0$ et $dp[i][0] = 0$. Une chaîne vide n'a aucune sous-séquence commune avec une autre chaîne.

**Complexité :**
* **Temporelle :** $\mathcal{O}(N \cdot M)$
* **Spatiale (Optimisée) :** $\mathcal{O}(\min(N, M))$ en ne conservant en mémoire que la ligne précédente lors de l'itération. *Attention : la matrice complète $\mathcal{O}(N \cdot M)$ reste nécessaire si l'objectif est de reconstruire (backtracker) la chaîne exacte et pas seulement d'obtenir sa longueur.*

```python
def longest_common_subsequence(s1: str, s2: str) -> int:
    """
    Calcule la longueur de la Plus Longue Sous-séquence Commune (LCS).
    Implémente une optimisation spatiale stricte.
    """
    # Optimisation spatiale : on s'assure que s2 est toujours la chaîne 
    # la plus courte pour allouer le plus petit tableau possible en mémoire.
    if len(s1) < len(s2):
        s1, s2 = s2, s1
        
    n, m = len(s1), len(s2)
    
    # dp représente la ligne i-1 de la matrice de programmation dynamique
    dp = [0] * (m + 1)
    curr_dp = [0] * (m + 1)

    for i in range(1, n + 1):
        # curr_dp représente la ligne i en cours de calcul
        curr_dp[0]=0
        for j in range(1, m + 1):
            if s1[i - 1] == s2[j - 1]:
                curr_dp[j] = dp[j - 1] + 1
            else:
                if curr_dp[j - 1] > dp[j]: 
                    curr_dp[j] = curr_dp[j - 1]
                else :
                    curr_dp[j]=dp[j]
        # Le calcul de la ligne courante est terminé, elle devient la ligne précédente
        dp = curr_dp
        
    return dp[m]
```

**Logique de l'optimisation spatiale :**
L'équation de transition mathématique montre que pour calculer la valeur de la case `dp[i][j]`, l'algorithme n'a besoin que de consulter la case située juste au-dessus (`dp[i-1][j]`), la case à sa gauche (`dp[i][j-1]`), et la case en diagonale (`dp[i-1][j-1]`). 

Il est donc inutile de stocker l'historique de toutes les lignes précédentes. En maintenant uniquement deux tableaux unidimensionnels — un pour la ligne précédente (`dp`) et un pour la ligne en cours d'évaluation (`curr_dp`) — l'empreinte mémoire (footprint) est drastiquement réduite de $\mathcal{O}(N \cdot M)$ à $\mathcal{O}(\min(N, M))$. Le "swap" final (`dp = curr_dp`) s'effectue simplement par réaffectation de référence sans coût de copie supplémentaire, garantissant une exécution performante.

## 3.3 DP sur les intervalles (Interval DP)

La DP sur les intervalles est utilisée lorsque la solution d'un problème sur un segment continu `[l, r]` dépend de la combinaison des solutions optimales de ses propres sous-segments. La particularité architecturale de cette technique réside dans le flux de contrôle : on n'itère pas de gauche à droite, mais par **longueur croissante de l'intervalle**. On résout d'abord tous les segments de taille 1, puis 2, etc., jusqu'à la taille $N$.

### Exemple classique : Multiplication d'une chaîne de matrices
Le problème consiste à trouver l'ordre optimal (le parenthésage) pour multiplier une suite de matrices afin de minimiser le nombre total d'opérations scalaires.

**Modélisation Mathématique**
Soit un tableau de dimensions $P$ de taille $N+1$, où la matrice $A_i$ a pour dimension $P[i-1] \times P[i]$.

* **État :** $dp[i][j]$ représente le coût minimum pour multiplier la sous-chaîne de matrices allant de l'indice $i$ à $j$ (inclus).
* **Transition :** Pour trouver le coût optimal sur le segment $[i, j]$, on teste tous les points de coupure possibles $k$ (avec $i \le k < j$).
  $$dp[i][j] = \min_{i \le k < j} \left( dp[i][k] + dp[k+1][j] + P[i-1] \cdot P[k] \cdot P[j] \right)$$
* **Cas de base :** $dp[i][i] = 0$ pour tout $i$. Le coût pour multiplier une matrice par elle-même n'existe pas.

**Complexité**
* **Temporelle :** $\mathcal{O}(N^3)$ (deux boucles pour définir les bornes de l'intervalle, une boucle interne pour chercher le point de coupure optimal).
* **Spatiale :** $\mathcal{O}(N^2)$ (la matrice des états).

### Exemple illustratif : L'impact critique du parenthésage

Considérons la multiplication de trois matrices $A$, $B$ et $C$ avec les dimensions suivantes :
* **$A$** : $10 \times 100$
* **$B$** : $100 \times 5$
* **$C$** : $5 \times 50$

Pour rappel, multiplier une matrice de dimension $p \times q$ par une matrice $q \times r$ requiert exactement $p \cdot q \cdot r$ opérations scalaires. Le produit matriciel étant associatif, nous pouvons calculer $A \cdot B \cdot C$ de deux manières. Voyons l'impact sur les performances :

**Option 1 : Parenthésage $(A \cdot B) \cdot C$**
1. Calcul de $(A \cdot B)$ : $10 \cdot 100 \cdot 5 = 5000$ opérations. La matrice résultante est de dimension $10 \times 5$.
2. Multiplication par $C$ : $10 \cdot 5 \cdot 50 = 2500$ opérations.
3. **Coût total :** $5000 + 2500 = 7500$ opérations.

**Option 2 : Parenthésage $A \cdot (B \cdot C)$**
1. Calcul de $(B \cdot C)$ : $100 \cdot 5 \cdot 50 = 25000$ opérations. La matrice résultante est de dimension $100 \times 50$.
2. Multiplication par $A$ : $10 \cdot 100 \cdot 50 = 50000$ opérations.
3. **Coût total :** $25000 + 50000 = 75000$ opérations.

**Conclusion :**
Le second choix de parenthésage nécessite **10 fois plus d'opérations** (75 000 contre 7 500) pour aboutir strictement au même résultat mathématique. Cet écart de performance croît de manière exponentielle avec le nombre de matrices (nombre de Catalan), ce qui rend une recherche exhaustive en force brute totalement inenvisageable. C'est précisément pour cela que l'optimisation par la DP sur les intervalles est requise.

```python
def matrix_chain_multiplication(p: list[int]) -> int:
    """
    Calcule le coût minimum pour multiplier une chaîne de matrices.
    Le tableau p contient les dimensions : la matrice i a pour taille p[i] x p[i+1].
    """
    # Le nombre de matrices est la longueur du tableau des dimensions moins 1
    n = len(p) - 1
    if n <= 0:
        return 0
        
    # Initialisation de la matrice DP. 
    # dp[i][j] stockera le coût min pour multiplier les matrices de i à j.
    # On utilise l'infini comme valeur de base pour la minimisation.
    dp = [[float('inf')] * n for _ in range(n)]
    
    # Cas de base : le coût pour multiplier une matrice seule est de 0
    for i in range(n):
        dp[i][i] = 0
        
    # Workflow spécifique à l'Interval DP : itération sur la LONGUEUR du sous-segment
    # La longueur varie de 2 (plus petit calcul possible) à n (le problème global)
    for length in range(2, n + 1):
        for i in range(n - length + 1):
            j = i + length - 1
            
            # Recherche du point de coupure k optimal divisant [i, j] en [i, k] et [k+1, j]
            for k in range(i, j):
                # Le coût combine la résolution des deux sous-problèmes plus le coût 
                # de la multiplication des deux matrices résultantes.
                cost = dp[i][k] + dp[k + 1][j] + p[i] * p[k + 1] * p[j + 1]
                
                if cost < dp[i][j]:
                    dp[i][j] = cost
                    
    return dp[0][n - 1]
```

## 6.1 Points, vecteurs, produits scalaire/vectoriel
Les notions de points et de vecteurs, ainsi que les produits scalaire et vectoriel, constituent les bases pour modéliser et analyser des objets géométriques dans le plan ou l’espace.

Dans un plan 2D, un point et un vecteur sont mathématiquement équivalents. Pour améliorer la réutilisabilité, nous définissons une classe avec surcharge d'opérateurs, s'inspirant de l'implémentation du Vecteur2D vue en cours .

Le produit scalaire est défini par $\vec{A} \cdot \vec{B} = x_1 x_2 + y_1 y_2$. Géométriquement, son signe reflète le cosinus de l'angle entre deux vecteurs : positif pour un angle aigu, négatif pour un angle obtus, et nul pour l'orthogonalité. Il est utile pour calculer les normes ou les projections.

Le produit vectoriel est l'âme de la géométrie algorithmique. En 2D, le signe de $x_1 y_2 - x_2 y_1$ détermine la rotation relative. Si $P \times Q > 0$, $Q$ est à gauche de $P$ (anti-horaire). C'est la base de tous les tests d'orientation.

La complexité temporelle des opérations de base est de $O(1)$, car le coût de calcul est constant quel que soit le nombre de points. La complexité spatiale est également de $O(1)$, ne nécessitant qu'une mémoire constante pour stocker les variables de coordonnées.

```python
class Point:
    """Classe de base pour les points et vecteurs"""
    __slots__ = ('x', 'y') # Optimisation de la mémoire

    def __init__(self, x, y): # Constructeur de la classe [cite: 353, 743, 823]
        self.x = x
        self.y = y

    def __sub__(self, other): # Surcharge de l'opérateur de soustraction [cite: 875]
        # Retourne un vecteur pointant de 'other' vers 'self'
        return Point(self.x - other.x, self.y - other.y)

    def __repr__(self): # Redéfinition de la méthode de conversion en chaîne [cite: 879, 880]
        return f"({self.x}, {self.y})"

def cross_product(a, b):
    """Calcul du produit vectoriel en 2D"""
    # Mesure l'aire signée du parallélogramme formé par les vecteurs a et b
    return a.x * b.y - a.y * b.x

def get_orientation(p, q, r):
    """
    Test d'orientation de trois points
    0 -> Colinéaire , 1 -> Horaire , 2 -> Anti-horaire 
    """
    val = cross_product(q - p, r - q)
    if val == 0: return 0
    return 1 if val < 0 else 2
```

## 6.2 Enveloppe convexe
L’enveloppe convexe d’un ensemble de points est le plus petit polygone convexe contenant tous les points, et peut être calculée efficacement à l’aide d’algorithmes dédiés.

Le problème de l'enveloppe convexe consiste à trouver le plus petit polygone convexe contenant tous les points. L'algorithme d'Andrew est privilégié en compétition pour sa clarté et sa performance. Il construit la frontière via un double balayage (coques inférieure et supérieure) avec une complexité de $O(n \log n)$.

L'algorithme trie d'abord les points par $x$, puis par $y$. Ensuite, on utilise une pile pour maintenir les points de la coque. Pour chaque point, tant que le nouveau point ne respecte pas la règle "anti-horaire", on retire le sommet de la pile. Cette approche gère parfaitement les points en double ou colinéaires.

```python
def convex_hull(points):
    """
    Calcul de l'enveloppe convexe (Andrew's Monotone Chain)
    Complexité : O(N log N) dû au tri initial
    """
    n = len(points)
    if n <= 2: return points
    
    # 1. Tri par coordonnées x (puis y en cas d'égalité)
    points.sort(key=lambda p: (p.x, p.y))
    
    def build_half_hull(sequence):
        hull = []
        for p in sequence:
            # Tant que les 3 derniers points ne forment pas un virage à gauche, on dépile
            while len(hull) >= 2 and get_orientation(hull[-2], hull[-1], p) != 2:
                hull.pop()
            hull.append(p)
        return hull

    # 2. Construction des coques inférieure et supérieure
    lower = build_half_hull(points)
    upper = build_half_hull(reversed(points))
    
    # Fusionner en omettant le dernier point de chaque coque (doublons)
    return lower[:-1] + upper[:-1]
```

## 6.3 Intersection de segments
Les algorithmes d’intersection de segments permettent de déterminer si deux segments se coupent, en s’appuyant sur des tests d’orientation et des comparaisons géométriques.

En compétition, pour vérifier si deux segments $P_1Q_1$ et $P_2Q_2$ se croisent, on privilégie le test d'enjambement basé sur le signe du produit vectoriel. Cette méthode détermine les positions relatives des extrémités par rapport à l'autre segment, offrant une robustesse maximale.

La logique se divise en deux : dans le cas général, l'intersection existe si $P_1, Q_1$ sont de part et d'autre de $P_2Q_2$ et inversement. Dans le cas colinéaire, si trois points sont alignés, on vérifie si le point appartient à l'intervalle des coordonnées $x$ et $y$ de l'autre segment.

Le processus de décision ne comporte qu'un nombre fixe d'opérations arithmétiques et de branchements conditionnels, sa complexité temporelle est donc de $O(1)$. Sa complexité spatiale est également de $O(1)$, ne nécessitant qu'un espace constant pour stocker les indicateurs de changement de direction intermédiaires.

```python
def is_on_segment(p, q, r):
    """Vérifie si q est sur le segment pr"""
    return min(p.x, r.x) <= q.x <= max(p.x, r.x) and \
           min(p.y, r.y) <= q.y <= max(p.y, r.y)

def do_intersect(p1, q1, p2, q2):
    """
    Vérifie si les segments p1q1 et p2q2 se coupent
    """
    o1 = get_orientation(p1, q1, p2)
    o2 = get_orientation(p1, q1, q2)
    o3 = get_orientation(p2, q2, p1)
    o4 = get_orientation(p2, q2, q1)

    # Cas général : les segments s'enjambent mutuellement
    if o1 != o2 and o3 != o4:
        return True

    # Cas particuliers : un point est colinéaire et situé sur l'autre segment
    if o1 == 0 and is_on_segment(p1, p2, q1): return True
    if o2 == 0 and is_on_segment(p1, q2, q1): return True
    if o3 == 0 and is_on_segment(p2, p1, q2): return True
    if o4 == 0 and is_on_segment(p2, q1, q2): return True

    return False
```
