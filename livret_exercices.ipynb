# Section 1 : Algorithmique sur les Graphes

## Problème 1.1 : Course Schedule

**Lien vers le problème :** [CSES - Course Schedule](https://cses.fi/problemset/task/1679)

**Énoncé :**
Vous devez valider $N$ cours pour valider votre année. Il existe $M$ prérequis indiquant qu'un cours $A$ doit impérativement être validé avant un cours $B$.
Votre tâche est de trouver un ordre valide pour suivre tous les cours. S'il est impossible de tous les suivre (présence de dépendances circulaires), signalez-le.

**Entrée :**
- La première ligne contient deux entiers $N$ et $M$. Les cours sont numérotés de $1$ à $N$.
- Les $M$ lignes suivantes décrivent les prérequis. Chaque ligne contient deux entiers $A$ et $B$ : le cours $A$ doit être complété avant le cours $B$.

**Sortie :**
- Affichez un ordre valide (les $N$ entiers séparés par des espaces). S'il n'y a pas de solution, affichez `IMPOSSIBLE`.

**Contraintes :**
- $1 \le N \le 10^5$
- $1 \le M \le 2 \cdot 10^5$

### Indication 1

Modélisez le problème sous forme de graphe orienté. Les cours sont les sommets et les prérequis sont des arêtes dirigées. Quel type de parcours permet de trier des dépendances ?

### Indication 2

Utilisez le tri topologique. Pour éviter la limite de récursion stricte de Python avec un DFS, privilégiez l'algorithme de Kahn basé sur le degré entrant (in-degree) des sommets et une file (Queue).

### Correction théorique

<b>Paradigme :</b> Tri Topologique (Algorithme de Kahn).

<b>Processus de réflexion :</b> 
Dans un graphe orienté, un tri topologique est un ordre linéaire des sommets tel que pour chaque arête dirigée $U \to V$, $U$ apparaît avant $V$. 
L'algorithme de Kahn consiste à compter le nombre d'arêtes entrantes (in-degree) pour chaque sommet. On place tous les sommets de degré entrant nul dans une file. On extrait un sommet, on l'ajoute au résultat, puis on "supprime" virtuellement ses arêtes sortantes en décrémentant le degré entrant de ses voisins. Si un voisin atteint un degré entrant nul, il est ajouté à la file.

<b>Preuve de justesse :</b>
Soit un graphe orienté acyclique (DAG). Il existe toujours au moins un sommet de degré entrant nul. L'algorithme extrait un tel sommet $u$. Comme aucun sommet restant n'a d'arête pointant vers $u$, placer $u$ en premier dans l'ordre partiel respecte toutes les dépendances. Par récurrence, l'invariant de boucle garantit l'ordre. Si à la fin, le nombre de sommets dans la solution est strictement inférieur à $N$, cela prouve mathématiquement l'existence d'un cycle (dépendance circulaire), rendant la complétion impossible.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}(N + M)$ en pire cas. Chaque sommet et chaque arête sont traités exactement une fois.
- <b>Espace :</b> $\mathcal{O}(N + M)$ pour la liste d'adjacence, le tableau des degrés entrants et la file de traitement.

## Problème 1.2 : Flight Discount

**Lien vers le problème :** [CSES - Flight Discount](https://cses.fi/problemset/task/1195)

**Énoncé :**
Vous voulez voyager de la ville $1$ à la ville $N$. Vous possédez un coupon de réduction qui permet de diviser par deux (division entière) le prix d'un seul vol pendant votre trajet. 
Trouvez le prix minimum pour atteindre la destination.

**Entrée :**
- La première ligne contient $N$ (villes) et $M$ (vols).
- Les $M$ lignes suivantes contiennent $A$, $B$, et $C$ : un vol de $A$ vers $B$ coûte $C$. (Graphe orienté).

**Sortie :**
- Affichez un entier : le coût minimum du trajet.

**Contraintes :**
- $1 \le N \le 10^5$
- $1 \le M \le 2 \cdot 10^5$
- $1 \le C \le 10^9$

### Indication 1

Un Dijkstra classique ne suffit pas car le choix d'utiliser le coupon modifie le coût futur. Il faut intégrer l'état du coupon dans le graphe lui-même.

### Indication 2

Au lieu de chercher le plus court chemin dans $G = (V, E)$, cherchez-le dans un graphe d'états où chaque sommet est défini par un tuple $(u, k)$ avec $u \in V$ et $k \in \{0, 1\}$ ($0$ = coupon disponible, $1$ = coupon utilisé).

### Correction théorique

<b>Paradigme :</b> Plus Court Chemin (Dijkstra sur Graphe d'États).

<b>Processus de réflexion :</b>
Chaque sommet $u$ est dédoublé : $u_0$ (on arrive à $u$ sans avoir utilisé le coupon) et $u_1$ (on arrive à $u$ en ayant déjà utilisé le coupon).
Pour chaque arête dirigée $u \to v$ de poids $w$ dans le graphe d'origine, on crée les transitions suivantes dans notre nouveau graphe :
1. $u_0 \to v_0$ avec le poids $w$ (on ne l'utilise toujours pas).
2. $u_0 \to v_1$ avec le poids $\lfloor w / 2 \rfloor$ (on l'utilise sur ce vol).
3. $u_1 \to v_1$ avec le poids $w$ (on l'a déjà utilisé, on paie le prix fort).
On lance ensuite l'algorithme de Dijkstra depuis le sommet $1_0$. Le résultat attendu sera le minimum entre la distance de $N_0$ et $N_1$ (bien que logiquement, $N_1$ sera toujours optimal).

<b>Preuve de justesse :</b>
L'algorithme de Dijkstra garantit la découverte du plus court chemin si tous les poids sont positifs. En modélisant l'état dans les nœuds, chaque transition reflète une action légale stricte. L'invariant de Dijkstra (lorsqu'un nœud $(u, k)$ est extrait de la file de priorité, sa distance est optimale et définitive) est conservé car l'utilisation conditionnelle du coupon (créant le poids $\lfloor w / 2 \rfloor$) respecte la positivité des arêtes.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}((V + E) \log V)$. Le graphe virtuel contient $2V$ sommets et $3E$ arêtes, ce qui ne modifie que la constante asymptotique.
- <b>Espace :</b> $\mathcal{O}(V + E)$ pour la liste d'adjacence et les tableaux de distances.
</details>

## Problème 1.3 : Road Reparation

**Lien vers le problème :** [CSES - Road Reparation](https://cses.fi/problemset/task/1675)

**Énoncé :**
Il y a $N$ villes et $M$ routes potentielles. Chaque route a un coût de réparation. Vous devez choisir un sous-ensemble de routes à réparer pour que toutes les villes soient connectées entre elles, en minimisant le coût total. S'il est impossible de tout connecter, signalez-le.

**Entrée :**
- $N$ (villes) et $M$ (routes potentielles).
- $M$ lignes contenant $A, B, C$ : une route entre $A$ et $B$ coûtant $C$.

**Sortie :**
- Affichez le coût minimum, ou `IMPOSSIBLE`.

**Contraintes :**
- $1 \le N \le 10^5$
- $1 \le M \le 2 \cdot 10^5$

### Indication 1

C'est la définition exacte du problème de l'Arbre Couvrant de Poids Minimum (MST - Minimum Spanning Tree). Pensez à l'algorithme de Kruskal.

### Indication 2

Pour implémenter Kruskal efficacement, triez les arêtes par poids croissant. Utilisez une structure de données Disjoint Set Union (DSU / Union-Find) avec compression de chemin et union par rang pour détecter les cycles en $\mathcal{O}(1)$ amorti.

### Correction théorique

<b>Paradigme :</b> Algorithme Glouton et Disjoint Set Union (Algorithme de Kruskal).

<b>Processus de réflexion :</b>
On initialise un graphe vide composé de $N$ composantes connexes (chaque sommet est isolé). On trie toutes les arêtes par coût croissant. On itère sur ces arêtes : pour chaque arête $(u, v)$ de poids $w$, on vérifie si $u$ et $v$ appartiennent à la même composante via un DSU. Si ce n'est pas le cas, on inclut l'arête dans notre MST (ce qui fusionne les composantes de $u$ et $v$) et on ajoute $w$ au coût total. On s'arrête quand on a sélectionné $N-1$ arêtes.

<b>Preuve de justesse :</b>
La justesse de Kruskal repose sur la propriété de coupure (Cut Property) des matroïdes. Soit une partition des sommets du graphe en deux ensembles $S$ et $V \setminus S$. L'arête de poids minimum traversant cette coupe appartient obligatoirement à tout MST. En traitant les arêtes par ordre croissant, Kruskal garantit que la première arête reliant deux composantes distinctes est l'arête de poids minimum de la coupe séparant ces deux composantes. L'absence de cycle est garantie par le DSU.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}(M \log M)$ dominé par le tri initial des arêtes. Les opérations du DSU prennent un temps $\mathcal{O}(M \alpha(N))$, où $\alpha$ est la fonction inverse d'Ackermann, assimilable à une constante.
- <b>Espace :</b> $\mathcal{O}(N + M)$ pour stocker les arêtes et les tableaux `parent` et `rank` du DSU.

## Problème 1.4 : Planets and Kingdoms

**Lien vers le problème :** [CSES - Planets and Kingdoms](https://cses.fi/problemset/task/1683)

**Énoncé :**
Un réseau stellaire est modélisé par $N$ planètes et $M$ téléporteurs unidirectionnels. Deux planètes appartiennent au même royaume si et seulement s'il existe un chemin aller-retour entre elles. Déterminez le nombre de royaumes et assignez un identifiant de royaume à chaque planète.

**Entrée :**
- $N$ planètes et $M$ téléporteurs.
- $M$ lignes : $a \to b$.

**Sortie :**
- L'entier $K$ : le nombre de royaumes.
- Une ligne de $N$ entiers indiquant l'identifiant du royaume ($1$ à $K$) de chaque planète.

**Contraintes :**
- $1 \le N \le 10^5$
- $1 \le M \le 2 \cdot 10^5$

### Indication 1

La relation "chemin de $u$ vers $v$ ET de $v$ vers $u$" définit mathématiquement une Composante Fortement Connexe (SCC - Strongly Connected Component) dans un graphe orienté.

### Indication 2

Utilisez l'algorithme de Kosaraju. Il requiert deux passes de DFS : une première sur le graphe d'origine pour enregistrer l'ordre de fin de traitement (post-order), et une seconde sur le graphe transposé (arêtes inversées) en traitant les sommets dans l'ordre inverse du post-order.

### Correction théorique

<b>Paradigme :</b> Parcours de Graphe (Algorithme de Kosaraju).

<b>Processus de réflexion :</b>
On veut regrouper les sommets mutuellement accessibles. Kosaraju exploite le graphe de condensation (où chaque SCC est réduite à un méta-sommet), qui est un DAG (Directed Acyclic Graph). 
1. Lancer un DFS global. Empiler les sommets une fois que tous leurs voisins ont été visités (post-order).
2. Inverser toutes les arêtes du graphe pour obtenir $G^T$.
3. Dépiler les sommets. Si un sommet n'est pas encore visité, lancer un DFS sur $G^T$ depuis ce sommet. Tous les sommets atteints lors de ce DFS forment exactement une SCC.

<b>Preuve de justesse :</b>
Les SCC d'un graphe $G$ sont exactement les mêmes que celles de $G^T$. Le tri topologique implicite du premier DFS garantit que le sommet dépilé en dernier appartient à une SCC "source" dans le graphe de condensation. En inversant les arêtes, cette source devient un "puits". Ainsi, un parcours depuis ce puits dans $G^T$ ne peut s'échapper vers d'autres SCC, confinant la découverte strictement aux sommets de la SCC courante.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}(N + M)$. Deux parcours DFS complets et la transposition du graphe, tous linéaires.
- <b>Espace :</b> $\mathcal{O}(N + M)$ pour la construction de $G$, de $G^T$, et de la pile d'appels.

## Problème 1.5 : Flight Routes

**Lien vers le problème :** [CSES - Flight Routes](https://cses.fi/problemset/task/1196)

**Énoncé :**
Votre tâche est de trouver les $k$ itinéraires de vol les moins chers de Syrjälä (1) à Metsäpolku ($N$). Un itinéraire peut visiter la même ville ou emprunter le même vol plusieurs fois.

**Entrée :**
- $N$ (villes), $M$ (vols), et $k$.
- $M$ lignes : $A \to B$ avec le prix $C$.

**Sortie :**
- Affichez les $k$ coûts minimaux triés par ordre croissant.

**Contraintes :**
- $1 \le N \le 10^5$
- $1 \le M \le 2 \cdot 10^5$
- $1 \le k \le 10$

### Indication 1

Dijkstra trouve le chemin le plus court car il fige la distance d'un nœud dès sa première extraction. Pour trouver les $k$ plus courts, il faut autoriser un nœud à être visité et relâché jusqu'à $k$ fois.

### Indication 2

Au lieu de stocker un tableau `dist[v]`, maintenez pour chaque sommet une file de priorité locale (Min-Heap) de taille maximale $k$. Lors de la relaxation $u \to v$, si la nouvelle distance est meilleure que la pire des $k$ distances de $v$, ajoutez-la et poussez $(nouvelle\_distance, v)$ dans la file de priorité globale de Dijkstra.

### Correction théorique

<b>Paradigme :</b> Plus Court Chemin (Dijkstra Modifié pour $k$-chemins).

<b>Processus de réflexion :</b>
On modifie Dijkstra : l'état dans la file de priorité globale est `(distance_actuelle, u)`. Pour chaque voisin $v$, le coût potentiel est $d = distance\_actuelle + poids(u \to v)$. Au lieu d'un seul `dist[v]`, on a un tableau de files `best_dists[v]`. On n'accepte $d$ que si on a trouvé moins de $k$ chemins vers $v$, ou si $d$ est strictement inférieur à la distance la plus longue parmi les $k$ stockées pour $v$. Le processus s'arrête naturellement.

<b>Preuve de justesse :</b>
Dijkstra fonctionne par relaxation monotone. En conservant les $k$ meilleures distances pour tout sommet intermédiaire, on s'assure que toutes les combinaisons préfixiales valides pouvant générer un chemin global optimal vers $N$ sont explorées. Le principe de Bellman s'applique : le $i$-ème plus court chemin vers un nœud est composé d'une arête finale ajoutée à l'un des $j$-ièmes plus courts chemins ($j \le i$) vers son prédécesseur.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}(K \cdot M \log(K \cdot N))$. Chaque arête peut être relâchée $K$ fois. La file de priorité globale contient au maximum $K \cdot N$ éléments.
- <b>Espace :</b> $\mathcal{O}(K \cdot N + M)$ pour conserver jusqu'à $K$ distances par sommet.

# Section 2 : Programmation Dynamique

## Problème 2.1 : Increasing Subsequence

**Lien vers le problème :** [CSES - Increasing Subsequence](https://cses.fi/problemset/task/1145)

**Énoncé :**
Étant donné un tableau de $N$ entiers, trouvez la longueur de la plus longue sous-séquence strictement croissante (LIS - Longest Increasing Subsequence).
*Note : Les éléments de la sous-séquence ne sont pas obligés d'être contigus.*

**Entrée :**
- La première ligne contient un entier $N$.
- La ligne suivante contient les $N$ entiers du tableau $X_1, X_2, \dots, X_n$.

**Sortie :**
- Affichez un seul entier : la longueur maximale de la sous-séquence croissante.

**Contraintes :**
- $1 \le N \le 2 \cdot 10^5$
- $1 \le X_i \le 10^9$

### Indication 1

Une approche naïve de DP en $\mathcal{O}(N^2)$ (où `dp[i]` est la LIS se terminant à l'index `i`) obtiendra un Time Limit Exceeded (TLE) à cause du $N=2 \cdot 10^5$.

### Indication 2

Maintenez un tableau `tails` où `tails[k]` stocke la plus petite valeur de fin parmi toutes les sous-séquences croissantes de longueur `k+1`. Ce tableau sera naturellement trié, ce qui permet d'utiliser la recherche dichotomique (`bisect_left` en Python).
</details>

### Correction théorique

<b>Paradigme :</b> Programmation Dynamique optimisée par Recherche Dichotomique (Patience Sorting).

<b>Processus de réflexion :</b>
On maintient un tableau `tails`. À l'étape $i$, si $X_i$ est strictement supérieur à tous les éléments de `tails`, on l'ajoute à la fin (allongeant ainsi la plus longue sous-séquence trouvée). Sinon, on remplace le premier élément dans `tails` qui est supérieur ou égal à $X_i$ par $X_i$. 
Pourquoi ? Parce qu'une sous-séquence se terminant par une valeur plus petite offre plus de "potentiel" pour être prolongée par les éléments futurs.

<b>Preuve de justesse (Invariant) :</b>
Le tableau `tails` maintient la propriété suivante : `tails[j]` est l'élément minimal qui termine une séquence croissante de longueur $j+1$. L'opération de remplacement via la recherche dichotomique préserve le fait que `tails` est strictement croissant. À la fin du traitement, la taille du tableau `tails` correspond exactement à la longueur de la LIS globale.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}(N \log N)$. Pour chacun des $N$ éléments, on effectue une recherche dichotomique coûtant $\mathcal{O}(\log L)$ où $L$ est la longueur actuelle de `tails`.
- <b>Espace :</b> $\mathcal{O}(N)$ dans le pire cas (si le tableau est déjà strictement croissant) pour stocker le tableau `tails`.

## Problème 2.2 : Caesar's Legions

**Lien vers le problème :** [Codeforces - 118D](https://codeforces.com/problemset/problem/118/D)

**Énoncé :**
Gaius Julius Caesar souhaite aligner son armée pour une parade. L'armée est composée de $N_1$ fantassins et $N_2$ cavaliers.
Cependant, pour l'esthétique de la parade, il refuse d'avoir plus de $K_1$ fantassins consécutifs ou plus de $K_2$ cavaliers consécutifs.
Combien de dispositions différentes existent pour aligner toute l'armée ? Le résultat doit être donné modulo $10^8$.

**Entrée :**
- Une ligne contenant quatre entiers séparés par des espaces : $N_1$, $N_2$, $K_1$, et $K_2$.

**Sortie :**
- Affichez le nombre de dispositions valides modulo $10^8$.

**Contraintes :**
- $1 \le N_1, N_2 \le 100$
- $1 \le K_1, K_2 \le 10$

### Indication 1

C'est un problème de dénombrement. La dimension de votre tableau DP doit capturer l'état complet de la construction de la file. De quelles informations avez-vous besoin pour savoir si ajouter un fantassin ou un cavalier est valide ?

### Indication 2

Il vous faut un DP à 4 dimensions (ou 3 avec une astuce) : `dp[n1][n2][consec][type]`. 
Optimisation : Vous n'avez pas besoin de garder le compte exact des consécutifs dans l'état si vous générez des "blocs" de fantassins ou de cavaliers entiers d'un seul coup. `dp[i][j][0]` = nombre de façons de placer $i$ fantassins et $j$ cavaliers en terminant par un fantassin.


### Correction théorique

<b>Paradigme :</b> Programmation Dynamique avec États.

<b>Processus de réflexion :</b>
Soit `dp[i][j][0]` le nombre de configurations valides utilisant $i$ fantassins, $j$ cavaliers et se terminant par un fantassin.
Soit `dp[i][j][1]` la même chose mais en se terminant par un cavalier.

Pour calculer `dp[i][j][0]`, le dernier bloc ajouté à la fin de la file est constitué de $x$ fantassins consécutifs ($1 \le x \le K_1$). Avant ce bloc, la file devait obligatoirement se terminer par un cavalier pour casser la consécutivité.
La transition est donc :
$dp[i][j][0] = \sum_{x=1}^{\min(i, K_1)} dp[i-x][j][1]$
De même pour les cavaliers :
$dp[i][j][1] = \sum_{y=1}^{\min(j, K_2)} dp[i][j-y][0]$

Cas de base :
- $dp[0][0][0] = dp[0][0][1] = 1$ (état initial abstrait)

<b>Preuve de justesse (Sous-structure optimale) :</b>
Toute configuration valide de longueur $L$ se termine nécessairement par un bloc contigu du même type d'unité de longueur $\le K$. En retirant ce bloc, le préfixe restant de la file est par définition une configuration valide plus courte qui se termine par l'autre type d'unité. Le décompte récursif est exhaustif et mutuellement exclusif.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}(N_1 \times N_2 \times (K_1 + K_2))$. Avec les contraintes ($100 \times 100 \times 20 = 2 \cdot 10^5$ opérations), le calcul est instantané.
- <b>Espace :</b> $\mathcal{O}(N_1 \times N_2)$ en ne stockant que l'état $i$ et $j$ et la dimension binaire du type final.
</details>

## Problème 2.3 : Edit Distance

**Lien vers le problème :** [CSES - Edit Distance](https://cses.fi/problemset/task/1639)

**Énoncé :**
Étant donné deux chaînes de caractères de tailles $N$ et $M$, calculez le nombre minimal d'opérations requises pour transformer la première chaîne en la seconde. 
Les opérations autorisées (de coût 1) sont : ajouter un caractère, supprimer un caractère, remplacer un caractère.

**Entrée :**
- La première ligne contient une chaîne de longueur $N$.
- La deuxième ligne contient une chaîne de longueur $M$.

**Sortie :**
- Un seul entier : la distance d'édition minimale.

**Contraintes :**
- $1 \le N, M \le 5000$

### Indication 1

Comparez les préfixes des deux chaînes. `dp[i][j]` représentera la distance d'édition entre les $i$ premiers caractères de la première chaîne et les $j$ premiers caractères de la seconde.

### Indication 2

Les transitions découlent directement des opérations. Si $S1[i] == S2[j]$, l'ajout est gratuit : `dp[i-1][j-1]`. Sinon, on prend le min entre Insertion `dp[i][j-1]`, Suppression `dp[i-1][j]`, et Substitution `dp[i-1][j-1]`, le tout $+1$.

### Correction théorique

<b>Paradigme :</b> Programmation Dynamique 2D.

<b>Processus de réflexion :</b>
On construit une grille DP de taille $(N+1) \times (M+1)$. 
Cas de base : `dp[i][0] = i` (supprimer $i$ caractères) et `dp[0][j] = j` (insérer $j$ caractères).
Transition pour $i>0, j>0$ :
- Soit les caractères concordent, la matrice se propage en diagonale sans coût.
- Soit ils divergent, et on minimise localement le coût des 3 opérations d'édition.

<b>Preuve de justesse :</b>
La sous-structure optimale est stricte. L'alignement optimal de deux préfixes implique nécessairement l'alignement optimal de leurs sous-préfixes, modifiés par une seule opération terminale. La formulation mathématique couvre l'intégralité de l'espace de recherche des mutations. 

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}(N \times M)$. Chaque état est visité et évalué en temps constant.
- <b>Espace :</b> $\mathcal{O}(\min(N, M))$. Puisque le calcul de la ligne $i$ ne nécessite que la ligne $i-1$, le tableau 2D peut être réduit à 2 vecteurs 1D permutés à chaque itération.

## Problème 2.4 : Projects

**Lien vers le problème :** [CSES - Projects](https://cses.fi/problemset/task/1140)

**Énoncé :**
On vous propose $N$ projets. Chaque projet a un jour de début $a$, un jour de fin $b$, et une récompense $p$. Vous ne pouvez assister qu'à un seul projet par jour (les intervalles valides sont strictement disjoints).
Trouvez la récompense maximale réalisable.

**Entrée :**
- $N$ (nombre de projets).
- $N$ lignes : $a_i, b_i, p_i$.

**Sortie :**
- Un entier : la récompense maximale.

**Contraintes :**
- $1 \le N \le 2 \cdot 10^5$
- $1 \le a_i \le b_i \le 10^9$
- $1 \le p_i \le 10^9$

### Indication 1

Si tous les poids étaient de 1, une simple approche gloutonne en triant par jour de fin suffirait. Avec des poids, le glouton échoue. Il faut combiner le tri par jour de fin avec de la programmation dynamique.

### Indication 2

`dp[i]` sera la récompense max en considérant les $i$ premiers projets triés par heure de fin. Pour calculer l'état si on *inclut* le projet $i$, on doit trouver le dernier projet $k < i$ qui se termine avant $a_i$. Utilisez la recherche dichotomique pour le trouver en $\mathcal{O}(\log N)$.

### Correction théorique

<b>Paradigme :</b> Programmation Dynamique sur Intervalles & Recherche Dichotomique.

<b>Processus de réflexion :</b>
On trie tous les projets par leur jour de fin $b$. Soit `dp[i]` le gain maximal obtenu avec un sous-ensemble des $i$ premiers projets. Deux choix s'offrent à nous :
1. Ne pas faire le projet $i$ : on hérite de `dp[i-1]`.
2. Faire le projet $i$ : on gagne $p_i$, mais on doit y additionner le gain optimal des projets compatibles. On cherche via `bisect_right` le plus grand index $k$ tel que $b_k < a_i$. Le gain devient `dp[k] + p_i`.
Transition : `dp[i] = max(dp[i-1], dp[k] + p_i)`.

<b>Preuve de justesse :</b>
Le tri garantit l'ordre chronologique de la complétion des états. L'équation de transition couvre un arbre binaire de décision exhaustif (inclure vs exclure). La recherche dichotomique cible formellement le meilleur état valide existant. Par récurrence, chaque `dp[i]` est la solution optimale au sous-problème confiné dans l'intervalle temporel allant jusqu'à $b_i$.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}(N \log N)$. Le tri prend $\mathcal{O}(N \log N)$, et le remplissage du DP demande $N$ itérations effectuant chacune une dichotomie en $\mathcal{O}(\log N)$.
- <b>Espace :</b> $\mathcal{O}(N)$ pour stocker les intervalles et le tableau DP.

## Problème 2.5 : Sleeping Schedule

**Lien vers le problème :** [Codeforces - 1324E](https://codeforces.com/problemset/problem/1324/E)

**Énoncé :**
Vova a $N$ cycles de sommeil prévus. Une journée dure $H$ heures. Actuellement, il est l'heure $0$. Pour le $i$-ème cycle, il peut choisir d'aller dormir exactement $A_i$ heures après son réveil précédent, ou $A_i - 1$ heures après. 
Un sommeil est "bon" s'il va se coucher à une heure de la journée comprise dans l'intervalle inclusif $[L, R]$.
Maximisez le nombre de bons sommeils.

**Entrée :**
- $N, H, L, R$.
- Les $N$ entiers $A_i$.

**Sortie :**
- Le nombre maximal de bons sommeils possibles.

**Contraintes :**
- $1 \le N \le 2000$
- $3 \le H \le 2000$

### Indication 1

Chaque décision locale (choisir $A_i$ ou $A_i - 1$) modifie l'heure de tous les couchers futurs. Il faut encapsuler l'heure absolue de la journée dans la dimension du DP.

### Indication 2

Utilisez un DP 2D : `dp[i][h]` = le nombre maximal de bons sommeils obtenus après $i$ cycles, sachant que le dernier réveil a eu lieu à l'heure $h$ de la journée ($0 \le h < H$). Les transitions s'opèrent via l'arithmétique modulaire.

### Correction théorique

<b>Paradigme :</b> Programmation Dynamique.

<b>Processus de réflexion :</b>
On initialise un tableau `dp[N+1][H]` à $-1$ (état inaccessible) et `dp[0][0] = 0`.
Pour chaque étape $i$ de $0$ à $N-1$, on itère sur toutes les heures possibles $h \in [0, H-1]$. Si `dp[i][h] != -1`, deux branches temporelles s'ouvrent :
1. Coucher après $A_{i+1}$ heures : l'heure devient $h_1 = (h + A_{i+1}) \pmod H$.
2. Coucher après $A_{i+1} - 1$ heures : l'heure devient $h_2 = (h + A_{i+1} - 1) \pmod H$.
Le gain conditionnel $g_k = 1$ si $L \le h_k \le R$, sinon $0$.
La relaxation s'écrit : `dp[i+1][hk] = max(dp[i+1][hk], dp[i][h] + gk)`.
La réponse est le $\max$ de la dernière ligne `dp[N]`.

<b>Preuve de justesse :</b>
La sous-structure optimale est valide car le décompte des points ne nécessite aucun accès à l'historique des décisions passées, seule la congruence modulo $H$ de l'horloge détermine le score courant et le point de départ du cycle suivant. Les deux choix temporels forment un graphe dirigé acyclique (DAG) dont les arêtes respectent les opérations du groupe additif cyclique $\mathbb{Z}/H\mathbb{Z}$.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}(N \times H)$. Avec $N, H = 2000$, cela fait $4 \cdot 10^6$ itérations de boucle, ce qui s'exécute en quelques millisecondes.
- <b>Espace :</b> $\mathcal{O}(H)$. La matrice 2D peut être optimisée en ne conservant que la ligne courante $i$ et la ligne précédente $i-1$.

## Problème 3.1 : Dynamic Range Sum Queries

**Lien vers le problème :** [CSES - Dynamic Range Sum Queries](https://cses.fi/problemset/task/1648)

**Énoncé :**
Étant donné un tableau de $N$ entiers, vous devez traiter $Q$ requêtes de deux types :
1. Mettre à jour la valeur à la position $k$ pour qu'elle devienne $u$.
2. Calculer la somme des valeurs dans l'intervalle $[a, b]$.

**Entrée :**
- La première ligne contient $N$ et $Q$.
- La deuxième ligne contient les $N$ entiers initiaux.
- Les $Q$ lignes suivantes contiennent les requêtes de type `1 k u` ou `2 a b`.

**Sortie :**
- Pour chaque requête de somme, affichez le résultat sur une nouvelle ligne.

**Contraintes :**
- $1 \le N, Q \le 2 \cdot 10^5$
- $1 \le x_i, u \le 10^9$

### Indication 1

Les opérations de mise à jour et de somme de préfixes doivent être effectuées en temps logarithmique. Un tableau classique favorise l'une au détriment de l'autre (mise à jour $\mathcal{O}(1)$ et somme $\mathcal{O}(N)$ ou inversement).

### Indication 2

Implémentez un Arbre de Fenwick (Binary Indexed Tree). La somme sur $[a, b]$ s'obtient par la différence des sommes de préfixes $S(b) - S(a-1)$.

### Correction théorique

<b>Paradigme :</b> Structure de données dynamique (Arbre de Fenwick / BIT).

<b>Processus de réflexion :</b>
L'Arbre de Fenwick est une structure qui représente un tableau des sommes partielles implicite. Chaque index $i$ de l'arbre stocke la somme d'un sous-tableau de longueur $P(i)$, où $P(i)$ est la valeur du bit de poids faible de $i$ (isolable via l'opération bit à bit `i & -i`). Pour mettre à jour l'arbre avec un delta (différence entre nouvelle et ancienne valeur), on ajoute ce delta à tous les intervalles couvrant l'indice en remontant vers la droite. Pour interroger la somme d'un préfixe $S(k)$, on somme les intervalles disjoints formant le préfixe en descendant vers la gauche.

<b>Preuve de justesse :</b>
Tout entier $k$ s'écrit de manière unique comme une somme de puissances de 2 (décomposition binaire). De manière mathématiquement analogue, l'intervalle des indices $[1, k]$ se partitionne en un nombre exact d'intervalles précalculés dans l'arbre (au plus $\lfloor \log_2(k) \rfloor + 1$). L'invariant structurel garantit que l'opération de remontée met à jour l'exhaustivité des préfixes englobants, rendant le calcul télescopique $S(b) - S(a-1)$ rigoureusement exact sur tout segment arbitraire.

<b>Complexité :</b>
- <b>Temps :</b> Dans le pire cas et en espérance, $\mathcal{O}(\log N)$ par requête ou mise à jour, le nombre de bits à 1 ou de sauts vers le parent étant strictement borné par la profondeur logarithmique de l'arbre.
- <b>Espace :</b> Dans le pire cas et en espérance, $\mathcal{O}(N)$ pour allouer le tableau mémoire du BIT de dimension $N+1$.

## Problème 3.2 : Range Update Queries

**Lien vers le problème :** [CSES - Range Update Queries](https://cses.fi/problemset/task/1651)

**Énoncé :**
Étant donné un tableau de $N$ entiers, vous devez traiter $Q$ requêtes de deux types :
1. Ajouter la valeur $u$ à chaque élément de l'intervalle $[a, b]$.
2. Afficher la valeur actuelle à la position $k$.

**Entrée :**
- La première ligne contient $N$ et $Q$.
- La deuxième ligne contient les $N$ entiers initiaux.
- Les $Q$ lignes suivantes contiennent les requêtes de type `1 a b u` ou `2 k`.

**Sortie :**
- Pour chaque requête de type 2, affichez la valeur requise.

**Contraintes :**
- $1 \le N, Q \le 2 \cdot 10^5$
- $1 \le u \le 10^9$

### Indication 1

Mettre à jour explicitement chaque élément de l'intervalle génère un coût asymptotique de $\mathcal{O}(N)$ par requête, causant un TLE. Réfléchissez aux modifications strictement locales sur les bornes de l'intervalle.

### Indication 2

Utilisez le paradigme du tableau des différences (Difference Array). Si $D[i] = A[i] - A[i-1]$, une augmentation de $+u$ sur le segment $[a, b]$ ne modifie que deux valeurs : $D[a]$ et $D[b+1]$.

### Correction théorique

<b>Paradigme :</b> Tableau des différences couplé à un Arbre de Fenwick.

<b>Processus de réflexion :</b>
Le problème inverse la logique classique : on a des mises à jour sur intervalles et des requêtes ponctuelles. On définit le tableau des différences $D$ avec $D[1] = A[1]$ et $D[i] = A[i] - A[i-1]$ pour $i > 1$. L'opération d'ajout $+u$ sur $[a, b]$ modifie localement : $D[a]$ augmente de $u$, et $D[b+1]$ diminue de $u$. La valeur ponctuelle de $A[k]$ est simplement la somme préfixe de $D$ jusqu'à $k$. On maintient donc le tableau $D$ au sein d'un Arbre de Fenwick.

<b>Preuve de justesse :</b>
Le lemme des sommes télescopiques démontre algébriquement que $\sum_{j=1}^k D[j] = A[k]$. Lors d'une mise à jour de $[a, b]$, ajouter $u$ à $D[a]$ augmente la somme préfixe $\sum D[j]$ de $u$ pour tout indice $\ge a$. Soustraire $u$ à $D[b+1]$ annule cette augmentation pour tout indice $> b$. L'invariant de la valeur courante de $A[k]$ est ainsi formellement préservé et confiné exclusivement à l'intervalle cible.

<b>Complexité :</b>
- <b>Temps :</b> Dans le pire cas et en espérance, $\mathcal{O}(\log N)$ par requête. L'ajout sur intervalle demande deux mises à jour ponctuelles dans le BIT, et la lecture demande une requête de préfixe.
- <b>Espace :</b> Dans le pire cas et en espérance, $\mathcal{O}(N)$ pour la structure de l'arbre.

## Problème 3.3 : Hotel Queries

**Lien vers le problème :** [CSES - Hotel Queries](https://cses.fi/problemset/task/1143)

**Énoncé :**
Il y a $N$ hôtels, chacun avec un nombre de chambres libres $h_i$. $M$ groupes de touristes arrivent séquentiellement. Chaque groupe demande un nombre $r_j$ de chambres et sera hébergé dans le premier hôtel (d'indice minimum) ayant suffisamment de chambres. S'il trouve un hôtel, il réserve les chambres (le nombre de chambres libres diminue).
Trouvez l'hôtel attribué à chaque groupe (ou 0 s'il n'y en a aucun).

**Entrée :**
- La première ligne contient $N$ et $M$.
- La deuxième ligne contient $h_1, h_2, \dots, h_n$.
- La troisième ligne contient $r_1, r_2, \dots, r_m$.

**Sortie :**
- Affichez l'indice de l'hôtel assigné pour chaque groupe.

**Contraintes :**
- $1 \le N, M \le 2 \cdot 10^5$
- $1 \le h_i, r_j \le 10^9$

### Indication 1

Une recherche linéaire sur le tableau pour chaque groupe est en $\mathcal{O}(NM)$ et produira un TLE. Il est impératif d'utiliser une structure hiérarchique capable d'élaguer les segments du tableau ne possédant pas de capacité suffisante.

### Indication 2

Construisez un Segment Tree (Arbre de Segments) où chaque nœud stocke le maximum des chambres libres de son intervalle. Utilisez une descente prioritaire sur le sous-arbre gauche pour garantir l'indice minimum.

### Correction théorique

<b>Paradigme :</b> Arbre de Segments (Requête de maximum et descente binaire orientée).

<b>Processus de réflexion :</b>
On modélise un arbre binaire complet sur les capacités des hôtels. Pour loger un groupe nécessitant $X$ chambres, on inspecte la racine : si son maximum est $< X$, aucun hôtel n'est valide (retourner 0). Sinon, on traverse l'arbre de haut en bas : on inspecte d'abord le fils gauche. S'il satisfait $T[gauche] \ge X$, l'hôtel optimal se trouve nécessairement dans la moitié gauche, on y descend. Sinon, on descend dans le fils droit. Arrivé sur la feuille correspondante, on décrémente sa valeur de $X$ et on remet à jour les maximums parentaux en remontant.

<b>Preuve de justesse :</b>
L'arbre respecte l'invariant $T[v] = \max(T[2v], T[2v+1])$. La preuve repose sur un argument d'adversaire : l'énoncé exige l'indice minimum. Lors de la bifurcation à un nœud, le sous-arbre gauche représente le préfixe strict de l'intervalle actuel. Si le maximum du sous-arbre gauche est supérieur ou égal à la demande, l'existence de la solution dans ce sous-espace est garantie. L'espace de recherche droit peut être mathématiquement ignoré, respectant formellement l'optimalité de l'indice minimal sans collision.

<b>Complexité :</b>
- <b>Temps :</b> La construction initiale coûte $\mathcal{O}(N)$. La résolution de chaque requête force la visite d'exactement un chemin racine-feuille de longueur $\lceil \log_2 N \rceil$, puis la mise à jour remonte ce même chemin. La complexité est donc $\mathcal{O}(\log N)$ en pire cas et en espérance.
- <b>Espace :</b> Strictement $\mathcal{O}(N)$ (l'arbre complet nécessite une allocation d'au maximum $4N$ nœuds) dans le pire cas et en espérance.

## Problème 3.4 : List Removals

**Lien vers le problème :** [CSES - List Removals](https://cses.fi/problemset/task/1749)

**Énoncé :**
Vous disposez d'une liste de $N$ entiers. On vous donne $N$ positions consécutives à supprimer. La position spécifiée est indexée en fonction des éléments *restants* dans la liste. À chaque suppression, vous devez afficher la valeur de l'élément supprimé.

**Entrée :**
- La première ligne contient $N$.
- La deuxième ligne contient le contenu initial de la liste.
- La troisième ligne contient les $N$ positions $p_i$ à supprimer séquentiellement.

**Sortie :**
- Affichez l'élément supprimé pour chaque requête.

**Contraintes :**
- $1 \le N \le 2 \cdot 10^5$
- $1 \le x_i \le 10^9$
- $1 \le p_i \le N - i + 1$

### Indication 1

Le décalage de mémoire (`pop(index)`) sur un tableau Python ou `std::vector` classique opère en $\mathcal{O}(N)$ par suppression, menant inévitablement à un temps total $\mathcal{O}(N^2)$.

### Indication 2

Maintenez le tableau initial intact. Utilisez un Arbre de Fenwick indexé sur la taille $N$, rempli de $1$ (présent) et de $0$ (supprimé). Trouver le $k$-ième élément restant revient à trouver le premier index $i$ tel que la somme du préfixe soit égale à $k$. Optimisez la recherche dichotomique avec la technique de "Binary Lifting" sur l'arbre de Fenwick.

### Correction théorique

<b>Paradigme :</b> Sommes préfixes binaires et Binary Lifting (Descente Dichotomique Structurelle).

<b>Processus de réflexion :</b>
Un tableau binaire abstrait $B$ maintient l'état d'existence de chaque élément (initialisé à $1$). Le nombre d'éléments restants avant et incluant l'indice $i$ est la somme préfixe $\sum_{j=1}^i B[j]$. Trouver la cible $k$ implique d'isoler l'indice minimal $i$ satisfaisant $\sum_{j=1}^i B[j] = k$. Une approche naïve combinant `bisect` et une requête BIT nécessiterait $\mathcal{O}(\log^2 N)$. En appliquant le Binary Lifting sur l'arbre de Fenwick, on parcourt directement les bits de poids fort de l'indice courant vers les bits de poids faible, accumulant les valeurs sans franchir la cible $k$, réduisant la requête à $\mathcal{O}(\log N)$. On termine par $B[i] = 0$.

<b>Preuve de justesse :</b>
Soit $S(i) = \sum_{j=1}^i B[j]$. Puisque $B[j] \in \{0, 1\}$, l'invariant fondamental garantit que la fonction $S(i)$ est monotone et croissante au sens large. Le théorème des valeurs intermédiaires sur le domaine discret assure l'existence d'au moins une racine pour la cible $k$. Le Binary Lifting agit comme un parcours structurel en exploitant l'exhaustivité des sous-arbres dyadiques, ce qui garantit d'isoler le borne inférieure exacte, validant l'unicité du $k$-ième élément actif.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}(\log N)$ en pire cas et espérance par suppression grâce au parcours bit à bit direct sur le tableau sous-jacent du BIT. Le temps total s'établit à $\mathcal{O}(N \log N)$.
- <b>Espace :</b> $\mathcal{O}(N)$ formellement borné pour l'allocation du BIT et la sauvegarde des valeurs réelles du tableau.

## Problème 3.5 : Salary Queries

**Lien vers le problème :** [CSES - Salary Queries](https://cses.fi/problemset/task/1144)

**Énoncé :**
Une entreprise compte $N$ employés. Vous disposez de $Q$ requêtes :
1. `! k x` : Modifier le salaire de l'employé d'indice $k$ pour qu'il devienne $x$.
2. `? a b` : Compter le nombre d'employés dont le salaire actuel est compris dans l'intervalle $[a, b]$.

**Entrée :**
- La première ligne contient $N$ et $Q$.
- La deuxième ligne contient les salaires initiaux.
- Les $Q$ lignes suivantes contiennent les requêtes.

**Sortie :**
- Pour chaque requête `?`, affichez le compte.

**Contraintes :**
- $1 \le N, Q \le 2 \cdot 10^5$
- $1 \le p_i, x, a, b \le 10^9$

### Indication 1

Les requêtes s'exécutent sur le domaine de valeurs des salaires ($10^9$). Un Arbre de Fenwick indexé directement sur les salaires occasionnerait un `MemoryLimitExceeded` car le tableau requiert plus de mémoire RAM que les 512 MB tolérés par le juge.

### Indication 2

Réalisez une passe de traitement hors ligne (Offline). Accumulez tous les salaires qui existeront à n'importe quel instant du programme (salaires initiaux + modifications $x$ des requêtes) et effectuez une Compression de Coordonnées. Cela garantit un domaine de recherche dense borné par $\mathcal{O}(N + Q)$ valeurs uniques.

### Correction théorique

<b>Paradigme :</b> Compression de Coordonnées (Offline Processing) et Arbre de Fenwick des fréquences.

<b>Processus de réflexion :</b>
La contrainte forte est l'éparsité de l'espace des valeurs. On parcourt d'abord (offline) l'ensemble de l'entrée pour identifier l'intégralité des salaires générés. On trie et dé-duplique cette liste pour définir une application bijective $\phi(s)$ mappant de façon contiguë chaque salaire vers un entier de $1$ à $K$, où l'univers projeté $K \le N + Q$. Un BIT alloué sur cette dimension $K$ maintient la fréquence d'apparition active de chaque salaire compressé. Le comptage sur l'intervalle continu $[A, B]$ devient la somme des fréquences sur l'intervalle discret $[\phi(\min(s \ge A)), \phi(\max(s \le B))]$, nécessitant deux appels `bisect` supplémentaires.

<b>Preuve de justesse :</b>
La robustesse du modèle repose sur le lemme de l'isomorphisme de l'ordre ordinal. Soit $S$ l'ensemble fini et totalement ordonné des salaires dictés par les entrées. L'application surjective $\phi : S \to \{1, \dots, |S|\}$ générée par l'étape de compression est strictement croissante. Mathématiquement, pour tout doublet $x, y \in S$, la relation $x \le y \iff \phi(x) \le \phi(y)$ reste inviolée. Les bornes extrêmes projetées via dichotomie interceptent rigoureusement les mêmes sous-ensembles de l'espace abstrait. L'invariant des données est ainsi conservé avec zéro perte d'information quant au décompte.

<b>Complexité :</b>
- <b>Temps :</b> Le tri initial hors ligne pour construire le dictionnaire coûte $\mathcal{O}((N+Q) \log(N+Q))$ en pire cas et en espérance. Ensuite, grâce à la dichotomie $\mathcal{O}(\log K)$ et l'arbre de Fenwick $\mathcal{O}(\log K)$, le traitement séquentiel de chaque requête de mise à jour ou de lecture ponctionne un temps absolu de $\mathcal{O}(\log(N+Q))$ au pire cas et en espérance.
- <b>Espace :</b> L'utilisation totale de la mémoire est contrainte de façon déterministe à $\mathcal{O}(N+Q)$ en pire cas et espérance, circonscrivant le tableau inverse de compression et l'Arbre de Fenwick, évitant l'impossible allocation dense sur le domaine limitrophe de $10^9$.

## Problème 4.1 : T-primes

**Lien vers le problème :** [Codeforces - 230B](https://codeforces.com/problemset/problem/230/B)

**Énoncé :**
Un nombre est appelé "T-prime" s'il possède exactement trois diviseurs positifs distincts. 
On vous donne un tableau de $N$ entiers. Pour chaque entier, déterminez s'il est un T-prime ou non.

**Entrée :**
- Un entier $N$.
- $N$ entiers $X_i$ séparés par des espaces.

**Sortie :**
- Pour chaque entier, affichez `YES` s'il est T-prime, sinon `NO`.

**Contraintes :**
- $1 \le N \le 10^5$
- $1 \le X_i \le 10^{12}$

### Indication 1

Analysez la structure des diviseurs. Les diviseurs vont par paires $(d, X/d)$. Pour qu'un nombre ait un nombre impair de diviseurs, il doit nécessairement être un carré parfait.

### Indication 2

Un T-prime a exactement 3 diviseurs : $1$, lui-même, et un seul autre nombre. Ce nombre est donc sa racine carrée, et cette racine carrée doit obligatoirement être un nombre premier. Utilisez un Crible d'Ératosthène précalculé jusqu'à $\sqrt{\max(X_i)} = 10^6$.

### Correction théorique

<b>Paradigme :</b> Crible d'Ératosthène et Analyse Algébrique.

<b>Processus de réflexion :</b>
Étant donné la limite $X_i \le 10^{12}$, une factorisation naïve en $\mathcal{O}(\sqrt{X_i})$ par requête donnerait un temps total de $\mathcal{O}(N \sqrt{X_i}) \approx 10^{11}$ opérations (Time Limit Exceeded). 
L'observation mathématique clé : un entier $n$ possède exactement 3 diviseurs si et seulement si $n = p^2$ où $p$ est un nombre premier. On précalcule tous les nombres premiers jusqu'à $10^6$ avec un Crible d'Ératosthène. Pour chaque requête $X_i$, on vérifie en $\mathcal{O}(1)$ si $\sqrt{X_i}$ est un entier, puis si cet entier figure dans notre tableau booléen criblé.

<b>Preuve de justesse :</b>
Soit la décomposition en facteurs premiers $X = p_1^{k_1} p_2^{k_2} \dots p_m^{k_m}$. Le nombre total de diviseurs est donné par $\prod (k_i + 1)$. Pour que ce produit vaille exactement 3, il faut qu'il n'y ait qu'un seul facteur premier $p_1$, et que $k_1 + 1 = 3$, d'où $k_1 = 2$. Donc $X = p_1^2$. Cette condition est à la fois nécessaire et suffisante.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}(M \log \log M + N)$ où $M = 10^6$. Le crible s'exécute une seule fois en amont. Chaque requête est ensuite traitée en $\mathcal{O}(1)$.
- <b>Espace :</b> $\mathcal{O}(M)$ pour stocker le tableau de booléens du crible.

## Problème 4.2 : Exponentiation II

**Lien vers le problème :** [CSES - Exponentiation II](https://cses.fi/problemset/task/1712)

**Énoncé :**
On vous donne $N$ requêtes. Chaque requête consiste en trois entiers $A, B$, et $C$. Vous devez calculer $A^{B^C} \pmod{10^9+7}$.

**Entrée :**
- La première ligne contient $N$.
- Les $N$ lignes suivantes contiennent $A, B$, et $C$.

**Sortie :**
- Affichez le résultat pour chaque requête modulo $10^9+7$.

**Contraintes :**
- $1 \le N \le 10^5$
- $0 \le A, B, C \le 10^9$

### Indication 1

Attention à la non-associativité de l'opérateur modulo sur les exposants : $A^{B^C} \pmod M \neq A^{(B^C \pmod M)} \pmod M$.

### Indication 2

Le modulo $M = 10^9+7$ est un nombre premier. Utilisez le Petit Théorème de Fermat qui stipule que $A^{M-1} \equiv 1 \pmod M$. Par conséquent, les exposants "bouclent" avec une période de $M-1$.

### Correction théorique

<b>Paradigme :</b> Arithmétique Modulaire et Exponentiation Rapide (Binary Exponentiation).

<b>Processus de réflexion :</b>
On doit calculer $A^{B^C} \pmod M$ avec $M = 10^9+7$. Le calcul direct de $B^C$ causera un dépassement de mémoire (overflow). On utilise le corollaire du Petit Théorème de Fermat : si $A$ n'est pas divisible par le nombre premier $M$, alors $A^K \equiv A^{K \pmod{M-1}} \pmod M$. 
L'algorithme se divise en deux étapes :
1. Calculer $K = B^C \pmod{M-1}$ en utilisant l'exponentiation binaire.
2. Calculer $A^K \pmod M$ de la même manière.
Attention au cas particulier où $A$ est un multiple de $M$ : le résultat est $0$.

<b>Preuve de justesse :</b>
Par le Petit Théorème de Fermat, si $\gcd(A, M) = 1$, alors $A^{M-1} \equiv 1 \pmod M$. L'exposant $K$ peut être écrit sous la forme $K = q(M-1) + r$ via la division euclidienne, avec $r = K \pmod{M-1}$.
Ainsi, $A^K = A^{q(M-1) + r} = (A^{M-1})^q \times A^r \equiv 1^q \times A^r \equiv A^r \pmod M$.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}(\log C + \log K) = \mathcal{O}(\log M)$ par requête. L'exponentiation binaire résout $x^y$ en un nombre d'itérations proportionnel au nombre de bits de $y$.
- <b>Espace :</b> $\mathcal{O}(1)$. Les appels récursifs peuvent être remplacés par une boucle itérative.

## Problème 4.3 : The Football Season

**Lien vers le problème :** [Codeforces - 1244C](https://codeforces.com/problemset/problem/1244/C)

**Énoncé :**
Une équipe de football a joué $N$ matchs et a obtenu exactement $P$ points. Une victoire rapporte $W$ points, un match nul rapporte $D$ points, et une défaite rapporte $0$ point. 
Trouvez un triplet $(x, y, z)$ (victoires, nuls, défaites) tel que : 
$x \cdot W + y \cdot D = P$
$x + y + z = N$
Si plusieurs solutions existent, affichez-en une. S'il n'y en a aucune, affichez `-1`.

**Entrée :**
- Quatre entiers : $N, P, W, D$.

**Sortie :**
- Trois entiers $x, y, z$ ou `-1`.

**Contraintes :**
- $1 \le N \le 10^{12}$
- $0 \le P \le 10^{17}$
- $1 \le D < W \le 10^5$

### Indication 1

C'est un système d'équations. Puisque $z$ est contraint par $N - (x + y)$ et n'influence pas le score, le problème se réduit à trouver une solution à l'équation diophantienne linéaire $xW + yD = P$ avec $x, y \ge 0$ et $x + y \le N$.

### Indication 2

Il existe une infinité de solutions à $xW + yD = P$. Remarquez que si l'on a $W$ matchs nuls, on obtient $W \times D$ points. C'est le même nombre de points que si l'on avait $D$ victoires ($D \times W$). On peut donc toujours échanger $W$ nuls contre $D$ victoires pour minimiser le nombre total de matchs non-perdus. Cela implique que si une solution existe, il en existe obligatoirement une où le nombre de nuls $y$ est strictement inférieur à $W$.

### Correction théorique

<b>Paradigme :</b> Borne Supérieure Algébrique / Recherche Exhaustive bornée.

<b>Processus de réflexion :</b>
On cherche $xW + yD = P$. Au lieu d'implémenter l'Algorithme d'Euclide Étendu (qui donne n'importe quelle solution), on exploite les contraintes du problème. On itère sur toutes les valeurs possibles de $y$ (le nombre de matchs nuls). L'observation clé est qu'il suffit de tester $y \in [0, W-1]$. Pour chaque $y$, on calcule $R = P - yD$. Si $R \ge 0$ et $R \equiv 0 \pmod W$, alors on a trouvé notre nombre de victoires $x = R / W$. Il ne reste plus qu'à vérifier que $x + y \le N$. Si c'est le cas, on pose $z = N - x - y$ et on retourne la solution.

<b>Preuve de justesse :</b>
Supposons qu'il existe une solution valide $(x', y')$ avec $y' \ge W$. On peut réécrire cette solution en remarquant que $W$ matchs nuls rapportent $WD$ points, tout comme $D$ victoires. On peut donc transformer la solution en $(x' + D, y' - W)$ sans altérer le score $P$. Le nombre total de matchs joués $(x' + D) + (y' - W) = x' + y' - (W - D)$. Puisque $W > D$, le nombre total de matchs requis a strictement diminué, respectant d'autant plus la contrainte $N$. Par descente infinie, on arrivera toujours à un $y < W$. Tester cet intervalle garantit l'exhaustivité.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}(W)$ au pire. Puisque $W \le 10^5$, cela nécessite au maximum $10^5$ itérations, ce qui passe instantanément.
- <b>Espace :</b> $\mathcal{O}(1)$ mémoire constante.

## Problème 4.4 : Beautiful Numbers

**Lien vers le problème :** [Codeforces - 300C](https://codeforces.com/problemset/problem/300/C)

**Énoncé :**
Deux chiffres $A$ et $B$ sont considérés comme "excellents". Un nombre est "excellent" s'il est uniquement composé des chiffres $A$ et $B$. Un nombre est "magnifique" (beautiful) s'il est de longueur $N$ (ne contenant que $A$ et $B$) et que la **somme** de ses $N$ chiffres est un nombre "excellent".
Comptez le nombre de nombres magnifiques modulo $10^9+7$.

**Entrée :**
- Trois entiers : $A, B, N$.

**Sortie :**
- Un entier : la réponse modulo $10^9+7$.

**Contraintes :**
- $1 \le A < B \le 9$
- $1 \le N \le 10^6$

### Indication 1

Un nombre magnifique de longueur $N$ est composé de $i$ chiffres $A$ et de $(N-i)$ chiffres $B$. Itérez sur $i$ de $0$ à $N$. Si la somme totale pour un certain $i$ est un nombre excellent, alors toutes les permutations de ces $N$ chiffres forment des nombres magnifiques distincts.

### Indication 2

Le nombre de permutations est donné par le coefficient binomial $\binom{N}{i}$. Pour calculer $\binom{N}{i} \pmod M$, vous devez utiliser la division modulaire : $\binom{N}{i} = N! \cdot (i!)^{-1} \cdot ((N-i)!)^{-1} \pmod M$, en précalculant les factorielles et leurs inverses modulaires via l'exponentiation rapide.

### Correction théorique

<b>Paradigme :</b> Dénombrement et Inverses Modulaires.

<b>Processus de réflexion :</b>
On précalcule toutes les factorielles de $0$ à $N$ modulo $10^9+7$. 
Pour chaque configuration $i$ (nombre de chiffres $A$), le nombre de chiffres $B$ est $N-i$. La somme de cette configuration est $S_i = i \cdot A + (N-i) \cdot B$.
On extrait les chiffres de $S_i$ pour vérifier s'il est "excellent" (composé uniquement de $A$ et $B$). Si c'est le cas, on ajoute le coefficient binomial $\binom{N}{i}$ au total. 
Le calcul de ce coefficient se fait en temps constant en multipliant $N!$ par l'inverse modulaire de $i!$ et $(N-i)!$. L'inverse de $X$ modulo un premier $M$ est $X^{M-2} \pmod M$ (Corollaire de Fermat).

<b>Preuve de justesse :</b>
L'unicité des nombres comptés est garantie par la bijection entre une combinaison d'indices (les emplacements des $i$ chiffres $A$ dans la chaîne de longueur $N$) et le nombre lui-même. La somme des chiffres est invariante par permutation. Le calcul algébrique dans le corps $\mathbb{Z}/p\mathbb{Z}$ garantit l'exactitude de la division grâce à la bijectivité de la fonction $x \mapsto x^{p-2} \pmod p$ pour tout $x \not\equiv 0$.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}(N \log M)$. Le précalcul prend $\mathcal{O}(N)$. La boucle principale fait $N$ itérations. L'inversion modulaire coûte $\mathcal{O}(\log M)$, mais peut être optimisée en $\mathcal{O}(1)$ si on précalcule les inverses des factorielles à l'envers. La vérification de la somme coûte $\mathcal{O}(\log_{10}(S_i))$. Temps total dominé par $N$.
- <b>Espace :</b> $\mathcal{O}(N)$ pour les tableaux de factorielles.

## Problème 4.5 : Same GCDs

**Lien vers le problème :** [Codeforces - 1295D](https://codeforces.com/problemset/problem/1295/D)

**Énoncé :**
Étant donné deux entiers $A$ et $M$, trouvez le nombre d'entiers $x$ tels que $0 \le x < M$ et $\gcd(A, M) = \gcd(A + x, M)$.
On vous donne $T$ cas de test.

**Entrée :**
- La première ligne contient un entier $T$ (nombre de requêtes).
- Les $T$ lignes suivantes contiennent $A$ et $M$.

**Sortie :**
- Pour chaque requête, affichez le nombre de $x$ valides.

**Contraintes :**
- $1 \le T \le 50$
- $1 \le A < M \le 10^{10}$

### Indication 1

L'étape clé de l'algorithme d'Euclide est $\gcd(U, V) = \gcd(U \pmod V, V)$. Par conséquent, au fur et à mesure que $x$ varie de $0$ à $M-1$, l'expression $(A + x)$ parcourt cycliquement tous les restes modulo $M$. L'ensemble des valeurs $\gcd(A + x, M)$ est exactement l'ensemble des valeurs $\gcd(Y, M)$ pour $Y \in [0, M-1]$.

### Indication 2

Soit $g = \gcd(A, M)$. On cherche le nombre de $Y < M$ tels que $\gcd(Y, M) = g$. En divisant l'équation par $g$, cela revient à trouver le nombre de $Y'$ tels que $\gcd(Y', M/g) = 1$, où $Y' = Y/g$. C'est exactement la définition de la Fonction Indicatrice (Totiente) d'Euler, notée $\varphi(M/g)$.

### Correction théorique

<b>Paradigme :</b> Propriétés du PGCD et Arithmétique Modulaire.

<b>Processus de réflexion :</b>
Le problème demande de compter le nombre d'entiers dans l'intervalle semi-ouvert $[A, A+M-1]$ copremiers avec $M$ une fois factorisés par $g$. 
Puisque l'intervalle est de taille exacte $M$, on peut décaler l'intervalle entier par symétrie modulo $M$. Le nombre d'éléments vérifiant $\gcd(X, M) = g$ dans $[A, A+M-1]$ est strictement identique au nombre d'éléments le vérifiant dans $[0, M-1]$. 
L'égalité $\gcd(X, M) = g$ équivaut à $\gcd(X/g, M/g) = 1$. Le nombre d'entiers strictement inférieurs à $M/g$ et copremiers avec $M/g$ est donné par $\varphi(M/g)$.
Pour calculer $\varphi(K)$ pour $K = M/g$, on factorise $K$ en nombres premiers. $\varphi(K) = K \prod_{p|K} (1 - \frac{1}{p})$.

<b>Preuve de justesse :</b>
La fonction Indicatrice d'Euler est prouvée par le Principe d'Inclusion-Exclusion. Si $K = p_1^{a_1} p_2^{a_2} \dots$, on retire de l'intervalle les multiples de $p_1$, puis on retire les multiples de $p_2$, et on rajoute les multiples de $(p_1 p_2)$ qui ont été soustraits deux fois, etc. Ce développement algébrique factorisé donne exactement la formule du produit de Fermat / Euler.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}(\sqrt{M})$ par requête. Le calcul de $g$ prend $\mathcal{O}(\log M)$. La factorisation en nombres premiers de $M/g$ demande d'itérer jusqu'à sa racine carrée.
- <b>Espace :</b> $\mathcal{O}(1)$ mémoire constante.

# Section 5 : Algorithmique du Texte (String Processing)

## Problème 5.1 : String Matching

**Lien vers le problème :** [CSES - String Matching](https://cses.fi/problemset/task/1753)

**Énoncé :**
Étant donné une chaîne de caractères de longueur $N$ et un motif de longueur $M$, comptez le nombre d'occurrences du motif dans la chaîne.

**Entrée :**
- La première ligne contient la chaîne de caractères.
- La deuxième ligne contient le motif.

**Sortie :**
- Un seul entier : le nombre d'occurrences.

**Contraintes :**
- $1 \le N, M \le 10^6$

### Indication 1

Une recherche naïve fenêtre par fenêtre vous coûtera $\mathcal{O}(N \times M)$ opérations, ce qui résultera en un Time Limit Exceeded. Il faut un algorithme linéaire.

### Indication 2

Deux approches optimales : l'algorithme de Knuth-Morris-Pratt (KMP) qui utilise un tableau de préfixes pour éviter les retours en arrière, ou le hachage polynomial (Rabin-Karp) qui compare les empreintes numériques des sous-chaînes en $\mathcal{O}(1)$ grâce aux sommes préfixes.

### Correction théorique

<b>Paradigme :</b> Automate de reconnaissance (KMP) ou Hachage Roulant (Rabin-Karp).

<b>Processus de réflexion (KMP) :</b>
On calcule d'abord le tableau $\pi$ (fonction préfixe) pour le motif. $\pi[i]$ stocke la longueur du plus long bord (un préfixe qui est aussi un suffixe) du sous-motif se terminant à l'indice $i$. Lors du parcours de la chaîne principale, si une divergence survient après avoir validé $k$ caractères, au lieu de tout recommencer, on recule notre pointeur de motif à l'indice $\pi[k-1]$. Cela garantit qu'on ne re-teste pas des caractères déjà validés.

<b>Preuve de justesse :</b>
La fonction préfixe assure l'invariant de boucle suivant : à tout instant, la valeur de notre pointeur de motif correspond exactement à la longueur du plus long préfixe du motif qui matche le suffixe de la chaîne lue jusqu'à présent. L'exhaustivité est garantie car la transition d'échec (fallback) via le tableau $\pi$ teste strictement tous les préfixes valides potentiels par ordre décroissant de longueur.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}(N + M)$. Chaque caractère de la chaîne et du motif est lu une fois. Le pointeur de motif ne peut être décrémenté plus de fois qu'il n'a été incrémenté.
- <b>Espace :</b> $\mathcal{O}(M)$ pour stocker le tableau préfixe du motif.

## Problème 5.2 : Password

**Lien vers le problème :** [Codeforces - 126B](https://codeforces.com/problemset/problem/126/B)

**Énoncé :**
Étant donné une chaîne $S$ de longueur $N$, trouvez la plus longue sous-chaîne $T$ qui est à la fois :
1. Un préfixe de $S$
2. Un suffixe de $S$
3. Présente au moins une fois à l'intérieur de $S$ (ni en tant que préfixe, ni en tant que suffixe).
Si aucune telle chaîne n'existe, affichez `Just a legend`.

**Entrée :**
- Une seule ligne contenant la chaîne $S$.

**Sortie :**
- La chaîne $T$ correspondante, ou le message d'erreur.

**Contraintes :**
- $1 \le N \le 10^6$

### Indication 1

Si $T$ est un préfixe et un suffixe de la chaîne complète, cela correspond exactement à la définition d'un "bord" (border) de la chaîne complète. Que représente alors la dernière case du tableau préfixe $\pi[N-1]$ de l'algorithme KMP ?

### Indication 2

Calculez le tableau $\pi$ pour toute la chaîne $S$. La longueur du plus long préfixe-suffixe est $L = \pi[N-1]$. 
Si cette longueur $L$ apparaît ailleurs dans le tableau $\pi$ (pour un indice $i < N-1$), alors c'est la solution. Sinon, la solution pourrait être le "bord du bord", c'est-à-dire $\pi[L-1]$.

### Correction théorique

<b>Paradigme :</b> Analyse structurelle de la fonction préfixe (KMP).

<b>Processus de réflexion :</b>
Soit $\pi$ le tableau des préfixes de $S$. $L = \pi[N-1]$ est la longueur du plus long préfixe qui est aussi suffixe. 
Ce préfixe $S[0 \dots L-1]$ existe-t-il au milieu de $S$ ? Il suffit de chercher si la valeur $L$ est présente dans le tableau $\pi$ à un indice $i \in [1, N-2]$. Si c'est le cas, c'est gagné. 
Si ce n'est pas le cas, le candidat suivant n'est pas $L-1$, mais le plus long bord de ce préfixe, qui est donné par $L' = \pi[L-1]$. Puisque $L'$ est un bord du suffixe terminal, il est garanti d'exister à l'intérieur (car il existe au début de $S$ et à la fin de $S$, et par définition de $L$, il fait partie du préfixe initial qui se trouve être confiné avant la fin de $S$). Si $L' > 0$, la réponse est $L'$.

<b>Preuve de justesse :</b>
Par l'absurde, supposons qu'il existe un candidat optimal $T$ de longueur $k$ qui ne soit pas dans la séquence des bords itérés de $S$. Puisque $T$ est préfixe et suffixe de $S$, c'est par définition un bord de $S$. Le lemme fondamental des algorithmes de type KMP stipule que l'ensemble de tous les bords d'une chaîne peut être généré exhaustivement par l'application récursive de la fonction préfixe : $\pi[N-1], \pi[\pi[N-1]-1], \dots$ jusqu'à atteindre $0$. La recherche sur ces candidats spécifiques est donc mathématiquement exhaustive.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}(N)$. Le calcul du tableau $\pi$ se fait en temps linéaire. La vérification ultérieure nécessite un parcours de $\mathcal{O}(N)$ itérations ou une simple extraction en $\mathcal{O}(1)$.
- <b>Espace :</b> $\mathcal{O}(N)$ pour allouer le tableau $\pi$ des longueurs.

## Problème 5.3 : Sereja and Brackets

**Lien vers le problème :** [Codeforces - 380C](https://codeforces.com/problemset/problem/380/C)

**Énoncé :**
On vous donne une séquence de parenthèses de longueur $N$. Vous devez répondre à $Q$ requêtes. Chaque requête est définie par un intervalle $[L, R]$. Vous devez trouver la longueur de la plus longue sous-séquence de parenthèses correcte (bien parenthésée) qui peut être formée en supprimant zéro ou plusieurs caractères de la sous-chaîne $S[L \dots R]$.

**Entrée :**
- Une chaîne de parenthèses `(` et `)`.
- Un entier $Q$.
- $Q$ lignes contenant $L$ et $R$.

**Sortie :**
- Pour chaque requête, la longueur maximale trouvée.

**Contraintes :**
- $1 \le N, Q \le 10^6$

### Indication 1

Un parcours linéaire avec une pile (stack) prend $\mathcal{O}(N)$ par requête, ce qui provoquera un TLE. Le problème réclame des requêtes sur intervalles, c'est le signal d'alarme pour utiliser un Arbre de Segments (Segment Tree).

### Indication 2

Que doit stocker un nœud de l'arbre pour fusionner deux sous-intervalles ? Connaître uniquement le nombre de paires valides dans le fils gauche et le fils droit ne suffit pas. 
Un nœud doit stocker un triplet : `(paires_valides, ouvrantes_non_apparriees, fermantes_non_apparriees)`.

### Correction théorique

<b>Paradigme :</b> Structure de Données (Arbre de Segments avec fonction de fusion algébrique).

<b>Processus de réflexion :</b>
On construit un Segment Tree. Chaque feuille représentant le caractère `(` aura pour état `(0, 1, 0)`. Une feuille `)` aura `(0, 0, 1)`.
Lors de la fusion d'un nœud gauche $G$ et d'un nœud droit $D$ pour former le parent $P$, de nouvelles paires valides se forment en combinant les parenthèses ouvrantes non-appariées de $G$ avec les parenthèses fermantes non-appariées de $D$.
Le nombre de nouvelles paires est $k = \min(G.ouvrantes, D.fermantes)$.
L'état de $P$ devient :
- $P.paires = G.paires + D.paires + k$
- $P.ouvrantes = G.ouvrantes + D.ouvrantes - k$
- $P.fermantes = G.fermantes + D.fermantes - k$
Le résultat final pour une requête est $2 \times P.paires$.

<b>Preuve de justesse :</b>
L'opérateur de fusion est formellement associatif. Le parenthésage valide respecte une structure de langage hors-contexte (grammaire de Dyck). Toute paire de parenthèses valide qui chevauche la frontière entre le segment gauche et le segment droit connecte nécessairement une ouvrante "libre" située à gauche et une fermante "libre" située à droite. L'opération $\min$ capture la borne stricte de cette résolution de liaison. Par associativité, le partitionnement binaire de l'arbre conserve la justesse globale sur tout intervalle arbitraire.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}(N)$ pour la construction ascendante de l'arbre. $\mathcal{O}(\log N)$ par requête pour descendre et fusionner $\mathcal{O}(\log N)$ nœuds. Temps total $\mathcal{O}(N + Q \log N)$.
- <b>Espace :</b> $\mathcal{O}(N)$ strictement borné pour l'allocation des $4N$ nœuds de l'arbre de segments.

## Problème 5.4 : Word Combinations

**Lien vers le problème :** [CSES - Word Combinations](https://cses.fi/problemset/task/1731)

**Énoncé :**
On vous donne une chaîne de caractères de longueur $N$ et un dictionnaire contenant $K$ mots. 
Calculez le nombre de façons de former la chaîne donnée en concaténant des mots du dictionnaire. Le résultat doit être modulo $10^9+7$.

**Entrée :**
- La chaîne de caractères de longueur $N$.
- Un entier $K$.
- $K$ lignes contenant les mots du dictionnaire.

**Sortie :**
- Le nombre de combinaisons modulo $10^9+7$.

**Contraintes :**
- $1 \le N \le 5000$
- $1 \le K \le 10^5$
- La longueur totale de tous les mots du dictionnaire ne dépasse pas $10^6$.

### Indication 1

Soit `dp[i]` le nombre de façons de former le suffixe de la chaîne à partir de l'indice $i$. Pour calculer `dp[i]`, on doit vérifier tous les mots du dictionnaire qui correspondent à un préfixe de la sous-chaîne $S[i \dots N-1]$. Une comparaison naïve mot par mot entraînera un Time Limit Exceeded.

### Indication 2

Stockez le dictionnaire dans un Arbre Préfixe (Trie). À partir de chaque indice $i$, descendez dans le Trie en lisant les caractères $S[i], S[i+1], \dots$. Chaque fois que vous rencontrez un nœud terminal (fin de mot) dans le Trie à la profondeur $L$, vous pouvez ajouter la valeur de `dp[i + L]` à `dp[i]`.

### Correction théorique

<b>Paradigme :</b> Structure de Données (Trie) couplée à la Programmation Dynamique.

<b>Processus de réflexion :</b>
L'approche naïve en vérifiant l'égalité des sous-chaînes via hachage prendrait un temps potentiellement quadratique dominé par des requêtes inutiles. Le Trie permet un élagage structurel. 
On construit un automate fini déterministe (le Trie) à partir du dictionnaire. L'état DP se calcule de droite à gauche (de l'indice $N$ à $0$). `dp[N] = 1`. 
Pour un indice $i$, on initialise un pointeur à la racine du Trie. On lit séquentiellement $S[i+j]$ pour $j \ge 0$. Si l'arête correspondante n'existe pas dans le Trie, on coupe immédiatement la recherche (élagage). Si elle existe et qu'on atteint un nœud terminal, on exécute la transition : `dp[i] = (dp[i] + dp[i + j + 1]) % MOD`.

<b>Preuve de justesse :</b>
L'espace des solutions est partitionné par le premier mot choisi pour couvrir le préfixe de la sous-chaîne restante. Le Trie garantit que l'on n'explore que les mots mathématiquement existants dans le dictionnaire. La sous-structure optimale est stricte : le nombre de façons de construire le reste de la chaîne est indépendant des mots choisis précédemment.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}(\sum |W_k| + N^2)$. La construction du Trie prend un temps proportionnel à la somme des longueurs des mots. La DP effectue au pire $N$ itérations, et à chaque étape explore le Trie sur une profondeur maximale $N$. En pratique, l'élagage rend la boucle interne extrêmement courte.
- <b>Espace :</b> $\mathcal{O}(\sum |W_k|)$ pour l'allocation des nœuds du Trie et $\mathcal{O}(N)$ pour le tableau DP.

## Problème 5.5 : Prefixes and Suffixes

**Lien vers le problème :** [Codeforces - 432D](https://codeforces.com/problemset/problem/432/D)

**Énoncé :**
On vous donne une chaîne $S$ de longueur $N$. Pour chaque sous-chaîne qui est à la fois un préfixe et un suffixe de $S$, vous devez afficher sa longueur et le nombre exact de fois où elle apparaît dans $S$ en tant que sous-chaîne continue.

**Entrée :**
- Une seule ligne contenant la chaîne $S$.

**Sortie :**
- L'entier $K$ (le nombre de sous-chaînes valides).
- $K$ lignes contenant la longueur et la fréquence, triées par ordre croissant de longueur.

**Contraintes :**
- $1 \le N \le 10^5$

### Indication 1

Comme vu précédemment, identifier les préfixes qui sont aussi des suffixes se fait directement en inspectant les transitions de retour (fallback) du tableau $\pi$ (KMP) ou du tableau $Z$ à la fin de la chaîne.

### Indication 2

Pour compter la fréquence de chaque préfixe de longueur $L$ dans toute la chaîne, utilisez un tableau de comptage `count` initialisé à $1$ pour chaque valeur du tableau $\pi$ trouvée. Puis, propagez ces comptes de la fin vers le début du tableau $\pi$ : si un préfixe de longueur $i$ apparaît, il implique mathématiquement l'apparition de son propre plus long préfixe-suffixe (de longueur $\pi[i-1]$).

### Correction théorique

<b>Paradigme :</b> Propagation de fréquence sur l'Arbre des Bords (KMP).

<b>Processus de réflexion :</b>
1. Calculer le tableau $\pi$ classique en $\mathcal{O}(N)$.
2. Initialiser un tableau `dp` de taille $N+1$ à $1$ (chaque préfixe correspond au moins à lui-même lors de sa première lecture).
3. Pour $i$ allant de $N-1$ à $1$ en descendant : la valeur de la fonction préfixe est $p = \pi[i]$. Cela signifie qu'un préfixe de longueur $p$ est trouvé à cet endroit. On incrémente donc le compte de ce préfixe court : `dp[p] += dp[i+1]`.
4. Extraire les solutions valides : on commence avec la chaîne entière (longueur $N$). Ensuite, on remonte les bords terminaux $L_{k+1} = \pi[L_k - 1]$ jusqu'à $0$. Les longueurs de ces bords constituent l'ensemble des préfixes-suffixes valides.

<b>Preuve de justesse :</b>
Le tableau $\pi$ encode implicitement un arbre (l'Arbre des Bords) où le parent du nœud $i$ est $\pi[i-1]$. Apparaître avec une longueur $L$ implique la présence simultanée de tous les bords de $L$ à ce même indice d'arrivée. La propagation dynamique de l'indice $N$ vers $0$ correspond formellement à une somme récursive des tailles des sous-arbres dans cet Arbre des Bords. Cette bijection assure le décompte total exact de toutes les occurrences.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}(N)$. Le calcul de KMP est linéaire, la passe de programmation dynamique en ordre décroissant est linéaire, et la remontée des bords l'est également.
- <b>Espace :</b> $\mathcal{O}(N)$ pour le tableau $\pi$ et le tableau des fréquences `dp`.

## Problème 6.1 : Point Location Test

**Lien vers le problème :** [CSES - Point Location Test](https://cses.fi/problemset/task/2189)

**Énoncé :**
Étant donné une ligne définie par deux points $P_1$ et $P_2$, et un point $P_3$, déterminez si le point $P_3$ se trouve à gauche de la ligne, à droite de la ligne, ou s'il est colinéaire. La direction est de $P_1$ vers $P_2$.

**Entrée :**
- Le nombre de requêtes $T$.
- $T$ lignes contenant chacune six entiers $X_1, Y_1, X_2, Y_2, X_3, Y_3$.

**Sortie :**
- Pour chaque requête, affichez `LEFT`, `RIGHT`, ou `TOUCH`.

**Contraintes :**
- $1 \le T \le 10^5$
- $-10^9 \le X_i, Y_i \le 10^9$

### Indication 1

N'utilisez sous aucun prétexte des flottants, des divisions ou la fonction arc tangente (`atan2`). La précision des flottants IEEE 754 échouera sur de grandes coordonnées. Vous devez résoudre ce problème avec l'algèbre linéaire dans l'espace $\mathbb{Z}$.

### Indication 2

Formez les vecteurs $\vec{u} = \vec{P_1P_2}$ et $\vec{v} = \vec{P_1P_3}$. Le produit vectoriel 2D $\vec{u} \times \vec{v}$ (soit le déterminant de la matrice $2 \times 2$ formée par ces vecteurs) donne l'orientation géométrique. Étudiez le signe du résultat.

### Correction théorique

<b>Paradigme :</b> Produit Vectoriel 2D (Déterminant).

<b>Processus de réflexion :</b>
Soit le vecteur de référence $\vec{u} = (x_2 - x_1, y_2 - y_1)$ et le vecteur cible $\vec{v} = (x_3 - x_1, y_3 - y_1)$. 
Le produit vectoriel dans le plan est défini par : $C = u_x \cdot v_y - u_y \cdot v_x$.
Si $C > 0$, le vecteur $\vec{v}$ forme un angle orienté positif avec $\vec{u}$ (sens trigonométrique), donc le point est à gauche (`LEFT`).
Si $C < 0$, l'angle est négatif (sens horaire), donc le point est à droite (`RIGHT`).
Si $C = 0$, l'aire du parallélogramme sous-tendu est nulle, donc les points sont colinéaires (`TOUCH`).

<b>Preuve de justesse :</b>
Le produit vectoriel $\vec{u} \times \vec{v}$ équivaut au déterminant de la base $(\vec{u}, \vec{v})$. Un déterminant strictement positif signifie que la base conserve l'orientation canonique du plan euclidien (donc un virage à gauche). Comme toutes les opérations se limitent à la multiplication et la soustraction sur des entiers initiaux bornés à $10^9$, le calcul intermédiaire nécessite exactement un type entier de 64 bits (jusqu'à $2 \cdot 10^{18}$), garantissant une précision absolue sans aucun biais de troncature.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}(1)$ par requête.
- <b>Espace :</b> $\mathcal{O}(1)$ mémoire constante.

## Problème 6.2 : Polygon Area

**Lien vers le problème :** [CSES - Polygon Area](https://cses.fi/problemset/task/2191)

**Énoncé :**
Calculez l'aire exacte d'un polygone donné (qui peut être non convexe) représenté par ses $N$ sommets donnés dans le sens trigonométrique ou anti-trigonométrique.
Comme l'aire peut ne pas être entière, affichez le résultat sous la forme $2 \times \text{Aire}$ (qui est garanti d'être un entier).

**Entrée :**
- Le nombre de sommets $N$.
- $N$ lignes contenant les coordonnées $X_i, Y_i$.

**Sortie :**
- L'entier représentant $2 \times \text{Aire}$.

**Contraintes :**
- $3 \le N \le 1000$
- $-10^9 \le X_i, Y_i \le 10^9$

### Indication 1

Trianguler un polygone potentiellement concave est un algorithme complexe et lourd en $\mathcal{O}(N^3)$ naïf. Pensez plutôt à intégrer le contour du polygone par rapport à l'origine $(0,0)$.

### Indication 2

Utilisez la Formule de l'Arpenteur (Shoelace Formula) qui dérive du calcul vectoriel. La somme algébrique des produits vectoriels successifs des sommets adjacents par rapport à l'origine donnera directement $2 \times \text{Aire}$.

### Correction théorique

<b>Paradigme :</b> Formule de l'Arpenteur (Shoelace Algorithm).

<b>Processus de réflexion :</b>
On itère sur les arêtes formées par $P_i$ et $P_{i+1}$ (en fermant la boucle avec $P_{N}$ et $P_1$). Pour chaque arête, on calcule le produit vectoriel $\vec{OP_i} \times \vec{OP_{i+1}} = x_i \cdot y_{i+1} - x_{i+1} \cdot y_i$. On accumule ces valeurs.
Si les sommets sont dans le sens horaire, la somme sera négative. Dans tous les cas, $2 \times \text{Aire}$ est la valeur absolue de cette somme globale. Tout se fait en entiers 64 bits.

<b>Preuve de justesse :</b>
Cette formule est la traduction discrète du Théorème de Green-Riemann. L'aire délimitée par une courbe fermée simple $\mathcal{C}$ s'écrit $\frac{1}{2} \oint_{\mathcal{C}} (x\,dy - y\,dx)$. Sur un segment de droite paramétré entre $P_i$ et $P_{i+1}$, cette intégrale linéaire se simplifie exactement en $\frac{1}{2} (x_i y_{i+1} - x_{i+1} y_i)$. L'additivité des intégrales sur les contours orientés prouve que les aires formées avec l'origine à l'extérieur du polygone s'annulent algébriquement, laissant invariante la surface interne fermée.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}(N)$ où chaque arête est évaluée une seule fois.
- <b>Espace :</b> $\mathcal{O}(N)$ ou $\mathcal{O}(1)$ si les calculs sont accumulés en flux de lecture (stream) sans stocker les tableaux.

## Problème 6.3 : Convex Hull

**Lien vers le problème :** [CSES - Convex Hull](https://cses.fi/problemset/task/2195)

**Énoncé :**
Étant donné un ensemble de $N$ points dans le plan, identifiez l'enveloppe convexe (Convex Hull) de ces points, c'est-à-dire le plus petit polygone convexe qui contient tous les points de l'ensemble.
Affichez le nombre de sommets de cette enveloppe, puis leurs coordonnées.

**Entrée :**
- $N$ (le nombre de points).
- $N$ lignes de coordonnées $X_i, Y_i$.

**Sortie :**
- Affichez l'enveloppe convexe.

**Contraintes :**
- $3 \le N \le 2 \cdot 10^5$

### Indication 1

Évitez les algorithmes en $\mathcal{O}(N^2)$ comme le QuickHull non optimal ou Jarvis March. Les contraintes $2 \cdot 10^5$ nécessitent un algorithme basé sur le tri en $\mathcal{O}(N \log N)$.

### Indication 2

Privilégiez l'algorithme de la chaîne monotone d'Andrew (Monotone Chain), une variante robuste du Graham Scan. 
1. Triez les points par abscisse $X$ puis ordonnée $Y$.
2. Construisez l'enveloppe inférieure (lower hull) puis supérieure (upper hull) en utilisant une pile (stack) et le produit vectoriel pour éliminer les virages concaves (à droite).

### Correction théorique

<b>Paradigme :</b> Tri Lexicographique et Pile Monotone (Monotone Chain / Algorithme d'Andrew).

<b>Processus de réflexion :</b>
On trie tous les points de gauche à droite. On maintient une pile `hull`. Pour construire l'enveloppe inférieure, on itère sur les points. Avant d'ajouter un point $P_i$, on vérifie l'orientation des 3 derniers points : les deux derniers de la pile $A$ et $B$, et la cible $P_i$. Le produit vectoriel $\vec{AB} \times \vec{BP_i}$ doit garantir un virage à gauche ou une ligne droite (selon si l'énoncé tolère les points colinéaires). Si l'orientation est un virage à droite (concave), on dépile $B$ (Pop) et on re-vérifie. On répète le processus en ordre inverse pour l'enveloppe supérieure. On concatène les deux moitiés.

<b>Preuve de justesse :</b>
Le tri lexicographique garantit que la progression balaie l'espace de manière séquentielle, réduisant la construction de l'enveloppe 2D à deux problèmes 1D de maintien de pente monotone. L'invariant de boucle de la pile exige que toute séquence de 3 sommets constitutifs maintienne une courbure convexe stricte (déterminant $\ge 0$). Le processus d'éviction par dépilage corrige de manière gloutonne toute violation locale, convergeant vers la forme convexe minimale englobante globale.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}(N \log N)$. Dominé par le tri initial. La construction avec la pile est linéaire en $\mathcal{O}(N)$ car chaque point est empilé et dépilé au maximum une fois (Analyse Amortie).
- <b>Espace :</b> $\mathcal{O}(N)$ pour allouer le tableau trié et la pile des sommets.

## Problème 6.4 : Vanya and Triangles

**Lien vers le problème :** [Codeforces - 552D](https://codeforces.com/problemset/problem/552/D)

**Énoncé :**
Vanya a un ensemble de $N$ points distincts dans le plan cartésien. Il veut savoir combien de triangles stricts d'aire non nulle il peut former en choisissant 3 points parmi les $N$.

**Entrée :**
- Un entier $N$.
- $N$ lignes contenant les coordonnées $X_i, Y_i$.

**Sortie :**
- Le nombre de triangles d'aire non nulle.

**Contraintes :**
- $1 \le N \le 2000$
- $-100 \le X_i, Y_i \le 100$

### Indication 1

Le nombre total de triplets est $\binom{N}{3}$. Un triplet échoue à former un triangle si et seulement si les 3 points sont alignés (colinéaires). Le problème revient à compter le nombre de triplets colinéaires et à les soustraire du total.

### Indication 2

Une approche en $\mathcal{O}(N^3)$ (vérifier l'aire de chaque triplet via le produit vectoriel) passera tout juste ou fera un TLE en Python selon l'implémentation. 
Optimisation $\mathcal{O}(N^2 \log N)$ : Pour un point $A$ fixé, calculez la pente réduite (fraction irréductible $\frac{\Delta y}{\Delta x}$) vers tous les autres points $B$. Groupez les points par pente avec un dictionnaire/Hashmap. Si $k$ points partagent la même pente depuis $A$, cela ajoute $\binom{k}{2}$ triplets colinéaires. Divisez le total final par 3 (pour éviter les doublons).

### Correction théorique

<b>Paradigme :</b> Dénombrement Complémentaire et Hachage de Fractions.

<b>Processus de réflexion :</b>
Trois points $A, B, C$ sont colinéaires s'ils appartiennent à la même droite. 
1. Fixer un point $P_i$ comme pivot.
2. Pour tout $j > i$, calculer le vecteur directeur $\vec{v} = (X_j - X_i, Y_j - Y_i)$. 
3. Normaliser ce vecteur en le divisant par son PGCD $g = \gcd(|X_j - X_i|, |Y_j - Y_i|)$ et imposer une direction canonique (par exemple, forcer $X > 0$, ou si $X=0$, forcer $Y > 0$). Cela encode la pente sous forme de tuple immuable $(dx', dy')$.
4. Compter le nombre de points atterrissant dans le même bac (bucket) de pente via un Dictionnaire.
5. Si $k$ points ont la même pente depuis $P_i$, ils forment $\binom{k}{2}$ triplets dégénérés avec le point pivot $P_i$. On soustrait cette somme de $\binom{N}{3}$.

<b>Preuve de justesse :</b>
Soit l'ensemble global $E$ de cardinal $\binom{N}{3}$. La condition de colinéarité pour 3 points implique le partage absolu d'une même pente orientée à partir de n'importe quel point du sous-ensemble. En forçant $j > i$, on s'assure qu'un segment n'est généré qu'une seule fois en tant qu'arrête fondatrice du groupe. L'utilisation d'une fraction rationnelle normalisée par PGCD contourne rigoureusement l'instabilité catastrophique des flottants (`y/x`) et prévient les erreurs de clés dans la table de hachage.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}(N^2 \log(\max(|dx|, |dy|)))$. Les boucles imbriquées prennent $\mathcal{O}(N^2)$. Le calcul du PGCD d'Euclide ajoute un facteur logarithmique mineur sur les coordonnées.
- <b>Espace :</b> $\mathcal{O}(N)$ pour allouer le dictionnaire des pentes à chaque itération du pivot.

## Problème 6.5 : Polygons

**Lien vers le problème :** [Codeforces - 166B](https://codeforces.com/problemset/problem/166/B)

**Énoncé :**
Étant donnés deux polygones strictement convexes $A$ et $B$ dans le plan. Vous devez déterminer si le polygone $B$ est *strictement* situé à l'intérieur du polygone $A$. Les frontières ne doivent pas se toucher.

**Entrée :**
- Un entier $N$, suivi des $N$ sommets du polygone $A$ (dans le sens anti-horaire).
- Un entier $M$, suivi des $M$ sommets du polygone $B$ (dans le sens anti-horaire).

**Sortie :**
- Affichez `YES` si $B$ est strictement à l'intérieur de $A$, sinon `NO`.

**Contraintes :**
- $3 \le N, M \le 10^5$
- $-10^9 \le X_i, Y_i \le 10^9$

### Indication 1

Dire que $B$ est strictement inclus dans $A$ revient à dire que l'enveloppe convexe de l'union $(A \cup B)$ est exactement identique à $A$, et qu'aucun sommet de $B$ n'appartient au bord de cette nouvelle enveloppe.

### Indication 2

Fusionnez les ensembles de points. Les sommets de $A$ ont un tag `0`, ceux de $B$ ont un tag `1`. Exécutez l'algorithme de Convex Hull (Monotone Chain) sur ce super-ensemble. Si un seul point taggé `1` finit par appartenir à la frontière convexe de l'enveloppe englobante, alors $B$ n'est pas strictement contenu dans $A$.

### Correction théorique

<b>Paradigme :</b> Réduction à l'Enveloppe Convexe (Convex Hull Union).

<b>Processus de réflexion :</b>
Le problème d'appartenance de $M$ points dans un polygone convexe peut se résoudre en $\mathcal{O}(M \log N)$ via recherche dichotomique angulaire, mais c'est complexe à implémenter. 
Une solution plus élégante exploite la Convex Hull : on assemble les $N + M$ points dans un tableau unique. On y attache des identifiants (appartenance à $A$ ou $B$). On calcule l'enveloppe convexe de cet amas de points.
La seule façon pour qu'un point de $B$ puisse s'échapper ou toucher la frontière de $A$ est que ce point de $B$ devienne lui-même un sommet constitutif (ou un point colinéaire d'une arête) de l'enveloppe externe de l'amas global. En tolérant les points colinéaires lors de la phase de tri et de dépilage de l'enveloppe, si un point de l'enveloppe résultante appartient à $B$, on retourne `NO`. Sinon `YES`.

<b>Preuve de justesse :</b>
Par définition, un polygone est strictement convexe si l'intersection de tous les demi-plans définis par ses arêtes contient son aire en exclusivité. Si $B \subset A$ strictement, alors la distance signée de chaque point de $B$ aux hyperplans d'appui de $A$ est strictement positive vers l'intérieur. L'opérateur Enveloppe Convexe enveloppe formellement les extrémums géométriques du nuage de points. Un sommet intérieur ne peut mathématiquement pas participer à la frontière convexe.

<b>Complexité :</b>
- <b>Temps :</b> $\mathcal{O}((N+M) \log(N+M))$. Tri du super-ensemble puis construction linéaire de l'enveloppe via chaîne monotone.
- <b>Espace :</b> $\mathcal{O}(N+M)$ pour rassembler l'intégralité des points avec leurs étiquettes d'appartenance.
